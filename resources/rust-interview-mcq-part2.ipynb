### Q101. What happens when a variable goes out of scope in Rust?
- A) Nothing
- B) The garbage collector frees it
- C) `drop` is called automatically
- D) It becomes null

<details>
<summary>Show Answer</summary>

**Q101. C)** `drop` is called automatically. Rust uses RAII — when a value's owner goes out of scope, its `drop` method runs.

</details>
