# Structs

Structs let you create custom data types by grouping related values together. Think of them as blueprints: you define the shape once, then create many instances. They're the foundation for building meaningful abstractions in Rust.

## What Are Structs and Why Do They Matter?

Structs bundle data with clear, named fields. Instead of passing five separate parameters, you pass one struct. They make code more readable, self-documenting, and easier to extend.

## Named Structs

The most common kind: define fields with names and types, then create instances by assigning values to each field.

```rust
struct User {
    username: String,
    email: String,
    active: bool,
}

let user = User {
    username: String::from("alice"),
    email: String::from("alice@example.com"),
    active: true,
};
println!("User: {} ({})", user.username, user.email);
```

## Tuple Structs

Structs without named fields — just types. Useful when you want a distinct type but don't need field names. Access with `.0`, `.1`, etc.

```rust
struct Color(i32, i32, i32);
struct Point(i32, i32);

let red = Color(255, 0, 0);
let origin = Point(0, 0);
println!("Red: ({}, {}, {}), Origin: ({}, {})", red.0, red.1, red.2, origin.0, origin.1);
```

## Unit Structs

Structs with no fields. Useful as markers or for implementing traits without storing data.

```rust
struct AlwaysEqual;
let subject = AlwaysEqual;
// No data to access — it's a type-level marker
```

## Field Init Shorthand

When a struct field and a variable have the same name, you can omit the repetition.

```rust
struct Person {
    name: String,
    age: u32,
}

let name = String::from("Bob");
let age = 30;
let person = Person { name, age };  // shorthand for name: name, age: age
println!("{} is {} years old", person.name, person.age);
```

## Struct Update Syntax `..`

Create a new struct from an existing one, changing only some fields. Use `..other` to copy the rest.

```rust
struct Config {
    host: String,
    port: u16,
    debug: bool,
}

let default = Config {
    host: String::from("localhost"),
    port: 8080,
    debug: false,
};
let dev = Config {
    debug: true,
    ..default
};
println!("Dev config: {}:{}, debug={}", dev.host, dev.port, dev.debug);
```

## `impl` Blocks and Methods

Methods are functions defined on a type. Use `&self` for read-only, `&mut self` for mutation, `self` when you consume the value.

```rust
struct Rectangle {
    width: u32,
    height: u32,
}

impl Rectangle {
    fn area(&self) -> u32 {
        self.width * self.height
    }
    fn scale(&mut self, factor: u32) {
        self.width *= factor;
        self.height *= factor;
    }
}

let mut rect = Rectangle { width: 3, height: 4 };
println!("Area: {}", rect.area());
rect.scale(2);
println!("Scaled area: {}", rect.area());
```

## Associated Functions (Constructors)

Functions in `impl` without `self` are **associated functions**. Call with `Type::name()`. The convention `::new()` is the standard constructor.

```rust
struct Counter {
    value: u32,
}

impl Counter {
    fn new() -> Counter {
        Counter { value: 0 }
    }
    fn with_start(start: u32) -> Counter {
        Counter { value: start }
    }
}

let c = Counter::new();
let c2 = Counter::with_start(10);
println!("Counter::new() and Counter::with_start(10)");
```

## Derived Traits: `#[derive(Debug, Clone)]`

The compiler can auto-implement common traits. `Debug` lets you print with `{:?}`; `Clone` lets you duplicate values.

```rust
#[derive(Debug, Clone)]
struct Item {
    name: String,
    count: i32,
}

let item = Item { name: String::from("apple"), count: 5 };
println!("Item: {:?}", item);
let copy = item.clone();
println!("Clone: {:?}", copy);
```

## Visibility with `pub`

By default, struct fields are private. Use `pub` to make them (or the struct itself) visible outside the module.

```rust
pub struct PublicStruct {
    pub name: String,
    id: i32,  // private by default
}

let s = PublicStruct { name: String::from("test"), id: 1 };
println!("name (public): {}", s.name);
```

## Structs vs Tuples

Use **structs** when you have named, meaningful fields. Use **tuples** for quick grouping of a few values or when order matters more than names.

```rust
// Tuple: fine for (x, y) coordinates
let point = (10, 20);

// Struct: better when fields have distinct roles
struct UserProfile {
    username: String,
    score: u32,
}
let profile = UserProfile { username: String::from("alice"), score: 100 };
println!("{} has score {}", profile.username, profile.score);
```

## Builder Pattern Intro

When a struct has many optional fields, a builder lets you set only what you need. Methods return `&mut self` for chaining.

```rust
#[derive(Debug)]
struct Query {
    table: String,
    limit: Option<u32>,
}

impl Query {
    fn new(table: &str) -> Query {
        Query { table: table.to_string(), limit: None }
    }
    fn limit(mut self, n: u32) -> Query {
        self.limit = Some(n);
        self
    }
}

let q = Query::new("users").limit(10);
println!("Query: {:?}", q);
```

## Printing with Debug and Display

`Debug` (`{:?}`) is for developers. `Display` (`{}`) is for user-facing output. `Debug` can be derived; `Display` must be implemented manually.

```rust
use std::fmt;

#[derive(Debug)]
struct Coords {
    x: i32,
    y: i32,
}

impl fmt::Display for Coords {
    fn fmt(&self, f: &mut fmt::Formatter) -> fmt::Result {
        write!(f, "({}, {})", self.x, self.y)
    }
}

let c = Coords { x: 3, y: 4 };
println!("Debug: {:?}", c);
println!("Display: {}", c);
```

## Common Mistakes Beginners Make

1. **Forgetting `#[derive(Debug)]`** when you want to print a struct
2. **Using `self` instead of `&self`** when you don't need to consume
3. **Making all fields `pub`** — prefer methods for controlled access
4. **Confusing tuple structs with tuples** — `Point(i32, i32)` is a type, not `(i32, i32)`

## Key Rules to Remember

- Named structs have named fields; tuple structs use `.0`, `.1`
- Use field init shorthand when names match
- `..other` copies remaining fields in struct update
- `&self` = borrow, `&mut self` = mutable borrow, `self` = consume
- Associated functions (no `self`) are called with `Type::name()`
- `#[derive(Debug)]` enables `{:?}`; implement `Display` for `{}`
- Use `pub` to expose structs and fields across module boundaries
