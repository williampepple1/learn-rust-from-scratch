# Control Flow

Control flow lets your program make decisions and repeat actions. In Rust, many constructs are **expressions** — they produce values. This makes the language elegant and reduces boilerplate.

## What Is Control Flow and Why Does It Matter?

Control flow determines which code runs and how many times. Without it, programs would execute top-to-bottom with no branching or loops. Rust's design — making `if` and `match` expressions — encourages clear, value-producing code.

## `if` / `else` Basics

Conditions must be `bool` — no implicit conversion from numbers. Use `if`, `else if`, and `else`.

```rust
let x = 7;
if x < 5 {
    println!("x is small");
} else if x < 10 {
    println!("x is medium");
} else {
    println!("x is large");
}
```

## `if` as an Expression (in `let`)

Because `if` is an expression, you can use it to assign a value. All branches must produce the same type.

```rust
let condition = true;
let number = if condition { 5 } else { 6 };
println!("number = {}", number);
```

## `match` — Exhaustive Pattern Matching

`match` compares a value against patterns and runs the first matching arm. The compiler enforces **exhaustiveness** — you must handle all possible cases.

```rust
let value = 3;
let description = match value {
    1 => "one",
    2 => "two",
    3 => "three",
    _ => "something else",  // _ catches everything
};
println!("{} is {}", value, description);
```

## Match Guards and Bindings

**Guards** add extra conditions with `if`. **Bindings** capture values with `@`.

```rust
let pair = (0, 5);
match pair {
    (0, y) => println!("x is 0, y is {}", y),
    (x, 0) => println!("x is {}, y is 0", x),
    (x, y) if x == y => println!("equal: {}", x),
    (x, y) => println!("x={}, y={}", x, y),
}

let n = 10;
match n {
    x @ 1..=5 => println!("Small: {}", x),
    x @ 6..=10 => println!("Medium: {}", x),
    _ => println!("Large"),
}
```

## Loops: `loop`, `while`, `for`

**`loop`** runs forever until `break`. **`while`** runs while a condition is true. **`for`** iterates over a collection.

```rust
let mut count = 0;
loop {
    count += 1;
    if count >= 3 { break; }
}
println!("Broke after {} iterations", count);
```

```rust
let mut i = 0;
while i < 3 {
    println!("i = {}", i);
    i += 1;
}
```

```rust
for item in [10, 20, 30] {
    println!("item = {}", item);
}
```

## Loop Labels and `break` with Values

Use labels like `'outer:` to break from nested loops. `loop` can return a value with `break value`.

```rust
let result = loop {
    break 42;
};
println!("loop returned: {}", result);

'outer: for i in 0..3 {
    for j in 0..3 {
        if i == 1 && j == 1 {
            break 'outer;
        }
        println!("i={}, j={}", i, j);
    }
}
```

## Ranges: `..` and `..=`

`..` is exclusive of the end; `..=` is inclusive.

```rust
for i in 0..5 {
    print!("{} ", i);
}
println!();
for i in 0..=5 {
    print!("{} ", i);
}
println!();
```

## `if let` — Match One Pattern

When you only care about one pattern, `if let` is shorter than `match`.

```rust
let some_value: Option<i32> = Some(7);
if let Some(x) = some_value {
    println!("Got value: {}", x);
}

let none_value: Option<i32> = None;
if let Some(x) = none_value {
    println!("Got value: {}", x);
} else {
    println!("Got None");
}
```

## `while let` — Loop While Pattern Matches

Useful for draining iterators or popping from a stack until empty.

```rust
let mut stack = vec![1, 2, 3];
while let Some(top) = stack.pop() {
    println!("Popped: {}", top);
}
```

## `?` Operator for Early Return

The `?` operator propagates errors: if the value is `Err`, return early; otherwise unwrap the `Ok`.

```rust
fn might_fail() -> Result<i32, String> {
    let x: Result<i32, String> = Ok(10);
    let y = x?;  // If Err, return early; else y = 10
    Ok(y * 2)
}
println!("Result: {:?}", might_fail());
```

## Iterating with `for`-in

`for` works with any type that implements `IntoIterator`: ranges, arrays, vectors, iterators.

```rust
let v = vec!["a", "b", "c"];
for (i, s) in v.iter().enumerate() {
    println!("{}: {}", i, s);
}
```

## Common Mistakes Beginners Make

1. **Using `if x` when x is a number** — Rust requires `bool`; use `if x != 0`
2. **Forgetting `_` in match** — exhaustiveness is required
3. **Using `break` without a value** in a `loop` when you want to return something
4. **Confusing `..` and `..=`** — 0..5 is 0–4; 0..=5 is 0–5

## Key Rules to Remember

- `if` conditions must be `bool`
- `if` and `match` are expressions — they produce values
- `match` must be exhaustive; use `_` as a catch-all
- `loop` can return a value with `break value`
- Use `'label:` for breaking out of nested loops
- `if let` and `while let` simplify single-pattern matching
- `?` propagates errors in `Result` and `Option`
