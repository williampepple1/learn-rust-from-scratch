# Enums & Pattern Matching

Welcome! Enums (short for *enumerations*) are one of Rust's superpowers. They let you define a type that can be one of several variants — like a traffic light that's either Red, Yellow, or Green. Combined with pattern matching, enums make your code expressive, safe, and impossible to forget a case.

## What Are Enums and Why Do They Matter?

In many languages, you might use strings or magic numbers to represent different states (e.g., `"red"`, `"yellow"`, `"green"`). Rust's enums are **type-safe**: the compiler knows exactly which values are valid and forces you to handle every possibility. No more "what if someone passes an invalid value?" — it simply cannot happen.

## Basic Enums: Variants with No Data

The simplest enum has variants that don't carry any extra data. Think of it like a list of named options.

```rust
#[derive(Debug)]
enum TrafficLight {
    Red,
    Yellow,
    Green,
}

let light = TrafficLight::Green;
println!("We have a traffic light: {:?}", light);
```

## Variants with Tuple Data

Enum variants can hold data! Tuple-style variants store values in order, like a tuple.

```rust
enum Message {
    Quit,
    Move { x: i32, y: i32 },
    Write(String),
    ChangeColor(i32, i32, i32),
}

let msg1 = Message::Write(String::from("Hello!"));
let msg2 = Message::ChangeColor(255, 0, 128);
let msg3 = Message::Move { x: 10, y: 20 };

println!("Messages created successfully!");
```

## Variants with Struct Data

You can use named fields (like a struct) for clarity when a variant has multiple pieces of data.

```rust
#[derive(Debug)]
enum Shape {
    Circle { radius: f64 },
    Rectangle { width: f64, height: f64 },
}

let circle = Shape::Circle { radius: 5.0 };
let rect = Shape::Rectangle { width: 10.0, height: 4.0 };
println!("circle: {:?}, rect: {:?}", circle, rect);
```

## Option&lt;T&gt;: Some and None

`Option<T>` is Rust's way of representing "maybe there's a value." Instead of `null` (which causes null pointer bugs), Rust uses:
- `Some(value)` — there is a value
- `None` — there is no value

The compiler forces you to handle both cases!

```rust
fn find_first_even(numbers: &[i32]) -> Option<i32> {
    for &n in numbers {
        if n % 2 == 0 { return Some(n); }
    }
    None
}

let nums = vec![1, 3, 5, 8, 9];
match find_first_even(&nums) {
    Some(n) => println!("First even: {}", n),
    None => println!("No even number found"),
}
```

## Result&lt;T, E&gt;: Ok and Err

`Result<T, E>` represents either success (`Ok(value)`) or failure (`Err(error)`). It's the standard way to handle recoverable errors in Rust.

```rust
fn divide(a: i32, b: i32) -> Result<i32, String> {
    if b == 0 {
        Err(String::from("Cannot divide by zero"))
    } else {
        Ok(a / b)
    }
}

println!("{:?}", divide(10, 2));
println!("{:?}", divide(10, 0));
```

## Match: Exhaustive Pattern Matching

`match` is like a supercharged switch. It must be **exhaustive** — you must handle every possible variant. The compiler will error if you forget one!

```rust
enum Coin {
    Penny,
    Nickel,
    Dime,
    Quarter,
}

fn value_in_cents(coin: Coin) -> u8 {
    match coin {
        Coin::Penny => 1,
        Coin::Nickel => 5,
        Coin::Dime => 10,
        Coin::Quarter => 25,
    }
}

println!("A dime is worth {} cents", value_in_cents(Coin::Dime));
```

## Match with Data: Extracting Values

When variants carry data, you can extract and use it in the match arms.

```rust
fn area(shape: &Shape) -> f64 {
    match shape {
        Shape::Circle { radius } => std::f64::consts::PI * radius * radius,
        Shape::Rectangle { width, height } => width * height,
    }
}

let c = Shape::Circle { radius: 3.0 };
let r = Shape::Rectangle { width: 4.0, height: 5.0 };
println!("Circle area: {:.2}", area(&c));
println!("Rectangle area: {}", area(&r));
```

## if let: Matching One Variant

When you only care about one variant, `if let` is more concise than `match`.

```rust
let some_value: Option<i32> = Some(42);

if let Some(x) = some_value {
    println!("We got the value: {}", x);
}

let none_value: Option<i32> = None;
if let Some(x) = none_value {
    println!("This won't run");
} else {
    println!("It was None");
}
```

## while let: Looping Until a Pattern Fails

`while let` keeps looping as long as the pattern matches. Great for popping from a stack!

```rust
let mut stack = vec![1, 2, 3];

while let Some(top) = stack.pop() {
    println!("Popped: {}", top);
}
println!("Stack is empty");
```

## Methods on Enums with impl

Enums can have methods, just like structs. Use `impl` to add behavior.

```rust
impl TrafficLight {
    fn next(&self) -> TrafficLight {
        match self {
            TrafficLight::Red => TrafficLight::Green,
            TrafficLight::Yellow => TrafficLight::Red,
            TrafficLight::Green => TrafficLight::Yellow,
        }
    }
}

let light = TrafficLight::Red;
let next = light.next();
println!("After Red comes: {:?}", next);
```

## Common Standard Library Enums

Besides `Option` and `Result`, you'll see enums like `std::cmp::Ordering` (Less, Equal, Greater), `IpAddr` (V4, V6), and `Option`/`Result` everywhere in the standard library.

```rust
use std::cmp::Ordering;

let x = 5;
let y = 10;
let ordering = x.cmp(&y);

match ordering {
    Ordering::Less => println!("x is less than y"),
    Ordering::Equal => println!("x equals y"),
    Ordering::Greater => println!("x is greater than y"),
}
```

## When to Use Enums vs Structs

- **Use enums** when a value can be *one of* several variants (traffic light, message type, optional value).
- **Use structs** when you have a fixed set of fields that always exist together (a point, a user).

Enums and structs work great together — enum variants can hold structs!

## Enum Sizing and the Never Type

Rust enums are sized to fit their largest variant. The **never type** `!` (also written as `std::convert::Infallible`) represents computations that never finish — useful for `Result<T, !>` when failure is impossible.

## Enums for State Machines

Enums are perfect for modeling state machines: each variant is a state, and transitions are just creating a new variant. The compiler ensures invalid states are unrepresentable!

```rust
#[derive(Debug)]
enum ConnectionState {
    Disconnected,
    Connecting,
    Connected { user_id: u32 },
}

let state = ConnectionState::Connected { user_id: 42 };
println!("State: {:?}", state);
```

## Common Mistakes Beginners Make

1. **Forgetting to handle all match arms** — use `_` or `other` as a catch-all if needed
2. **Confusing `Option` with default values** — `Option` means "maybe absent"; use `unwrap_or` for defaults
3. **Overusing `unwrap()`** — prefer `match` or `?` for proper error handling
4. **Not using `#[derive(Debug)]`** — add it to enums you want to print with `{:?}`

## Key Rules to Remember

- Enums define a type with multiple variants; each variant can carry data
- `Option<T>` = Some(value) or None; `Result<T,E>` = Ok(value) or Err(error)
- `match` must be exhaustive — handle every variant
- Use `if let` for single-variant checks; `while let` for loop-until-fail
- Enums can have methods via `impl`
- Enums are ideal for state machines and "one of many" data
