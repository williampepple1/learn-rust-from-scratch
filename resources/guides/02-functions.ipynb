# Functions

Functions are the heart of reusable logic in Rust. They let you package behavior into named blocks that you can call from anywhere. Think of them as recipes: you define the steps once, then follow them whenever you need the result.

## What Are Functions and Why Do They Matter?

Functions help you avoid repetition, organize code, and make programs easier to understand and test. In Rust, functions are first-class citizens — you can pass them around, store them, and return them.

## Basic `fn` Syntax

Define a function with `fn`, a name, parentheses for parameters, and optionally a return type.

````rust
fn greet() {
    println!("Hello, Rust!");
}
greet();
````

## Parameters and Return Types

Parameters go inside parentheses with type annotations. The return type comes after `->`.

````rust
fn add(a: i32, b: i32) -> i32 {
    a + b
}
let sum = add(3, 7);
println!("3 + 7 = {}", sum);
````

## Expressions vs Statements

In Rust, **expressions** produce a value; **statements** do not. The last expression in a function (no semicolon!) is the return value. A semicolon turns it into a statement, which returns `()`.

````rust
fn double(x: i32) -> i32 {
    x * 2   // No semicolon = expression, this is returned
}
// BAD: fn bad_double(x: i32) -> i32 { x * 2; }  // Semicolon = statement, returns () — type error!
println!("double(5) = {}", double(5));
````

## The Unit Type `()`

Functions that don't return a value implicitly return `()`, the unit type. It's like "nothing" but as a type.

````rust
fn do_nothing() {
    println!("Doing nothing useful");
}
let result = do_nothing();
println!("Result of do_nothing: {:?}", result);
````

## Early Return with `return`

Use `return` to exit a function early with a value. Most of the time, the trailing expression is preferred.

````rust
fn abs(n: i32) -> i32 {
    if n < 0 {
        return -n;
    }
    n
}
println!("abs(-10) = {}, abs(10) = {}", abs(-10), abs(10));
````

## Multiple Return Values with Tuples

Return multiple values by packing them in a tuple.

````rust
fn min_max(a: i32, b: i32) -> (i32, i32) {
    if a < b { (a, b) } else { (b, a) }
}
let (min, max) = min_max(42, 17);
println!("min = {}, max = {}", min, max);
````

## Nested Functions

You can define functions inside other functions. Useful for small helpers that don't need to be visible elsewhere.

````rust
fn outer() {
    fn inner(x: i32) -> i32 {
        x + 1
    }
    println!("inner(10) = {}", inner(10));
}
outer();
````

## Function Pointers: The `fn` Type

Functions have a type: `fn(ArgTypes) -> ReturnType`. You can pass functions as arguments or store them in variables.

````rust
fn apply_twice(f: fn(i32) -> i32, x: i32) -> i32 {
    f(f(x))
}
fn add_one(x: i32) -> i32 { x + 1 }
let result = apply_twice(add_one, 5);
println!("apply_twice(add_one, 5) = {}", result);
````

## Diverging Functions `-> !`

A diverging function never returns. It has the "never" type `!`. Examples: `panic!()`, infinite loops.

````rust
fn forever() -> ! {
    loop {
        // Never returns
    }
}
// forever();  // Uncomment to run forever!
fn panic_example() -> ! {
    panic!("This function never returns normally");
}
// panic_example();  // Uncomment to panic
println!("Diverging functions never return — we skip calling them here.");
````

## Documentation Comments `///`

Use `///` for doc comments that appear in generated documentation. They support Markdown!

````rust
/// Computes the square of a number.
/// # Examples
/// ```
/// assert_eq!(square(4), 16);
/// ```
fn square(x: i32) -> i32 {
    x * x
}
println!("square(4) = {}", square(4));
````

## The `main` Function

Every executable Rust program has a `main` function. It's the entry point. In Jupyter/EvCxR, we don't need `main` — cells run directly.

## Passing by Value vs by Reference

By default, Rust **moves** or **copies** values. Use `&` to pass a reference (borrow) so the caller keeps ownership. Use `&mut` for mutable references.

````rust
fn by_value(x: i32) {
    println!("Received copy: {}", x);
}
fn by_reference(s: &str) {
    println!("Borrowed: {}", s);
}
fn by_mut_ref(n: &mut i32) {
    *n += 10;
}

let num = 5;
by_value(num);
println!("num still usable (Copy): {}", num);

let text = String::from("hello");
by_reference(&text);
println!("text still owned: {}", text);

let mut value = 7;
by_mut_ref(&mut value);
println!("value after mutation: {}", value);
````

## Common Mistakes Beginners Make

1. **Adding a semicolon** to the last expression — it becomes a statement returning `()`
2. **Forgetting return type** when the function returns something other than `()`
3. **Confusing `fn` type with closures** — `fn` is for function pointers, not closures
4. **Passing by value when you need the original** — use `&` to borrow

## Key Rules to Remember

- The last expression (no semicolon) is the return value
- Use `return` for early exits
- `()` is the unit type for "no value"
- Tuples let you return multiple values
- `fn(A, B) -> C` is the type of a function
- `-> !` means the function never returns
- Use `&` and `&mut` to pass by reference and avoid moving
