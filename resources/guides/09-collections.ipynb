# Collections

Welcome! Collections are Rust's way of storing multiple values together. Think of them as flexible containers — unlike arrays (fixed size) and tuples (fixed structure), collections can grow, shrink, and organize data in different ways. This guide covers the most important ones: `Vec`, `String`, `HashMap`, `HashSet`, and more.

## What Are Collections and Why Do They Matter?

Collections store data on the **heap** (except when optimized away). They're essential when you don't know the size at compile time or need dynamic data structures. Rust's standard library provides several collections, each optimized for different use cases.

## Vec&lt;T&gt; — The Dynamic Array

`Vec` is a growable, heap-allocated array. It's one of the most used collections in Rust. You can create it, push and pop elements, index into it, slice it, and iterate over it.

```rust
// Creating a Vec
let v1: Vec<i32> = Vec::new();
let v2 = vec![1, 2, 3];           // macro for convenience
let v3 = vec![0; 5];              // [0, 0, 0, 0, 0]
println!("v2: {:?}, v3: {:?}", v2, v3);
```

```rust
// push and pop
let mut nums = vec![10, 20];
nums.push(30);
nums.push(40);
println!("After push: {:?}", nums);
let last = nums.pop();
println!("Popped: {:?}, remaining: {:?}", last, nums);
```

```rust
// Indexing and slicing
let v = vec![10, 20, 30, 40, 50];
println!("v[2] = {}", v[2]);
let slice: &[i32] = &v[1..4];    // [20, 30, 40]
println!("slice: {:?}", slice);
// v[99] would panic! Use get() for safe access:
println!("v.get(99) = {:?}", v.get(99));
```

```rust
// Iterating over a Vec
let v = vec!["a", "b", "c"];
for s in &v {
    println!("  {}", s);
}
// Mutable iteration
let mut nums = vec![1, 2, 3];
for n in &mut nums {
    *n += 10;
}
println!("After +10: {:?}", nums);
```

## Capacity vs Length

A `Vec` has a **length** (how many elements) and a **capacity** (how much space is allocated). When you push beyond capacity, the Vec reallocates. Use `with_capacity(n)` when you know the size in advance to avoid reallocations.

```rust
let mut v = Vec::with_capacity(10);
println!("len: {}, cap: {}", v.len(), v.capacity());
v.push(1); v.push(2); v.push(3);
println!("len: {}, cap: {}", v.len(), v.capacity());
```

## String — Growable UTF-8 Text

`String` is a growable, owned UTF-8 encoded string. It's like `Vec<u8>` but guaranteed to be valid UTF-8. You can create, push, concatenate, and slice strings.

```rust
// Creating strings
let s1 = String::new();
let s2 = String::from("hello");
let s3 = "world".to_string();
let s4 = "rust".to_owned();
println!("{}, {}, {}, {}", s1, s2, s3, s4);
```

```rust
// Pushing and concatenation
let mut s = String::from("foo");
s.push_str("bar");
s.push('!');
println!("s = {}", s);
// Concatenation with +
let a = String::from("Hello, ");
let b = String::from("world!");
let c = a + &b;  // a is moved, b is borrowed
println!("c = {}", c);
```

```rust
// Slicing strings — be careful! Indices are byte positions, not char positions
let s = "hello";
let slice = &s[0..2];  // "he"
println!("slice: {}", slice);
// &s[0..1] on " café " would panic (é is 2 bytes in UTF-8)
// Use chars() or char_indices() for character-level access
```

```rust
// chars vs bytes
let s = "café";
println!("chars: {:?}", s.chars().collect::<Vec<_>>());
println!("bytes: {:?}", s.bytes().collect::<Vec<_>>());
```

## HashMap&lt;K, V&gt; — Key-Value Storage

A `HashMap` stores key-value pairs. Keys must be unique. Lookup is fast (average O(1)). You need `use std::collections::HashMap;`.

```rust
use std::collections::HashMap;

let mut scores = HashMap::new();
scores.insert("Alice", 100);
scores.insert("Bob", 85);
scores.insert("Charlie", 92);
println!("scores: {:?}", scores);
```

```rust
// get returns Option<&V>
let alice_score = scores.get("Alice");
println!("Alice: {:?}", alice_score);
// entry API: insert or update
scores.entry("Alice").or_insert(0);
scores.entry("Dave").or_insert(70);
println!("scores: {:?}", scores);
```

```rust
// Default values with entry
let text = "hello world wonderful world";
let mut word_count: HashMap<&str, u32> = HashMap::new();
for word in text.split_whitespace() {
    let count = word_count.entry(word).or_insert(0);
    *count += 1;
}
println!("word_count: {:?}", word_count);
```

```rust
// Iterating over a HashMap
for (name, score) in &scores {
    println!("{}: {}", name, score);
}
```

## HashSet&lt;T&gt; — Unique Values

A `HashSet` stores unique values (no duplicates). It's built on `HashMap<T, ()>`. Great for membership tests and deduplication.

```rust
use std::collections::HashSet;

let mut seen = HashSet::new();
seen.insert(1);
seen.insert(2);
seen.insert(1);  // duplicate, ignored
println!("seen: {:?}", seen);
println!("contains 2: {}", seen.contains(&2));
```

## BTreeMap and BTreeSet

`BTreeMap` and `BTreeSet` keep keys in sorted order. Use them when you need ordering or range queries. Lookup is O(log n) instead of O(1) average.

```rust
use std::collections::BTreeMap;

let mut map = BTreeMap::new();
map.insert(3, "three");
map.insert(1, "one");
map.insert(2, "two");
println!("BTreeMap (sorted): {:?}", map);
```

## VecDeque — Double-Ended Queue

`VecDeque` allows efficient push and pop from both ends. Use it when you need a queue or a sliding window.

```rust
use std::collections::VecDeque;

let mut deque: VecDeque<i32> = VecDeque::new();
deque.push_back(1);
deque.push_back(2);
deque.push_front(0);
println!("deque: {:?}", deque);
println!("pop_front: {:?}", deque.pop_front());
```

## When to Use Which Collection?

| Need | Use |
|------|-----|
| Dynamic list, indexed access | `Vec<T>` |
| Key-value lookup, no order | `HashMap<K,V>` |
| Unique values, membership | `HashSet<T>` |
| Sorted keys, range queries | `BTreeMap<K,V>` / `BTreeSet<T>` |
| Queue, push/pop both ends | `VecDeque<T>` |
| Growable text | `String` |

## Common Iterator Methods on Collections

Collections implement `IntoIterator` and often have `.iter()`, `.iter_mut()`, `.into_iter()`. You can chain adaptors and collect back into collections.

```rust
let v = vec![1, 2, 3, 4, 5];
let doubled: Vec<i32> = v.iter().map(|x| x * 2).collect();
let evens: Vec<&i32> = v.iter().filter(|x| **x % 2 == 0).collect();
let sum: i32 = v.iter().sum();
println!("doubled: {:?}, evens: {:?}, sum: {}", doubled, evens, sum);
```

## Common Mistakes Beginners Make

1. **Indexing with `[]` on Option** — use `.get()` or match on the result
2. **String slicing at invalid byte boundaries** — UTF-8 means one char ≠ one byte
3. **Forgetting that `HashMap::get` returns `Option<&V>`** — you need to handle `None`
4. **Using `Vec` when you need a queue** — `VecDeque` is better for FIFO
5. **Not reserving capacity** — use `with_capacity` when you know the size

## Key Rules to Remember

- `Vec<T>`: growable array, push/pop, indexing, slicing
- `String`: owned UTF-8 text; slice with care (byte indices)
- `HashMap<K,V>`: key-value, `insert`, `get`, `entry` API
- `HashSet<T>`: unique values, `insert`, `contains`
- Choose the right collection for your access pattern
- Use iterator methods (`map`, `filter`, `collect`) for clean, idiomatic code
