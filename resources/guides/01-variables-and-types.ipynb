# Variables & Data Types

Welcome! This guide introduces the building blocks of Rust: how we store data and what kinds of data we can work with. Think of variables as labeled boxes that hold values — and Rust gives you precise control over what goes in each box.

## What Are Variables and Why Do They Matter?

Variables let you store and reuse values in your program. In Rust, variables are **immutable by default** — once you assign a value, it cannot change unless you explicitly opt in. This design prevents many bugs and makes your code easier to reason about.

## Basic Syntax: `let` Bindings

Use `let` to create a variable binding. The name is bound to a value.

```rust
let x = 42;
println!("x = {}", x);
```

## Mutability with `mut`

By default, variables are immutable. To allow changes, add the `mut` keyword.

```rust
let mut count = 0;
count += 1;
count += 1;
println!("count = {}", count);
```

## Type Inference and Type Annotations

Rust infers types from how you use variables. You can also annotate types explicitly when you want clarity or when the compiler needs a hint.

```rust
let inferred = 100;           // Compiler infers i32
let explicit: i64 = 100;      // We specify i64
let float: f64 = 3.14;       // Must annotate for floats (default is f64)
println!("inferred: {}, explicit: {}, float: {}", inferred, explicit, float);
```

## Integer Types

Rust has signed (`i`) and unsigned (`u`) integers in sizes 8, 16, 32, 64, and 128 bits. `isize` and `usize` match the pointer size on your machine (32 or 64 bits).

```rust
let small: i8 = -128;
let medium: i32 = 1_000_000;
let big: u64 = 18_446_744_073_709_551_615;
let index: usize = 42;  // Common for array indices
println!("i8: {}, i32: {}, u64: {}, usize: {}", small, medium, big, index);
```

## Floating-Point Types

Rust has `f32` (32-bit) and `f64` (64-bit). The default for literals like `3.14` is `f64`.

```rust
let pi: f64 = 3.14159;
let small_float: f32 = 2.5;
println!("pi = {}, small_float = {}", pi, small_float);
```

## Booleans and Characters

`bool` is `true` or `false`. `char` is a single Unicode scalar value (4 bytes), not a single byte like in C.

```rust
let is_rust_fun: bool = true;
let letter: char = 'R';
let emoji: char = '🦀';
println!("bool: {}, letter: {}, emoji: {}", is_rust_fun, letter, emoji);
```

## `&str` vs `String`

`&str` is a **string slice** — a view into existing string data (immutable). `String` is an owned, growable string stored on the heap.

```rust
let slice: &str = "hello";           // Immutable view
let owned: String = String::from("world");  // Owned, growable
let mut growable = String::from("Hi");
growable.push_str(" there!");
println!("slice: {}, owned: {}, growable: {}", slice, owned, growable);
```

## Tuples and Arrays

**Tuples** group values of different types. **Arrays** hold a fixed number of values of the same type.

```rust
let tuple: (i32, f64, &str) = (42, 3.14, "hello");
println!("tuple: {:?}", tuple);
println!("First: {}, Second: {}", tuple.0, tuple.1);

let arr: [i32; 5] = [1, 2, 3, 4, 5];
println!("array: {:?}", arr);
println!("arr[2] = {}", arr[2]);
```

## Type Casting with `as`

Use `as` to convert between compatible types. Be careful with truncation!

```rust
let x: i32 = 100;
let y: f64 = x as f64;
let z: u8 = 255;
let w: i32 = z as i32;  // 255 fits in i32
println!("x as f64: {}, z as i32: {}", y, w);
```

## Numeric Literals: Hex, Octal, Binary, Underscores

Rust supports readable number formats. Underscores are ignored and help with readability.

```rust
let decimal = 1_000_000;
let hex = 0xFF;
let octal = 0o77;
let binary = 0b1111_0000;
println!("decimal: {}, hex: {}, octal: {}, binary: {}", decimal, hex, octal, binary);
```

## Constants vs Statics

**Constants** (`const`) are compile-time values, inlined everywhere. **Statics** (`static`) have a fixed memory location and can be mutable (with `unsafe`). Prefer `const` for simple values.

```rust
const MAX_POINTS: u32 = 100_000;
static GREETING: &str = "Hello, Rust!";
println!("MAX_POINTS: {}, GREETING: {}", MAX_POINTS, GREETING);
```

## Shadowing vs Mutability

**Shadowing** lets you reuse a variable name with a new binding. The old value is hidden. This is different from `mut` — shadowing creates a new variable and can change the type.

```rust
let x = 5;
let x = x + 1;   // Shadowing: new x = 6
let x = "six";  // Can even change type!
println!("x = {}", x);
```

## Common Mistakes Beginners Make

1. **Forgetting `mut`** when you need to modify a variable
2. **Confusing `&str` and `String`** — use `&str` for function parameters when you don't need ownership
3. **Integer overflow** — debug builds panic, release builds wrap
4. **Using `const` for computed values** — `const` must be evaluable at compile time

## Key Rules to Remember

- Variables are **immutable by default** — use `mut` when you need to change them
- Rust infers types, but you can annotate when needed
- `&str` is a slice; `String` is owned and growable
- Tuples group mixed types; arrays are fixed-size and same-type
- Shadowing creates a new variable; it's not the same as mutability
- Use `const` for compile-time constants; `static` for global state
