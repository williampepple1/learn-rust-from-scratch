### Exercise 142: Match Struct and Enum

Write `area(s: Shape) -> f64` matching Circle and Rect.

```rust
enum Shape { Circle(f64), Rect(f64, f64) }
fn area(s: Shape) -> f64 {
    match s {
        Shape::Circle(r) => std::f64::consts::PI * r * r,
        Shape::Rect(w, h) => w * h
    }
}
area(Shape::Rect(3.0, 4.0))
```

```rust
assert!((area(Shape::Circle(1.0)) - 3.14159).abs() < 0.01);
assert_eq!(area(Shape::Rect(3.0, 4.0)), 12.0);
println!("✅ Exercise 142 passed!");
```
