### Exercise 233: HashSet union and intersection

Compute union and intersection of two HashSets.

```rust
use std::collections::HashSet;

let a: HashSet<i32> = [1, 2, 3].into_iter().collect();
let b: HashSet<i32> = [2, 3, 4].into_iter().collect();
let union: HashSet<i32> = a.union(&b).copied().collect();
let inter: HashSet<i32> = a.intersection(&b).copied().collect();
let mut u: Vec<i32> = union.into_iter().collect();
let mut i: Vec<i32> = inter.into_iter().collect();
u.sort();
i.sort();
(u, i)
```

```rust
assert_eq!(u, vec![1, 2, 3, 4]);
assert_eq!(i, vec![2, 3]);
println!("✅ Exercise 233 passed!");
```

### Exercise 244: Word count with HashMap

```rust
use std::collections::HashMap;

let s = "hello world hello rust world hello";
let mut counts: HashMap<&str, i32> = HashMap::new();
for word in s.split_whitespace() {
    *counts.entry(word).or_insert(0) += 1;
}
counts.get("hello").copied()
```

### Exercise 250: Practical — Group by first letter

Group words `["apple", "apricot", "banana", "blueberry"]` by first letter into a HashMap.

```rust
use std::collections::HashMap;

let words = vec!["apple", "apricot", "banana", "blueberry"];
let mut groups: HashMap<char, Vec<&str>> = HashMap::new();
for w in &words {
    let c = w.chars().next().unwrap();
    groups.entry(c).or_insert_with(Vec::new).push(w);
}
let a_words = groups.get(&'a').cloned();
a_words
```

```rust
assert_eq!(groups.get(&'a').map(|v| v.len()), Some(2));
assert_eq!(groups.get(&'b').map(|v| v.len()), Some(2));
println!("✅ Exercise 250 passed!");
```
