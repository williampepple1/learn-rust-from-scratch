### Exercise 151: Define a Simple Trait

### Exercise 153: Trait with Multiple Methods

Define trait `Shape` with `area(&self) -> f64` and `perimeter(&self) -> f64`. Implement for `Rectangle { w: f64, h: f64 }`.

```rust
trait Shape {
    fn area(&self) -> f64;
    fn perimeter(&self) -> f64;
}

struct Rectangle { w: f64, h: f64 }

impl Shape for Rectangle {
    fn area(&self) -> f64 { self.w * self.h }
    fn perimeter(&self) -> f64 { 2.0 * (self.w + self.h) }
}

let rect = Rectangle { w: 4.0, h: 5.0 };
(rect.area(), rect.perimeter())
```

### Exercise 154: Implement Shape for Circle

Implement `Shape` for `Circle { r: f64 }`. Use π ≈ 3.14159 for area and perimeter.

```rust
trait Shape { fn area(&self) -> f64; fn perimeter(&self) -> f64; }
struct Rectangle { w: f64, h: f64 }
struct Circle { r: f64 }

impl Shape for Rectangle { fn area(&self) -> f64 { self.w * self.h } fn perimeter(&self) -> f64 { 2.0 * (self.w + self.h) } }
impl Shape for Circle { fn area(&self) -> f64 { self.r * self.r * 3.14159 } fn perimeter(&self) -> f64 { 2.0 * 3.14159 * self.r } }

let c = Circle { r: 1.0 };
c.area()
```

```rust
let pi = 3.14159;
assert!((c.area() - pi).abs() < 0.001);
assert!((c.perimeter() - 2.0 * pi).abs() < 0.001);
let c2 = Circle { r: 2.0 };
assert!((c2.area() - 4.0 * pi).abs() < 0.01);
println!("✅ Exercise 154 passed!");
```

### Exercise 184: From for Error Conversion

```rust
trait Greet { fn greet(&self) -> String; }
struct Person { name: String }
struct Robot { id: u32 }
impl Greet for Person { fn greet(&self) -> String { "Person".into() } }
impl Greet for Robot { fn greet(&self) -> String { "Robot".into() } }

fn pick(use_robot: bool) -> Box<dyn Greet> {
    if use_robot { Box::new(Robot { id: 0 }) } else { Box::new(Person { name: String::new() }) }
}

let g = pick(true);
g.greet()
```

```rust
assert_eq!(pick(true).greet(), "Robot");
assert_eq!(pick(false).greet(), "Person");
println!("✅ Exercise 194 passed!");
```

### Exercise 196: Vec of dyn Trait References

Create `Vec<&dyn Greet>` from a mix of Person and Robot references.

```rust
trait Greet { fn greet(&self) -> String; }
struct Person { name: String }
struct Robot { id: u32 }
impl Greet for Person { fn greet(&self) -> String { format!("P:{}", self.name) } }
impl Greet for Robot { fn greet(&self) -> String { format!("R:{}", self.id) } }

let p = Person { name: String::from("A") };
let r = Robot { id: 1 };
let greeters: Vec<&dyn Greet> = vec![&p, &r];
greeters.iter().map(|g| g.greet()).collect::<Vec<_>>()
```

```rust
let msgs: Vec<String> = greeters.iter().map(|g| g.greet()).collect();
assert_eq!(msgs[0], "P:A");
assert_eq!(msgs[1], "R:1");
println!("✅ Exercise 196 passed!");
```

### Exercise 200: Practical Trait Object Use

Create a function `run_greeters(greeters: Vec<Box<dyn Greet>>)` that returns a Vec of their greet strings.

```rust
trait Greet { fn greet(&self) -> String; }
struct Person { name: String }
struct Robot { id: u32 }
impl Greet for Person { fn greet(&self) -> String { format!("Person {}", self.name) } }
impl Greet for Robot { fn greet(&self) -> String { format!("Robot {}", self.id) } }

fn run_greeters(greeters: Vec<Box<dyn Greet>>) -> Vec<String> {
    greeters.into_iter().map(|g| g.greet()).collect()
}

let input = vec![
    Box::new(Person { name: String::from("Alice") }) as Box<dyn Greet>,
    Box::new(Robot { id: 42 }) as Box<dyn Greet>,
];
run_greeters(input)
```

```rust
let result = run_greeters(vec![Box::new(Person { name: String::from("A") }), Box::new(Robot { id: 1 })]);
assert_eq!(result[0], "Person A");
assert_eq!(result[1], "Robot 1");
println!("✅ Exercise 200 passed!");
```
