### Exercise 300: Practical — Parse config

Parse a "key=value" string into (key, value). Return proper errors for invalid format.

```rust
fn parse_kv(s: &str) -> Result<(&str, &str), String> {
    let mut parts = s.splitn(2, '=');
    let key = parts.next().ok_or("empty")?;
    let value = parts.next().ok_or("missing =")?;
    if key.is_empty() { return Err("empty key".into()); }
    Ok((key, value))
}

(parse_kv("port=8080"), parse_kv("invalid"))
```

```rust
assert_eq!(parse_kv("port=8080"), Ok(("port", "8080")));
assert!(parse_kv("invalid").is_err());
assert!(parse_kv("=value").is_err());
println!("✅ Exercise 300 passed!");
```
