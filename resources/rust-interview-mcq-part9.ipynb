# 🦀 Rust Interview MCQ — Part 9: Macros, Unsafe & FFI (Q801–Q900)


### Q801. What fragment specifier matches an expression like `1 + 2` or `foo()`?

- A) expr
- B) ty
- C) ident
- D) stmt

<details>
<summary>Show Answer</summary>

**Q801. A)** The `expr` fragment specifier matches Rust expressions: literals, function calls, binary ops, blocks, etc. `ty` matches types; `ident` matches identifiers; `stmt` matches statements.

</details>


### Q802. In `macro_rules! foo { ($x:expr) => { $x }; }`, what does `$x:expr` mean?

- A) x is a variable that must be an expression; expr is the fragment specifier
- B) x is a type named expr
- C) expr is a metavariable
- D) x and expr are both identifiers

<details>
<summary>Show Answer</summary>

**Q802. A)** In macro patterns, `$name:specifier` binds input to `$name`; the specifier (expr, ty, ident, etc.) constrains what the macro accepts. `expr` means the match must be a valid expression.

</details>


### Q803. What does the `tt` (token tree) fragment specifier match?

- A) Only identifiers
- B) Any single token or delimited token tree (e.g., `(a + b)` or `foo`)
- C) Only type expressions
- D) Only statements

<details>
<summary>Show Answer</summary>

**Q803. B)** `tt` is the most general fragment—it matches a single token or a balanced delimited group `(...)`, `{...}`, `[...]`. Used in TT munchers and when you need maximum flexibility.

</details>


### Q804. What does this macro produce when invoked as `repeat!(3, x + 1)`?

```rust
macro_rules! repeat {
    ($n:expr, $e:expr) => { (0..$n).map(|_| $e).collect::<Vec<_>>() };
}
```

- A) [1, 1, 1]
- B) [1, 2, 3]
- C) Compilation error
- D) [x+1, x+1, x+1]

<details>
<summary>Show Answer</summary>

**Q804. A)** The macro expands to (0..3).map(|_| x+1).collect::<Vec<_>>(). For each of 0,1,2, it evaluates x+1. If x is in scope and e.g. 0, we get [1,1,1]. The key is $e is repeated and evaluated each time.

</details>

### Q810. What does `ident` fragment specifier match?

### Q811. What is a 'TT muncher' in Rust macros?

- D) A macro that uses only tt specifiers

<details>
<summary>Show Answer</summary>

**Q811. A)** A TT muncher is a recursive macro that matches `$first:tt $rest:tt*`, processes `$first`, and recursively invokes itself on `$rest`. Used to parse variable-length input without fixed patterns.

</details>

### Q813. In macro hygiene, what is 'hygienic'?

<details>
<summary>Show Answer</summary>

**Q813. B)** Rust's macro hygiene means identifiers *introduced* by the macro (e.g., a temp variable) don't conflict with caller's names. But paths and types from the macro expansion are resolved in the caller's scope. It's partial hygiene.

</details>


### Q814. What does this recursive macro do when given `sum!(1, 2, 3)`?

```rust
macro_rules! sum {
    ($first:expr) => { $first };
    ($first:expr, $($rest:expr),*) => { $first + sum!($($rest),*) };
}
```

- A) 6
- B) Compilation error: recursive macro
- C) 1 + 2 + 3 (expands to that expression)
- D) Panics at compile time

<details>
<summary>Show Answer</summary>

**Q814. C)** The macro recursively expands: sum!(1,2,3) -> 1 + sum!(2,3) -> 1 + 2 + sum!(3) -> 1 + 2 + 3. So it produces the expression 1+2+3, which evaluates to 6. The expansion works; recursion terminates on the base case.

</details>

### Q819. What happens if a macro has multiple matching arms and the input matches more than one?
