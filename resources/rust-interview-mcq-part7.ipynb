### Q601. What is the output of the following code?

```rust
let x = 10;
let add_x = |y| x + y;
let result = add_x(5);
println!("{}", result);
```

- A) 15
- B) Compilation error: closure cannot capture `x`
- C) 10
- D) 5

<details>
<summary>Show Answer</summary>

**Q601. A)** The closure captures `x` by immutable reference (the default). Since `x` is a Copy type (i32), the closure can use it. The result is 10 + 5 = 15.

</details>


### Q602. Which closure trait is required when a closure must be called multiple times and mutates captured state?

```rust
let mut counter = 0;
let mut inc = || { counter += 1; counter };
inc(); inc();
```

- A) Fn
- B) FnMut
- C) FnOnce
- D) FnRef

<details>
<summary>Show Answer</summary>

**Q602. B)** FnMut is required because the closure mutates `counter`. Fn requires immutable capture; FnOnce can only be called once. FnMut allows mutable borrowing of captured variables and multiple calls.

</details>


### Q603. What happens when you compile this code?

```rust
let s = String::from("hello");
let f = move || println!("{}", s);
println!("{}", s);
```

- C) Prints "hello" once (from the closure call)

**Q603. B)** The `move` keyword transfers ownership of `s` into the closure. After the closure is created, `s` is no longer valid. The second `println!` tries to use the moved `s`, causing a compilation error.

</details>


### Q604. Which trait bound allows a closure to be called only once?

- B) F: FnMut
- C) F: FnOnce

### Q606. Which statement about closure trait bounds is correct?

- A) Fn is a subtrait of FnMut, which is a subtrait of FnOnce

- C) Fn, FnMut, and FnOnce are independent traits

<details>
<summary>Show Answer</summary>

**Q606. A)** The trait hierarchy is Fn: FnMut: FnOnce. FnOnce is the most general (can consume the closure); FnMut allows mutation; Fn allows only immutable borrowing. Every closure implements at least FnOnce.

### Q611. What is the key difference between `impl Fn(i32) -> i32` and `dyn Fn(i32) -> i32` as function parameters?

- A) impl Fn is dynamically dispatched; dyn Fn is statically dispatched
- B) impl Fn is statically dispatched and can inline; dyn Fn is dynamically dispatched and has a vtable
- C) dyn Fn cannot be used as a parameter
- D) impl Fn requires the closure to be Sized

<details>
<summary>Show Answer</summary>

**Q611. B)** impl Fn uses static dispatch—the compiler monomorphizes for each concrete type, enabling inlining. dyn Fn uses dynamic dispatch via a vtable at runtime, which has indirection cost but allows heterogeneous collections of closures.

</details>

</details>
