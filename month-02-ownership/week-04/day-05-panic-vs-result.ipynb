# 🦀 Day 5: `panic!` vs Recoverable Errors

---

## 💥 When to Panic

| Panic (`panic!`, `unwrap`, `expect`) | Recoverable (`Result`, `Option`) |
|---|---|
| Programmer bugs (logic errors) | Expected failure modes |
| Violated invariants | User input errors |
| Unrecoverable state | File/network errors |
| Tests and examples | Any error the caller can handle |

```rust
// PANIC: when something is fundamentally wrong

// panic!("This should never happen!");

// Array out of bounds → panic
// let v = vec![1, 2, 3];
// v[10]; // 💥 panics!

// unwrap on None → panic
// let x: Option<i32> = None;
// x.unwrap(); // 💥 panics!

println!("Panics crash the program immediately!");
println!("Use them for bugs, not for expected errors.");
```

```rust
// RECOVER: when failure is expected and handleable

// Good: File might not exist
use std::fs;
match fs::read_to_string("maybe_exists.txt") {
    Ok(content) => println!("Content: {}", content),
    Err(_) => println!("File doesn't exist — that's OK!"),
}

// Good: User input might be wrong
fn parse_input(s: &str) -> Result<i32, String> {
    s.trim().parse::<i32>().map_err(|e| format!("Bad input: {}", e))
}

println!("{:?}", parse_input("42"));
println!("{:?}", parse_input("abc"));
```

```rust
// The "parse or default" pattern
let configs = vec![
    ("port", "8080"),
    ("timeout", "abc"),   // bad value
    ("retries", "3"),
    ("workers", ""),      // empty
];

for (key, val) in &configs {
    let parsed: u32 = val.parse().unwrap_or_else(|_| {
        println!("  ⚠️ Invalid '{}' value '{}', using default", key, val);
        match *key {
            "port" => 3000,
            "timeout" => 30,
            "retries" => 1,
            "workers" => 4,
            _ => 0,
        }
    });
    println!("  {} = {}", key, parsed);
}
```

```rust
// Decision flowchart in code
fn error_strategy(error_type: &str) {
    match error_type {
        "array_out_of_bounds" => println!("  → PANIC: bug in your code"),
        "null_pointer"        => println!("  → PANIC: Rust prevents this with Option"),
        "file_not_found"      => println!("  → RESULT: expected, let caller decide"),
        "network_timeout"     => println!("  → RESULT: retry or report to user"),
        "invalid_user_input"  => println!("  → RESULT: show error message"),
        "impossible_state"    => println!("  → PANIC: unreachable!() or panic!()"),
        "division_by_zero"    => println!("  → RESULT: validate input first"),
        _ => println!("  → RESULT: when in doubt, use Result"),
    }
}

println!("Error Strategy Guide:\n");
for scenario in ["array_out_of_bounds", "file_not_found", "network_timeout",
                  "invalid_user_input", "impossible_state", "division_by_zero"] {
    print!("{:24}", scenario);
    error_strategy(scenario);
}
```

## 🎯 Key Takeaways

1. **Panic** for bugs and impossible states
2. **Result** for expected failures (I/O, parsing, validation)
3. `unwrap()`/`expect()` in tests and examples is OK
4. In production: always handle errors explicitly
5. When in doubt → use `Result`
