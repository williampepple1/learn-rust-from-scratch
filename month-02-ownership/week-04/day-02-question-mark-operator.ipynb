# 🦀 Day 2: The `?` Operator

---

```rust
// Without ? — verbose error handling
use std::fs;
use std::io;

fn read_file_verbose(path: &str) -> Result<String, io::Error> {
    let content = match fs::read_to_string(path) {
        Ok(c) => c,
        Err(e) => return Err(e),
    };
    Ok(content)
}

// With ? — same thing, much cleaner!
fn read_file_clean(path: &str) -> Result<String, io::Error> {
    let content = fs::read_to_string(path)?;  // Returns Err early if error
    Ok(content)
}

// Even shorter:
fn read_file_short(path: &str) -> Result<String, io::Error> {
    fs::read_to_string(path)
}

println!("? operator works identically — just less noise!");
```

```rust
// Chaining ? for multiple operations
fn read_first_line(path: &str) -> Result<String, io::Error> {
    let content = fs::read_to_string(path)?;
    let first = content.lines()
        .next()
        .ok_or(io::Error::new(io::ErrorKind::InvalidData, "Empty file"))?;
    Ok(first.to_string())
}

// Create a temp file to test
fs::write("test_q.txt", "First line\nSecond line").unwrap();
match read_first_line("test_q.txt") {
    Ok(line) => println!("First line: {}", line),
    Err(e) => println!("Error: {}", e),
}
fs::remove_file("test_q.txt").ok();
```

```rust
// ? with different error types — need conversion
fn parse_file_number(path: &str) -> Result<i32, Box<dyn std::error::Error>> {
    let content = fs::read_to_string(path)?;  // io::Error
    let number = content.trim().parse::<i32>()?;  // ParseIntError
    Ok(number)
}

// Box<dyn Error> accepts any error type!
fs::write("num.txt", "42").unwrap();
match parse_file_number("num.txt") {
    Ok(n) => println!("Parsed: {}", n),
    Err(e) => println!("Error: {}", e),
}
fs::remove_file("num.txt").ok();
```

```rust
// ? with Option — works too!
fn find_nth_word(text: &str, n: usize) -> Option<&str> {
    let word = text.split_whitespace().nth(n)?;
    Some(word)
}

println!("Word 0: {:?}", find_nth_word("hello world", 0));
println!("Word 5: {:?}", find_nth_word("hello world", 5));
```

```rust
// Practical: multi-step validation
fn process_user_input(input: &str) -> Result<String, Box<dyn std::error::Error>> {
    let trimmed = input.trim();
    if trimmed.is_empty() {
        return Err("Input cannot be empty".into());
    }
    let age: u32 = trimmed.parse()?;
    if age > 150 {
        return Err(format!("Age {} is unrealistic", age).into());
    }
    Ok(format!("Valid age: {}", age))
}

for input in ["25", "  42  ", "", "abc", "200"] {
    match process_user_input(input) {
        Ok(msg) => println!("✅ '{}' → {}", input, msg),
        Err(e) => println!("❌ '{}' → {}", input, e),
    }
}
```

## 🎯 Key Takeaways

1. `?` returns `Err` early if the `Result` is `Err`
2. `?` also works with `Option` (returns `None` early)
3. `?` can auto-convert error types via `From` trait
4. `Box<dyn Error>` accepts any error type (quick & dirty)
5. The function must return `Result` (or `Option`) to use `?`
