```rust
// Result refresher
fn divide(a: f64, b: f64) -> Result<f64, String> {
    if b == 0.0 {
        Err("Division by zero".to_string())
    } else {
        Ok(a / b)
    }
}

match divide(10.0, 3.0) {
    Ok(result) => println!("10/3 = {:.4}", result),
    Err(e) => println!("Error: {}", e),
}

match divide(10.0, 0.0) {
    Ok(result) => println!("10/0 = {}", result),
    Err(e) => println!("Error: {}", e),
}
```

```rust
// unwrap() — panics on Err (only use in examples/tests!)
let x: Result<i32, &str> = Ok(42);
println!("unwrap: {}", x.unwrap());

// expect() — panics with custom message
let y: Result<i32, &str> = Ok(100);
println!("expect: {}", y.expect("Should have a value"));

// unwrap_or() — default value on error
let bad: Result<i32, &str> = Err("oops");
println!("unwrap_or: {}", bad.unwrap_or(0));

// unwrap_or_else() — compute default lazily
let bad2: Result<i32, &str> = Err("oops");
println!("unwrap_or_else: {}", bad2.unwrap_or_else(|_| 42));

// unwrap_or_default() — uses Default trait
let bad3: Result<String, &str> = Err("oops");
println!("unwrap_or_default: '{}'", bad3.unwrap_or_default());
```

```rust
// Combinators: map, and_then, map_err
fn parse_and_double(s: &str) -> Result<i32, String> {
    s.parse::<i32>()
        .map(|n| n * 2)
        .map_err(|e| format!("Parse error for '{}': {}", s, e))
}

println!("{:?}", parse_and_double("21"));
println!("{:?}", parse_and_double("abc"));
```

```rust
// and_then for chaining operations that return Result
fn validate_age(s: &str) -> Result<u32, String> {
    s.parse::<u32>()
        .map_err(|_| format!("'{}' is not a number", s))
        .and_then(|age| {
            if age > 0 && age < 150 {
                Ok(age)
            } else {
                Err(format!("Age {} is out of range", age))
            }
        })
}

for input in ["25", "0", "200", "abc"] {
    println!("validate_age('{}') = {:?}", input, validate_age(input));
}
```

```rust
// Converting between Result and Option
let ok: Result<i32, &str> = Ok(42);
let opt: Option<i32> = ok.ok();  // Result → Option (discards error)
println!("ok() → {:?}", opt);

let some: Option<i32> = Some(42);
let res: Result<i32, &str> = some.ok_or("missing"); // Option → Result
println!("ok_or() → {:?}", res);

let none: Option<i32> = None;
let res2: Result<i32, String> = none.ok_or_else(|| "computed error".into());
println!("ok_or_else() → {:?}", res2);
```

## 🎯 Key Takeaways

1. `unwrap()` panics — use `expect()` or proper handling in production
2. `unwrap_or()`, `unwrap_or_else()`, `unwrap_or_default()` for safe defaults
3. `.map()` transforms `Ok` value, `.map_err()` transforms `Err` value
4. `.and_then()` chains operations that themselves return `Result`
5. Convert: `Result::ok()` → Option, `Option::ok_or()` → Result
