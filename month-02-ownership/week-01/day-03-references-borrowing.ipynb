## 📎 Immutable References (`&T`)

A reference lets you **use** a value without **owning** it.

```rust
// Instead of taking ownership...
fn calculate_length_owned(s: String) -> usize {
    s.len()
} // s is dropped! Caller loses it.

// ...borrow with a reference!
fn calculate_length(s: &String) -> usize {
    s.len()
} // s goes out of scope, but since it doesn't own the data, nothing is dropped

let s = String::from("hello");
let len = calculate_length(&s);  // & creates a reference
println!("'{}' has length {}", s, len); // s is still valid! ✅
```

```rust
// Multiple immutable references are fine!
let s = String::from("hello");
let r1 = &s;
let r2 = &s;
let r3 = &s;
println!("r1={}, r2={}, r3={}", r1, r2, r3); // All valid! ✅
println!("original: {}", s); // Also valid! ✅
```

```rust
// References cannot modify the data
fn try_modify(s: &String) {
    // s.push_str(" world"); // ❌ ERROR: cannot borrow as mutable
    println!("Can read: {}", s); // ✅ Reading is fine
}

let s = String::from("hello");
try_modify(&s);
println!("Still: {}", s);
```

## ✏️ Mutable References (`&mut T`)

```rust
fn append_world(s: &mut String) {
    s.push_str(" world");
}

let mut s = String::from("hello"); // Must be declared mut!
append_world(&mut s);              // Pass mutable reference
println!("{}", s);                 // "hello world" ✅
```

```rust
// Only ONE mutable reference at a time!
let mut s = String::from("hello");
let r1 = &mut s;
// let r2 = &mut s;  // ❌ ERROR: cannot borrow as mutable more than once
r1.push_str("!");
println!("{}", r1);
```

```rust
// Can't mix mutable and immutable references!
let mut s = String::from("hello");

// This works because r1 and r2 are used BEFORE r3 is created:
let r1 = &s;
let r2 = &s;
println!("{} and {}", r1, r2);
// r1 and r2 are no longer used after this point (NLL: Non-Lexical Lifetimes)

let r3 = &mut s;  // ✅ OK because r1, r2 are done
r3.push_str(" world");
println!("{}", r3);
```

## 📜 Borrowing Rules Summary

1. At any time, you can have **either**:
   - Any number of immutable references (`&T`), **OR**
   - Exactly one mutable reference (`&mut T`)
2. References must always be **valid** (no dangling)
3. A reference's scope ends at its **last use** (NLL)

## 📚 Practical: Functions with References

```rust
// Read-only operations: use &T
fn first_word(s: &String) -> &str {
    let bytes = s.as_bytes();
    for (i, &byte) in bytes.iter().enumerate() {
        if byte == b' ' {
            return &s[..i];
        }
    }
    &s[..]
}

// Modification: use &mut T
fn capitalize_first(s: &mut String) {
    if let Some(c) = s.chars().next() {
        let upper = c.to_uppercase().to_string();
        s.replace_range(..c.len_utf8(), &upper);
    }
}

let mut text = String::from("hello world");
println!("First word: {}", first_word(&text));
capitalize_first(&mut text);
println!("Capitalized: {}", text);
```

```rust
// Working with vectors by reference
fn sum(numbers: &[i32]) -> i32 {
    numbers.iter().sum()
}

fn double_all(numbers: &mut Vec<i32>) {
    for n in numbers.iter_mut() {
        *n *= 2;
    }
}

fn largest(numbers: &[i32]) -> Option<&i32> {
    numbers.iter().max()
}

let mut nums = vec![1, 2, 3, 4, 5];
println!("Sum: {}", sum(&nums));
println!("Largest: {:?}", largest(&nums));
double_all(&mut nums);
println!("Doubled: {:?}", nums);
```

## 🎯 Key Takeaways

1. `&T` borrows immutably — can have **many** at once
2. `&mut T` borrows mutably — can have **only one**
3. Can't mix `&T` and `&mut T` simultaneously
4. References prevent use-after-free and data races **at compile time**
5. Prefer borrowing over cloning for performance
