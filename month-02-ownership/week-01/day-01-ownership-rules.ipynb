## 📜 The Three Rules of Ownership

1. **Each value has exactly one owner** (a variable)
2. **There can only be one owner at a time**
3. **When the owner goes out of scope, the value is dropped** (freed)

````rust
// Rule 1 & 3: s owns the String, dropped at end of scope
{
    let s = String::from("hello"); // s owns this String
    println!("Inside scope: {}", s);
} // s goes out of scope → String is dropped (memory freed)

// println!("{}", s); // ❌ ERROR: s doesn't exist here
println!("s has been dropped!");
````

## 📦 Stack vs Heap

| | Stack | Heap |
|---|---|---|
| Speed | Very fast | Slower |
| Size | Known at compile time | Dynamic |
| Types | `i32`, `f64`, `bool`, `char`, tuples, arrays | `String`, `Vec`, `Box`, `HashMap` |
| Copy | Values are **copied** | Ownership is **moved** |

````rust
// Stack types: values are COPIED
let x = 5;       // x holds 5 on the stack
let y = x;        // y gets a COPY of 5
println!("x = {}, y = {}", x, y); // Both valid! ✅

// Heap types: ownership is MOVED
let s1 = String::from("hello"); // s1 owns the String on the heap
let s2 = s1;                     // ownership MOVES to s2
// println!("{}", s1);           // ❌ ERROR: s1 no longer owns the data
println!("s2 = {}", s2);         // ✅ s2 is the new owner
````

## 🔄 Ownership Moves — Visualized

```
let s1 = String::from("hello");

Stack:              Heap:
s1 ─────────────→  [h][e][l][l][o]
   ptr, len=5,
   capacity=5

let s2 = s1;  // MOVE!

Stack:              Heap:
s1 (invalid) ╌╌╌   [h][e][l][l][o]
s2 ─────────────→       ↑
   ptr, len=5,          │
   capacity=5    (same data, new owner)
```

````rust
// Move happens with assignment
let a = String::from("rust");
let b = a;  // a is moved to b
// a is now invalid
println!("b = {}", b);

// Move happens when passing to functions
fn take_ownership(s: String) {
    println!("I own: {}", s);
} // s is dropped here

let name = String::from("Alice");
take_ownership(name);
// println!("{}", name); // ❌ name was moved into the function!
````

````rust
// Move happens with return values too
fn give_ownership() -> String {
    let s = String::from("yours now");
    s  // ownership moves to the caller
}

fn take_and_give_back(s: String) -> String {
    println!("I borrowed: {}", s);
    s  // give it back!
}

let s1 = give_ownership();
println!("Got: {}", s1);

let s2 = String::from("hello");
let s3 = take_and_give_back(s2);
// s2 is invalid, s3 has the value
println!("Got back: {}", s3);
````

````rust
// Moves in collections
let mut v = vec![String::from("a"), String::from("b"), String::from("c")];

// Can't move out of a vector by indexing:
// let first = v[0];  // ❌ Would leave v[0] in invalid state

// Instead, use these:
let first = v.remove(0);           // Removes and returns ownership
println!("Removed: {}", first);
println!("Remaining: {:?}", v);

// Or swap_remove (faster, doesn't preserve order)
let mut v2 = vec![String::from("x"), String::from("y"), String::from("z")];
let item = v2.swap_remove(0);
println!("Swapped out: {}, remaining: {:?}", item, v2);
````

## 🔍 When Does a Move NOT Happen?

Types that implement the `Copy` trait are **copied** instead of moved.

````rust
// Copy types — all of these are still valid after "assignment"
let a: i32 = 42;
let b = a;  // COPY, not move
println!("a={}, b={}", a, b); // Both valid ✅

let x: f64 = 3.14;
let y = x;  // COPY
println!("x={}, y={}", x, y); // Both valid ✅

let flag: bool = true;
let flag2 = flag;  // COPY
println!("flag={}, flag2={}", flag, flag2); // Both valid ✅

let c: char = '🦀';
let c2 = c;  // COPY
println!("c={}, c2={}", c, c2); // Both valid ✅

// Tuples of Copy types are also Copy
let point = (3, 4);
let point2 = point;  // COPY
println!("point={:?}, point2={:?}", point, point2); // Both valid ✅

println!("\n📝 Copy types: all integer/float/bool/char types, and tuples/arrays containing only Copy types");
````

## 🧪 Ownership Detective

````rust
// Can you predict which lines would cause errors?

let s = String::from("hello");
let t = s;             // Q1: Is s still valid? → NO (moved)
let u = t.clone();     // Q2: Is t still valid? → YES (cloned)
println!("t={}, u={}", t, u);

let n = 42;
let m = n;             // Q3: Is n still valid? → YES (Copy type)
println!("n={}, m={}", n, m);

let v = vec![1, 2, 3];
let w = v;             // Q4: Is v still valid? → NO (Vec is moved)
// println!("{:?}", v); // ❌
println!("w={:?}", w);
````

## 🎯 Key Takeaways

1. Each value has **exactly one owner** at a time
2. When the owner goes out of scope, the value is **dropped**
3. Assignment of heap types **moves** ownership (invalidates the source)
4. Stack types (`i32`, `bool`, `f64`, etc.) implement `Copy` — they're **copied**
5. Use `.clone()` to explicitly deep-copy heap data
