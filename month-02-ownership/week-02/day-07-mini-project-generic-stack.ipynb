## 📚 Step 1: Core Stack Implementation

```rust
#[derive(Debug)]
struct Stack<T> {
    elements: Vec<T>,
    max_size: Option<usize>,
}

impl<T> Stack<T> {
    fn new() -> Self {
        Stack { elements: Vec::new(), max_size: None }
    }

    fn with_capacity(max: usize) -> Self {
        Stack { elements: Vec::with_capacity(max), max_size: Some(max) }
    }

    fn push(&mut self, item: T) -> Result<(), &'static str> {
        if let Some(max) = self.max_size {
            if self.elements.len() >= max {
                return Err("Stack overflow: max capacity reached");
            }
        }
        self.elements.push(item);
        Ok(())
    }

    fn pop(&mut self) -> Option<T> {
        self.elements.pop()
    }

    fn peek(&self) -> Option<&T> {
        self.elements.last()
    }

    fn peek_mut(&mut self) -> Option<&mut T> {
        self.elements.last_mut()
    }

    fn is_empty(&self) -> bool {
        self.elements.is_empty()
    }

    fn size(&self) -> usize {
        self.elements.len()
    }

    fn clear(&mut self) {
        self.elements.clear();
    }
}

println!("Stack<T> implemented! ✅");
```

## 🔄 Step 2: Iterator Support

```rust
// IntoIterator — consume the stack
impl<T> IntoIterator for Stack<T> {
    type Item = T;
    type IntoIter = std::vec::IntoIter<T>;

    fn into_iter(self) -> Self::IntoIter {
        self.elements.into_iter()
    }
}

// Drain method — pop all elements in LIFO order
impl<T> Stack<T> {
    fn drain(&mut self) -> Vec<T> {
        let mut result = Vec::new();
        while let Some(item) = self.pop() {
            result.push(item);
        }
        result
    }

    fn as_slice(&self) -> &[T] {
        &self.elements
    }
}

println!("Iterator support added! ✅");
```

## 🖨️ Step 3: Display for Displayable Types

```rust
use std::fmt;

impl<T: fmt::Display> fmt::Display for Stack<T> {
    fn fmt(&self, f: &mut fmt::Formatter) -> fmt::Result {
        write!(f, "Stack[")?;
        for (i, item) in self.elements.iter().enumerate() {
            if i > 0 { write!(f, ", ")?; }
            write!(f, "{}", item)?;
        }
        write!(f, "] (top →)")
    }
}

println!("Display trait implemented! ✅");
```

## 🧪 Step 4: Test It!

```rust
// Test with integers
let mut int_stack: Stack<i32> = Stack::new();
for i in 1..=5 {
    int_stack.push(i).unwrap();
}
println!("Int stack: {}", int_stack);
println!("Peek: {:?}", int_stack.peek());
println!("Pop: {:?}", int_stack.pop());
println!("After pop: {}\n", int_stack);
```

```rust
// Test with strings
let mut str_stack: Stack<String> = Stack::new();
str_stack.push("hello".into()).unwrap();
str_stack.push("world".into()).unwrap();
str_stack.push("rust".into()).unwrap();
println!("String stack: {}", str_stack);

while let Some(word) = str_stack.pop() {
    println!("  Popped: {}", word);
}
println!("Empty: {}\n", str_stack.is_empty());
```

```rust
// Test with capacity limit
let mut bounded: Stack<char> = Stack::with_capacity(3);
bounded.push('a').unwrap();
bounded.push('b').unwrap();
bounded.push('c').unwrap();
println!("Bounded: {}", bounded);
match bounded.push('d') {
    Ok(()) => println!("Added d"),
    Err(e) => println!("Error: {}", e),
}
```

## 🔧 Step 5: Practical — Balanced Parentheses

```rust
fn is_balanced(s: &str) -> bool {
    let mut stack: Stack<char> = Stack::new();
    for ch in s.chars() {
        match ch {
            '(' | '[' | '{' => { stack.push(ch).unwrap(); },
            ')' => if stack.pop() != Some('(') { return false; },
            ']' => if stack.pop() != Some('[') { return false; },
            '}' => if stack.pop() != Some('{') { return false; },
            _ => {}
        }
    }
    stack.is_empty()
}

let tests = [
    ("(())", true), ("[{()}]", true), ("(()", false),
    (")(", false), ("{[()]}", true), ("a + (b * [c])", true),
];

for (input, expected) in &tests {
    let result = is_balanced(input);
    let status = if result == *expected { "✅" } else { "❌" };
    println!("{} '{}' → balanced={}", status, input, result);
}
```

```rust
// Assertions (manual tests)
assert!(Stack::<i32>::new().is_empty());
assert_eq!(Stack::<i32>::new().size(), 0);

let mut s = Stack::new();
s.push(1).unwrap();
s.push(2).unwrap();
assert_eq!(s.size(), 2);
assert_eq!(s.peek(), Some(&2));
assert_eq!(s.pop(), Some(2));
assert_eq!(s.pop(), Some(1));
assert_eq!(s.pop(), None);
assert!(s.is_empty());

let mut bounded = Stack::with_capacity(1);
assert!(bounded.push(1).is_ok());
assert!(bounded.push(2).is_err());

println!("All tests passed! ✅");
```

## 🎯 Key Takeaways

1. Generic types enable **code reuse** without duplication
2. `impl<T>` for universal methods; `impl<T: Bound>` for constrained ones
3. Implement standard traits (`Display`, `IntoIterator`) for ergonomic APIs
4. Use `Result` for operations that can fail (like bounded push)
5. Generics are **zero-cost** — compiled to concrete types
