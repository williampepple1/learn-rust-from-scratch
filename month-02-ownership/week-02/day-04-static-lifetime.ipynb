# 🦀 Day 4: `'static` Lifetime & Lifetime Bounds

---

```rust
// 'static means "lives for the entire program"

// String literals are 'static
let s: &'static str = "I live forever";
println!("{}", s);

// Constants are 'static
const PI: f64 = 3.14159;
static APP_NAME: &str = "MyApp";
println!("{}: {}", APP_NAME, PI);
```

```rust
// 'static as a trait bound: the TYPE doesn't contain non-static references
// It does NOT mean the value lives forever!

fn print_static<T: std::fmt::Display + 'static>(val: T) {
    println!("Value: {}", val);
}

print_static(42);                     // ✅ i32 is 'static (owns its data)
print_static(String::from("hello"));  // ✅ String is 'static (owns its data)
print_static("literal");              // ✅ &'static str

// Won't work with non-static references:
// let local = String::from("temp");
// let r = &local;
// print_static(r);  // ❌ &String is not 'static
```

```rust
// Common misconception: 'static doesn't mean "never freed"
// Owned types like String satisfy 'static because they don't borrow

fn needs_static<T: 'static>(val: T) -> T {
    val  // T doesn't contain borrowed data
}

let s = needs_static(String::from("hello")); // ✅ String owns its data
println!("{}", s);

let n = needs_static(vec![1, 2, 3]); // ✅ Vec owns its data
println!("{:?}", n);
```

## 🔗 Lifetime Bounds on Generics

```rust
// T: 'a means "T must live at least as long as 'a"

#[derive(Debug)]
struct Ref<'a, T: 'a> {
    data: &'a T,
}

let value = 42;
let r = Ref { data: &value };
println!("{:?}", r);

let text = String::from("hello");
let r2 = Ref { data: &text };
println!("{:?}", r2);
```

```rust
// Combining lifetime and trait bounds
use std::fmt::Display;

fn announce_and_return<'a, T: Display>(value: &'a T, msg: &str) -> &'a T {
    println!("{}: {}", msg, value);
    value
}

let x = 42;
let result = announce_and_return(&x, "The answer is");
println!("Returned: {}", result);
```

## ⚠️ When to use `'static`

| Use Case | Example |
|----------|--------|
| String literals | `let s: &'static str = "hello"` |
| Thread spawning | `std::thread::spawn` requires `'static` |
| Error types | `Box<dyn Error + 'static>` |
| Lazy statics | Global state |

**Avoid** `'static` when a shorter lifetime would work!

## 🎯 Key Takeaways

1. `'static` means **lives for the entire program** (string literals, constants)
2. `T: 'static` means T **owns** all its data (no non-static borrows)
3. `String`, `Vec<T>`, `i32` are all `'static` because they own their data
4. `&'a String` is NOT `'static` (it borrows from something local)
5. Use `'static` sparingly — prefer shorter lifetimes when possible
