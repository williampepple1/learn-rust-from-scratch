```rust
// Generic struct
#[derive(Debug)]
struct Wrapper<T> {
    value: T,
}

impl<T> Wrapper<T> {
    fn new(value: T) -> Self {
        Wrapper { value }
    }

    fn into_inner(self) -> T {
        self.value
    }

    fn get(&self) -> &T {
        &self.value
    }
}

// Specific impl only for Display types
impl<T: std::fmt::Display> Wrapper<T> {
    fn print(&self) {
        println!("Wrapped: {}", self.value);
    }
}

let w1 = Wrapper::new(42);
let w2 = Wrapper::new("hello");
let w3 = Wrapper::new(vec![1, 2, 3]);
w1.print();
w2.print();
// w3.print(); // ❌ Vec doesn't implement Display
println!("w3: {:?}", w3);
```

```rust
// Multiple type parameters
#[derive(Debug)]
struct Pair<A, B> {
    first: A,
    second: B,
}

impl<A, B> Pair<A, B> {
    fn new(first: A, second: B) -> Self {
        Pair { first, second }
    }

    fn swap(self) -> Pair<B, A> {
        Pair { first: self.second, second: self.first }
    }
}

impl<A: std::fmt::Display, B: std::fmt::Display> Pair<A, B> {
    fn display(&self) {
        println!("({}, {})", self.first, self.second);
    }
}

let p = Pair::new("age", 30);
p.display();
let swapped = p.swap();
swapped.display();
```

```rust
// Generic enums — you already know Option and Result!

// Custom generic enum
#[derive(Debug)]
enum Validated<T> {
    Valid(T),
    Invalid(String),
    Pending,
}

impl<T: std::fmt::Display> Validated<T> {
    fn unwrap_or(self, default: T) -> T {
        match self {
            Validated::Valid(v) => v,
            _ => default,
        }
    }

    fn map<U, F: FnOnce(T) -> U>(self, f: F) -> Validated<U> {
        match self {
            Validated::Valid(v) => Validated::Valid(f(v)),
            Validated::Invalid(e) => Validated::Invalid(e),
            Validated::Pending => Validated::Pending,
        }
    }
}

let age: Validated<u32> = Validated::Valid(25);
let name: Validated<String> = Validated::Invalid("Name required".into());
let status: Validated<bool> = Validated::Pending;

println!("Age: {}", age.unwrap_or(0));
println!("Name: {}", name.unwrap_or("Unknown".into()));
println!("Status: {}", status.unwrap_or(false));
```

```rust
// Generic struct with trait bounds for specific behavior
#[derive(Debug)]
struct SortedVec<T: Ord> {
    data: Vec<T>,
}

impl<T: Ord> SortedVec<T> {
    fn new() -> Self {
        SortedVec { data: Vec::new() }
    }

    fn insert(&mut self, item: T) {
        let pos = self.data.binary_search(&item).unwrap_or_else(|e| e);
        self.data.insert(pos, item);
    }

    fn contains(&self, item: &T) -> bool {
        self.data.binary_search(item).is_ok()
    }

    fn as_slice(&self) -> &[T] {
        &self.data
    }
}

let mut sorted = SortedVec::new();
for n in [5, 3, 8, 1, 9, 2] {
    sorted.insert(n);
}
println!("Sorted: {:?}", sorted.as_slice());
println!("Contains 3: {}", sorted.contains(&3));
println!("Contains 7: {}", sorted.contains(&7));
```

## 🎯 Key Takeaways

1. Structs and enums can be generic: `struct Foo<T>`
2. `impl<T> Foo<T>` implements methods for all T
3. `impl Foo<i32>` implements methods only for a specific type
4. `impl<T: Bound> Foo<T>` implements for T with trait bounds
5. `Option<T>` and `Result<T, E>` are generic enums you use daily!
