```rust
// Without generics — duplication!
fn largest_i32(list: &[i32]) -> &i32 {
    let mut largest = &list[0];
    for item in &list[1..] {
        if item > largest { largest = item; }
    }
    largest
}

fn largest_f64(list: &[f64]) -> &f64 {
    let mut largest = &list[0];
    for item in &list[1..] {
        if item > largest { largest = item; }
    }
    largest
}

println!("Largest i32: {}", largest_i32(&[1, 5, 3, 9, 2]));
println!("Largest f64: {}", largest_f64(&[1.5, 3.7, 2.1]));
println!("\n😩 Same logic twice! Let's use generics...");
```

```rust
// With generics — one function for all comparable types!
fn largest<T: PartialOrd>(list: &[T]) -> &T {
    let mut largest = &list[0];
    for item in &list[1..] {
        if item > largest { largest = item; }
    }
    largest
}

println!("Largest i32: {}", largest(&[1, 5, 3, 9, 2]));
println!("Largest f64: {}", largest(&[1.5, 3.7, 2.1]));
println!("Largest char: {}", largest(&['a', 'z', 'm']));
println!("Largest str: {}", largest(&["apple", "banana", "cherry"]));
```

## 🧩 Generic Function Syntax

```rust
// Basic generic function
fn identity<T>(value: T) -> T {
    value
}

println!("{}", identity(42));
println!("{}", identity("hello"));
println!("{}", identity(3.14));
```

```rust
// Multiple type parameters
fn pair<A, B>(a: A, b: B) -> (A, B) {
    (a, b)
}

let p = pair("name", 42);
println!("{:?}", p);

let p2 = pair(true, vec![1, 2, 3]);
println!("{:?}", p2);
```

```rust
// With trait bounds — constrain what T can be
use std::fmt::Display;

fn print_labeled<T: Display>(label: &str, value: &T) {
    println!("{}: {}", label, value);
}

fn sum_all<T: std::ops::Add<Output = T> + Copy>(items: &[T]) -> T {
    let mut total = items[0];
    for &item in &items[1..] {
        total = total + item;
    }
    total
}

print_labeled("Number", &42);
print_labeled("Text", &"hello");

println!("Sum ints: {}", sum_all(&[1, 2, 3, 4, 5]));
println!("Sum floats: {}", sum_all(&[1.0, 2.0, 3.0]));
```

```rust
// where clause — cleaner for complex bounds
fn complex_fn<T, U>(t: &T, u: &U) -> String
where
    T: Display + Clone,
    U: Display + std::fmt::Debug,
{
    format!("{} and {:?}", t, u)
}

println!("{}", complex_fn(&42, &"hello"));
```

```rust
// Practical example: generic search
fn find_first<T: PartialEq>(items: &[T], target: &T) -> Option<usize> {
    items.iter().position(|item| item == target)
}

fn contains<T: PartialEq>(items: &[T], target: &T) -> bool {
    items.iter().any(|item| item == target)
}

println!("Find 3 in ints: {:?}", find_first(&[1, 2, 3, 4], &3));
println!("Find 'b' in chars: {:?}", find_first(&['a', 'b', 'c'], &'b'));
println!("Contains 'hello': {}", contains(&["hello", "world"], &"hello"));
```

## 🎯 Key Takeaways

1. `<T>` declares a type parameter — works with any type
2. Trait bounds (`T: Display`) constrain what operations T supports
3. `where` clauses improve readability for complex bounds
4. Generics have **zero runtime cost** — monomorphization at compile time
5. Common bounds: `PartialOrd`, `PartialEq`, `Display`, `Debug`, `Clone`, `Copy`
