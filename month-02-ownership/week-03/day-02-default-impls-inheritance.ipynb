```rust
// Traits with rich default implementations
trait Logger {
    fn name(&self) -> &str;
    fn log_level(&self) -> &str { "INFO" } // default

    fn log(&self, message: &str) {
        println!("[{}][{}] {}", self.log_level(), self.name(), message);
    }

    fn warn(&self, message: &str) {
        println!("[WARN][{}] ⚠️ {}", self.name(), message);
    }

    fn error(&self, message: &str) {
        println!("[ERROR][{}] ❌ {}", self.name(), message);
    }
}

struct AppLogger;
impl Logger for AppLogger {
    fn name(&self) -> &str { "App" }
    // Uses ALL defaults
}

struct DebugLogger;
impl Logger for DebugLogger {
    fn name(&self) -> &str { "Debug" }
    fn log_level(&self) -> &str { "DEBUG" } // Override
}

let app = AppLogger;
app.log("Started");
app.warn("Low memory");

let dbg = DebugLogger;
dbg.log("Trace point hit");
```

## 🧬 Trait Inheritance (Supertraits)

```rust
use std::fmt;

// Printable requires Display
trait Printable: fmt::Display {
    fn print(&self) {
        println!("{}", self); // Can use Display because it's a supertrait
    }

    fn print_labeled(&self, label: &str) {
        println!("{}: {}", label, self);
    }
}

struct Temperature {
    celsius: f64,
}

impl fmt::Display for Temperature {
    fn fmt(&self, f: &mut fmt::Formatter) -> fmt::Result {
        write!(f, "{:.1}°C", self.celsius)
    }
}

impl Printable for Temperature {}

let temp = Temperature { celsius: 23.5 };
temp.print();
temp.print_labeled("Room temp");
```

```rust
// Chain of supertraits
trait Named {
    fn name(&self) -> &str;
}

trait Aged {
    fn age(&self) -> u32;
}

// Person requires both Named and Aged
trait Person: Named + Aged {
    fn introduce(&self) -> String {
        format!("Hi, I'm {} and I'm {} years old.", self.name(), self.age())
    }
}

struct Student {
    name: String,
    age: u32,
    grade: String,
}

impl Named for Student {
    fn name(&self) -> &str { &self.name }
}

impl Aged for Student {
    fn age(&self) -> u32 { self.age }
}

impl Person for Student {} // Gets introduce() for free

let s = Student { name: "Alice".into(), age: 20, grade: "A".into() };
println!("{}", s.introduce());
```

```rust
// Default methods that call other default methods
trait Validator {
    fn validate(&self) -> Vec<String>; // Must implement

    fn is_valid(&self) -> bool {
        self.validate().is_empty()
    }

    fn validation_report(&self) -> String {
        let errors = self.validate();
        if errors.is_empty() {
            "✅ Valid".to_string()
        } else {
            format!("❌ {} errors:\n{}", errors.len(),
                errors.iter().map(|e| format!("  - {}", e)).collect::<Vec<_>>().join("\n"))
        }
    }
}

struct Email { address: String }

impl Validator for Email {
    fn validate(&self) -> Vec<String> {
        let mut errors = Vec::new();
        if !self.address.contains('@') { errors.push("Missing @".into()); }
        if !self.address.contains('.') { errors.push("Missing domain".into()); }
        if self.address.len() < 5 { errors.push("Too short".into()); }
        errors
    }
}

let good = Email { address: "user@example.com".into() };
let bad = Email { address: "bad".into() };
println!("{}", good.validation_report());
println!("{}", bad.validation_report());
```

## 🎯 Key Takeaways

1. Default methods reduce boilerplate — implementors only define what's unique
2. Default methods can call other trait methods (even non-default ones)
3. `trait B: A` means B **requires** A — supertrait relationship
4. Supertraits enable using A's methods in B's default implementations
5. Chain supertraits for complex behavior hierarchies
