```rust
// Define a trait — a set of behaviors
trait Describable {
    fn describe(&self) -> String;
}

trait Printable {
    fn print(&self) {
        println!("[Printable item]");  // default implementation
    }
}

// Implement traits for types
struct Book {
    title: String,
    author: String,
    pages: u32,
}

impl Describable for Book {
    fn describe(&self) -> String {
        format!("'{}' by {} ({} pages)", self.title, self.author, self.pages)
    }
}

impl Printable for Book {
    fn print(&self) {
        println!("📚 {}", self.describe());
    }
}

struct Movie {
    title: String,
    director: String,
    runtime_min: u32,
}

impl Describable for Movie {
    fn describe(&self) -> String {
        format!("'{}' dir. {} ({}min)", self.title, self.director, self.runtime_min)
    }
}

impl Printable for Movie {} // Uses default implementation

let book = Book { title: "Dune".into(), author: "Herbert".into(), pages: 412 };
let movie = Movie { title: "Blade Runner".into(), director: "Scott".into(), runtime_min: 117 };

book.print();
movie.print();
println!("Book: {}", book.describe());
println!("Movie: {}", movie.describe());
```

```rust
// Traits with multiple methods
trait Summary {
    fn title(&self) -> &str;
    fn author(&self) -> &str;
    fn summary(&self) -> String {
        format!("{} by {}", self.title(), self.author())
    }
}

struct Article {
    headline: String,
    writer: String,
    content: String,
}

impl Summary for Article {
    fn title(&self) -> &str { &self.headline }
    fn author(&self) -> &str { &self.writer }
    // Uses default summary()
}

struct Tweet {
    username: String,
    text: String,
}

impl Summary for Tweet {
    fn title(&self) -> &str { &self.text }
    fn author(&self) -> &str { &self.username }
    fn summary(&self) -> String {
        format!("@{}: {}", self.username, self.text)  // Override default
    }
}

let article = Article {
    headline: "Rust 2026".into(),
    writer: "Ferris".into(),
    content: "Big news...".into(),
};
let tweet = Tweet { username: "rustlang".into(), text: "Rust is awesome!".into() };

println!("{}", article.summary());
println!("{}", tweet.summary());
```

```rust
// Using traits as function parameters
fn notify(item: &impl Summary) {
    println!("Breaking: {}", item.summary());
}

// Equivalent with trait bound syntax:
fn notify_verbose<T: Summary>(item: &T) {
    println!("📢 {}", item.summary());
}

notify(&article);
notify(&tweet);
notify_verbose(&article);
```

```rust
// Implementing traits for existing types
trait IsEven {
    fn is_even(&self) -> bool;
}

impl IsEven for i32 {
    fn is_even(&self) -> bool {
        self % 2 == 0
    }
}

impl IsEven for u64 {
    fn is_even(&self) -> bool {
        self % 2 == 0
    }
}

println!("42.is_even() = {}", 42_i32.is_even());
println!("7.is_even() = {}", 7_i32.is_even());
```

## 🎯 Key Takeaways

1. Traits define **shared behavior** across types
2. `trait Foo { fn bar(&self); }` defines the interface
3. `impl Foo for Type` provides the implementation
4. Default methods can be overridden
5. You can implement your traits on external types (and vice versa)
