# 🦀 Day 6: Deriving Traits

---

```rust
// #[derive] auto-implements traits based on struct fields

#[derive(Debug, Clone, PartialEq, Eq, Hash)]
struct Player {
    name: String,
    score: i32,
    level: u32,
}

let p1 = Player { name: "Hero".into(), score: 100, level: 5 };
let p2 = p1.clone();

println!("Debug: {:?}", p1);
println!("Equal: {}", p1 == p2);

let p3 = Player { name: "Hero".into(), score: 200, level: 5 };
println!("p1 == p3: {}", p1 == p3); // false — different score
```

```rust
// What you can derive:
// Debug   — {:?} formatting
// Clone   — .clone() deep copy
// Copy    — implicit copy (only if ALL fields are Copy)
// PartialEq — == and !=
// Eq      — full equality (marker, requires PartialEq)
// PartialOrd — <, >, <=, >=
// Ord     — total ordering (requires Eq + PartialOrd)
// Hash    — for use in HashMap/HashSet keys
// Default — ::default() with field defaults

#[derive(Debug, Clone, Copy, PartialEq, PartialOrd)]
struct Vec2 {
    x: f64,
    y: f64,
}

let a = Vec2 { x: 1.0, y: 2.0 };
let b = a;  // Copy! Both valid
println!("a={:?}, b={:?}", a, b);
println!("a == b: {}", a == b);
```

```rust
// Default derive — all fields use their type's Default
#[derive(Debug, Default)]
struct Settings {
    volume: f32,       // Default: 0.0
    fullscreen: bool,  // Default: false
    username: String,  // Default: ""
    retries: u32,      // Default: 0
}

let s = Settings::default();
println!("{:?}", s);

// Partial override with struct update
let custom = Settings {
    volume: 0.8,
    username: "Player1".into(),
    ..Settings::default()
};
println!("{:?}", custom);
```

```rust
// Deriving on enums
#[derive(Debug, Clone, Copy, PartialEq, Eq, Hash)]
enum Direction {
    North,
    South,
    East,
    West,
}

let d = Direction::North;
let d2 = d;  // Copy!
println!("{:?} == {:?}: {}", d, d2, d == d2);

// Use as HashMap key
use std::collections::HashMap;
let mut distances: HashMap<Direction, f64> = HashMap::new();
distances.insert(Direction::North, 10.5);
distances.insert(Direction::East, 7.2);
println!("Distances: {:?}", distances);
```

```rust
// When derive doesn't work — implement manually
#[derive(Debug)]
struct CaseInsensitiveString(String);

impl PartialEq for CaseInsensitiveString {
    fn eq(&self, other: &Self) -> bool {
        self.0.to_lowercase() == other.0.to_lowercase()
    }
}
impl Eq for CaseInsensitiveString {}

let a = CaseInsensitiveString("Hello".into());
let b = CaseInsensitiveString("hello".into());
let c = CaseInsensitiveString("HELLO".into());
println!("a == b: {}", a == b); // true!
println!("a == c: {}", a == c); // true!
println!("b == c: {}", b == c); // true!
```

## 🎯 Key Takeaways

1. `#[derive(...)]` auto-implements traits from field types
2. Common derives: `Debug, Clone, PartialEq, Eq, Hash`
3. `Copy` only works if ALL fields are `Copy`
4. `Default` gives zero/empty values for all primitive types
5. Implement manually when you need custom behavior (e.g., case-insensitive equality)
