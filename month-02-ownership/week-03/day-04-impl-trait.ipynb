# 🦀 Day 4: `impl Trait` Syntax

---

```rust
use std::fmt::Display;

// impl Trait in argument position
fn notify(item: &impl Display) {
    println!("Notification: {}", item);
}
// Equivalent to: fn notify<T: Display>(item: &T)

notify(&42);
notify(&"breaking news");
```

```rust
// impl Trait in return position — return any type implementing the trait
fn make_iterator(up_to: i32) -> impl Iterator<Item = i32> {
    (0..up_to).filter(|x| x % 2 == 0)
}

fn make_greeting(name: &str) -> impl Display {
    format!("Hello, {}! Welcome.", name)
}

let evens: Vec<i32> = make_iterator(10).collect();
println!("Evens: {:?}", evens);
println!("{}", make_greeting("Rustacean"));
```

```rust
// Practical: returning closures
fn make_adder(n: i32) -> impl Fn(i32) -> i32 {
    move |x| x + n
}

fn make_multiplier(n: i32) -> impl Fn(i32) -> i32 {
    move |x| x * n
}

let add5 = make_adder(5);
let mul3 = make_multiplier(3);
println!("add5(10) = {}", add5(10));
println!("mul3(10) = {}", mul3(10));

// Chaining
let result = mul3(add5(7));
println!("mul3(add5(7)) = {}", result);
```

```rust
// impl Trait vs dyn Trait
// impl Trait: Compile-time dispatch (static, faster, no allocation)
// dyn Trait: Runtime dispatch (dynamic, flexible, needs Box/&)

// Static dispatch — compiler generates specialized code
fn process_static(item: &impl Display) {
    println!("Static: {}", item);
}

// Dynamic dispatch — uses vtable at runtime
fn process_dynamic(item: &dyn Display) {
    println!("Dynamic: {}", item);
}

process_static(&42);      // Calls process_static::<i32>
process_static(&"hello"); // Calls process_static::<&str>
process_dynamic(&42);     // Same function, different vtable
process_dynamic(&"hello"); // Same function, different vtable

// impl Trait = can only return ONE concrete type
// dyn Trait = can return different types
```

```rust
// When you need different return types → use dyn Trait
trait Animal {
    fn speak(&self) -> &str;
}

struct Dog;
struct Cat;

impl Animal for Dog { fn speak(&self) -> &str { "Woof!" } }
impl Animal for Cat { fn speak(&self) -> &str { "Meow!" } }

// Can't use impl Trait here (different concrete types)
fn get_pet(dog_person: bool) -> Box<dyn Animal> {
    if dog_person { Box::new(Dog) } else { Box::new(Cat) }
}

let pet = get_pet(true);
println!("{}", pet.speak());
let pet2 = get_pet(false);
println!("{}", pet2.speak());
```

## 🎯 Key Takeaways

1. `impl Trait` in args = sugar for `<T: Trait>` (static dispatch)
2. `impl Trait` in return = return a concrete type the compiler knows
3. `dyn Trait` = runtime dispatch via vtable (needs `Box` or `&`)
4. Use `impl Trait` when the return type is always the same
5. Use `Box<dyn Trait>` when returning different types
