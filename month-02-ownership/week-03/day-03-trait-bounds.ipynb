```rust
use std::fmt::{Display, Debug};

// Basic trait bound
fn print_twice<T: Display>(item: &T) {
    println!("{0} | {0}", item);
}

// Multiple bounds with +
fn debug_and_display<T: Display + Debug>(item: &T) {
    println!("Display: {}  Debug: {:?}", item, item);
}

print_twice(&42);
print_twice(&"hello");
debug_and_display(&42);
debug_and_display(&"text");
```

```rust
// where clause for complex bounds
fn process<T, U>(t: &T, u: &U) -> String
where
    T: Display + Clone,
    U: Debug + PartialEq,
{
    format!("T={}, U={:?}", t, u)
}

println!("{}", process(&42, &"hello"));
```

```rust
// Conditional method implementations
#[derive(Debug)]
struct Collection<T> {
    items: Vec<T>,
}

impl<T> Collection<T> {
    fn new() -> Self { Collection { items: Vec::new() } }
    fn add(&mut self, item: T) { self.items.push(item); }
    fn len(&self) -> usize { self.items.len() }
}

// Only for Display types
impl<T: Display> Collection<T> {
    fn print_all(&self) {
        for (i, item) in self.items.iter().enumerate() {
            println!("  [{}] {}", i, item);
        }
    }
}

// Only for sortable types
impl<T: Ord> Collection<T> {
    fn sorted(&self) -> Vec<&T> {
        let mut sorted: Vec<&T> = self.items.iter().collect();
        sorted.sort();
        sorted
    }
}

// Only for comparable types
impl<T: PartialEq> Collection<T> {
    fn contains(&self, target: &T) -> bool {
        self.items.iter().any(|item| item == target)
    }
}

let mut nums = Collection::new();
for n in [3, 1, 4, 1, 5] { nums.add(n); }
nums.print_all();
println!("Sorted: {:?}", nums.sorted());
println!("Contains 4: {}", nums.contains(&4));
```

```rust
// Returning types that implement traits
fn make_greeting(formal: bool) -> Box<dyn Display> {
    if formal {
        Box::new(format!("Good day, esteemed user."))
    } else {
        Box::new(format!("Hey there!"))
    }
}

println!("{}", make_greeting(true));
println!("{}", make_greeting(false));
```

## 🎯 Key Takeaways

1. `T: Trait` constrains generic T to types implementing Trait
2. `T: A + B` requires multiple traits
3. `where` clause improves readability for complex bounds
4. Conditional impls: `impl<T: Bound> Foo<T>` — methods only for matching T
5. `Box<dyn Trait>` for returning different types with same trait
