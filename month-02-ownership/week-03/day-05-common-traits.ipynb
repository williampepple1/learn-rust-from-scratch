```rust
// Display — user-facing string representation
use std::fmt;

struct Color { r: u8, g: u8, b: u8 }

impl fmt::Display for Color {
    fn fmt(&self, f: &mut fmt::Formatter) -> fmt::Result {
        write!(f, "#{:02X}{:02X}{:02X}", self.r, self.g, self.b)
    }
}

impl fmt::Debug for Color {
    fn fmt(&self, f: &mut fmt::Formatter) -> fmt::Result {
        write!(f, "Color(r={}, g={}, b={})", self.r, self.g, self.b)
    }
}

let red = Color { r: 255, g: 0, b: 0 };
println!("Display: {}", red);
println!("Debug: {:?}", red);
```

```rust
// Clone and PartialEq
#[derive(Debug, Clone, PartialEq)]
struct Point { x: f64, y: f64 }

impl Point {
    fn distance_to(&self, other: &Point) -> f64 {
        ((self.x - other.x).powi(2) + (self.y - other.y).powi(2)).sqrt()
    }
}

let p1 = Point { x: 0.0, y: 0.0 };
let p2 = p1.clone();
println!("p1 == p2: {}", p1 == p2);

let p3 = Point { x: 3.0, y: 4.0 };
println!("p1 == p3: {}", p1 == p3);
println!("Distance: {:.2}", p1.distance_to(&p3));
```

```rust
// From / Into — type conversions

struct Celsius(f64);
struct Fahrenheit(f64);

impl From<Celsius> for Fahrenheit {
    fn from(c: Celsius) -> Self {
        Fahrenheit(c.0 * 9.0 / 5.0 + 32.0)
    }
}

impl From<Fahrenheit> for Celsius {
    fn from(f: Fahrenheit) -> Self {
        Celsius((f.0 - 32.0) * 5.0 / 9.0)
    }
}

impl fmt::Display for Celsius {
    fn fmt(&self, f: &mut fmt::Formatter) -> fmt::Result {
        write!(f, "{:.1}°C", self.0)
    }
}

impl fmt::Display for Fahrenheit {
    fn fmt(&self, f: &mut fmt::Formatter) -> fmt::Result {
        write!(f, "{:.1}°F", self.0)
    }
}

let boiling = Celsius(100.0);
let f: Fahrenheit = boiling.into();  // Into is free when you implement From
println!("Boiling: {}", f);

let body = Fahrenheit(98.6);
let c = Celsius::from(body);
println!("Body temp: {}", c);
```

```rust
// Default — provides default values
#[derive(Debug)]
struct Config {
    host: String,
    port: u16,
    max_connections: usize,
    debug: bool,
}

impl Default for Config {
    fn default() -> Self {
        Config {
            host: "localhost".to_string(),
            port: 8080,
            max_connections: 100,
            debug: false,
        }
    }
}

let default_config = Config::default();
println!("Default: {:?}", default_config);

// Override just what you need:
let custom = Config {
    port: 3000,
    debug: true,
    ..Config::default()
};
println!("Custom: {:?}", custom);
```

```rust
// PartialOrd / Ord for comparison and sorting
#[derive(Debug, PartialEq, Eq, PartialOrd, Ord, Clone)]
struct Score {
    points: u32,
    name: String,
}

let mut scores = vec![
    Score { points: 85, name: "Alice".into() },
    Score { points: 92, name: "Bob".into() },
    Score { points: 78, name: "Charlie".into() },
];

scores.sort();  // Uses Ord — sorts by points first, then name
for s in &scores {
    println!("  {} — {} pts", s.name, s.points);
}
```

## 🎯 Key Takeaways

| Trait | Purpose |
|-------|--------|
| `Display` | User-facing string (used by `{}`) |
| `Debug` | Developer-facing string (used by `{:?}`) |
| `Clone` | Explicit deep copy (`.clone()`) |
| `Copy` | Implicit bitwise copy |
| `PartialEq`/`Eq` | Equality comparison (`==`) |
| `PartialOrd`/`Ord` | Ordering comparison (`<`, `>`, `.sort()`) |
| `From`/`Into` | Type conversions |
| `Default` | Default values |
