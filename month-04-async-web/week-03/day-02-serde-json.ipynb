## What is serde?

**serde** is Rust's premier serialization/deserialization framework. It lets you convert Rust structs and enums to/from JSON, TOML, YAML, and more with minimal boilerplate.

**serde_json** is the JSON implementation — the most common format for web APIs.

```rust
:dep serde = { version = "1", features = ["derive"] }
:dep serde_json = "1"

// Dependencies loaded.
```

## #[derive(Serialize, Deserialize)]

Add these derives to your structs and enums to enable JSON conversion.

```rust
use serde::{Deserialize, Serialize};

#[derive(Debug, Serialize, Deserialize)]
struct User {
    id: u32,
    name: String,
    email: String,
}

let user = User {
    id: 1,
    name: "Alice".to_string(),
    email: "alice@example.com".to_string(),
};

// Serialize to JSON string
let json_str = serde_json::to_string(&user).expect("serialization failed");
println!("Compact: {}", json_str);

// Pretty-print
let json_pretty = serde_json::to_string_pretty(&user).expect("serialization failed");
println!("Pretty:\n{}", json_pretty);
```

## Deserializing JSON to Structs

Use `serde_json::from_str()` to parse a JSON string into a Rust type.

```rust
use serde::{Deserialize, Serialize};

#[derive(Debug, Serialize, Deserialize)]
struct User {
    id: u32,
    name: String,
    email: String,
}

let json = r#"{"id": 2, "name": "Bob", "email": "bob@example.com"}"#;
let user: User = serde_json::from_str(json).expect("deserialization failed");
println!("Deserialized: {:?}", user);
```

## Working with serde_json::Value

When you don't know the structure ahead of time, use `serde_json::Value` for dynamic JSON.

```rust
let json_str = r#"{"name": "Rust", "versions": [1, 2, 3], "active": true}"#;
let value: serde_json::Value = serde_json::from_str(json_str).expect("parse failed");

// Access with indexing
println!("name: {:?}", value["name"]);
println!("versions: {:?}", value["versions"]);
println!("active: {:?}", value["active"]);

// Check types
if let Some(s) = value["name"].as_str() {
    println!("name as str: {}", s);
}
if let Some(arr) = value["versions"].as_array() {
    println!("versions count: {}", arr.len());
}
```

## Field Attributes

| Attribute | Purpose |
|-----------|----------|
| `#[serde(rename = "...")]` | Map Rust field to different JSON key |
| `#[serde(default)]` | Use Default if field is missing |
| `#[serde(skip)]` | Skip when serializing/deserializing |
| `#[serde(skip_serializing_if = "...")]` | Skip if condition is true |

```rust
use serde::{Deserialize, Serialize};

#[derive(Debug, Serialize, Deserialize)]
struct ApiResponse {
    #[serde(rename = "user_id")]
    id: u32,
    #[serde(default)]
    nickname: String,
    #[serde(skip)]
    internal_flag: bool,
}

// JSON uses "user_id", we use "id"
let json = r#"{"user_id": 42, "internal_flag": true}"#;
let resp: ApiResponse = serde_json::from_str(json).expect("parse failed");
println!("id: {}, nickname: '{}' (default)", resp.id, resp.nickname);
```

## Nested Structs and Enums

serde handles nested types and enums naturally.

```rust
use serde::{Deserialize, Serialize};

#[derive(Debug, Serialize, Deserialize)]
struct Address {
    street: String,
    city: String,
    zip: String,
}

#[derive(Debug, Serialize, Deserialize)]
struct Person {
    name: String,
    address: Address,
}

#[derive(Debug, Serialize, Deserialize)]
#[serde(tag = "type")]
enum Event {
    Click { x: i32, y: i32 },
    KeyPress { key: String },
}

let json = r#"{"name": "Carol", "address": {"street": "123 Main", "city": "Rustville", "zip": "12345"}}"#;
let person: Person = serde_json::from_str(json).expect("parse failed");
println!("{:?}", person);

let event_json = r#"{"type": "Click", "x": 10, "y": 20}"#;
let event: Event = serde_json::from_str(event_json).expect("parse failed");
println!("{:?}", event);
```

## Optional Fields with Option<T>

Use `Option<T>` for fields that may be missing or null.

```rust
use serde::{Deserialize, Serialize};

#[derive(Debug, Serialize, Deserialize)]
struct Product {
    name: String,
    price: f64,
    description: Option<String>,
    tags: Option<Vec<String>>,
}

let json1 = r#"{"name": "Widget", "price": 9.99}"#;
let p1: Product = serde_json::from_str(json1).expect("parse failed");
println!("description: {:?}, tags: {:?}", p1.description, p1.tags);

let json2 = r#"{"name": "Gadget", "price": 19.99, "description": "Cool stuff", "tags": ["new", "sale"]}"#;
let p2: Product = serde_json::from_str(json2).expect("parse failed");
println!("description: {:?}, tags: {:?}", p2.description, p2.tags);
```

## 🎯 Key Takeaways

1. **serde** is Rust's serialization framework; **serde_json** handles JSON
2. `#[derive(Serialize, Deserialize)]` on structs and enums
3. `serde_json::to_string()` / `to_string_pretty()` for serialization
4. `serde_json::from_str()` for deserialization
5. `serde_json::Value` for dynamic/unknown JSON structure
6. Attributes: `rename`, `default`, `skip` for flexible mapping
7. `Option<T>` for optional fields
