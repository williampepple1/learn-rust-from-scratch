# 🦀 Day 1: Threads

Today: `std::thread::spawn`, `join`, move closures for parallelism.

---

```rust
use std::thread;
use std::time::Duration;

// Spawn a thread
let handle = thread::spawn(|| {
    for i in 1..=5 {
        println!("  🧵 Spawned thread: count {}", i);
        thread::sleep(Duration::from_millis(50));
    }
    42 // Return value!
});

for i in 1..=3 {
    println!("  🏠 Main thread: count {}", i);
    thread::sleep(Duration::from_millis(50));
}

// Wait for thread and get result
let result = handle.join().unwrap();
println!("Thread returned: {}", result);
```

```rust
// move closures — transfer ownership to thread
let data = vec![1, 2, 3, 4, 5];

let handle = thread::spawn(move || {
    let sum: i32 = data.iter().sum();
    println!("Sum: {}", sum);
    sum
});
// println!("{:?}", data);  // ❌ data moved into thread
let result = handle.join().unwrap();
println!("Thread computed: {}", result);
```

```rust
// Parallel number crunching
let numbers: Vec<i64> = (1..=1_000_000).collect();
let chunks: Vec<Vec<i64>> = numbers.chunks(250_000).map(|c| c.to_vec()).collect();

let start = std::time::Instant::now();
let handles: Vec<_> = chunks.into_iter().map(|chunk| {
    thread::spawn(move || {
        chunk.iter().map(|&x| x * x).sum::<i64>()
    })
}).collect();

let total: i64 = handles.into_iter().map(|h| h.join().unwrap()).sum();
println!("Sum of squares (4 threads): {} in {:?}", total, start.elapsed());
```

## 🎯 Key Takeaways
1. `thread::spawn(|| {...})` creates a new OS thread
2. `.join()` waits for completion, returns `Result<T>`
3. `move` closures transfer ownership into threads
4. Split work into chunks for parallel computation
5. Rust prevents data races at **compile time**
