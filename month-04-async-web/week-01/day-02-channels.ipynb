# 🦀 Day 2: Message Passing with Channels
---

```rust
use std::sync::mpsc; // Multiple Producer, Single Consumer
use std::thread;
use std::time::Duration;

// Basic channel
let (tx, rx) = mpsc::channel();

thread::spawn(move || {
    let messages = vec!["hello", "from", "the", "thread"];
    for msg in messages {
        tx.send(msg.to_string()).unwrap();
        thread::sleep(Duration::from_millis(100));
    }
});

// Receive messages (blocks until available)
for received in rx {
    println!("Got: {}", received);
}
```

```rust
// Multiple producers
let (tx, rx) = mpsc::channel();

for id in 0..3 {
    let tx_clone = tx.clone();
    thread::spawn(move || {
        for i in 0..3 {
            tx_clone.send(format!("Worker {}: message {}", id, i)).unwrap();
            thread::sleep(Duration::from_millis(50));
        }
    });
}
drop(tx); // Drop original sender so rx knows when all senders are done

for msg in rx {
    println!("  {}", msg);
}
```

```rust
// Producer-consumer pattern
let (tx, rx) = mpsc::channel();

// Producer: generate work items
thread::spawn(move || {
    for i in 1..=10 {
        tx.send(i).unwrap();
    }
});

// Consumer: process work items
let mut results = Vec::new();
for item in rx {
    let processed = item * item;
    results.push(processed);
}
println!("Processed: {:?}", results);
```

## 🎯 Key Takeaways
1. `mpsc::channel()` creates a send/receive pair
2. `tx.send()` sends, `rx.recv()` or `for msg in rx` receives
3. Clone `tx` for multiple producers
4. Channel closes when all senders are dropped
5. **"Share memory by communicating"** — Go/Rust philosophy
