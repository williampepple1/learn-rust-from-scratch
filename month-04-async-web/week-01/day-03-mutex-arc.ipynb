# 🦀 Day 3: Shared State — `Mutex<T>` & `Arc<T>`
---

```rust
use std::sync::Mutex;

// Mutex — mutual exclusion lock
let counter = Mutex::new(0);

{
    let mut num = counter.lock().unwrap(); // Acquire lock
    *num += 1;
} // Lock released when MutexGuard is dropped

println!("Counter: {:?}", counter);
```

```rust
use std::sync::Arc; // Atomic Reference Counted
use std::thread;

// Arc<Mutex<T>> — thread-safe shared mutable state
let counter = Arc::new(Mutex::new(0));
let mut handles = vec![];

for _ in 0..10 {
    let counter = Arc::clone(&counter);
    let handle = thread::spawn(move || {
        let mut num = counter.lock().unwrap();
        *num += 1;
    });
    handles.push(handle);
}

for handle in handles { handle.join().unwrap(); }
println!("Final count: {}", *counter.lock().unwrap()); // 10
```

```rust
// Thread-safe shared data structure
let data = Arc::new(Mutex::new(Vec::<String>::new()));
let mut handles = vec![];

for id in 0..5 {
    let data = Arc::clone(&data);
    handles.push(thread::spawn(move || {
        let mut vec = data.lock().unwrap();
        vec.push(format!("Thread {} was here", id));
    }));
}

for h in handles { h.join().unwrap(); }
println!("Shared data: {:?}", data.lock().unwrap());
```

## 🎯 Key Takeaways
1. `Mutex<T>` provides mutual exclusion — one thread at a time
2. `.lock()` returns a `MutexGuard` that auto-unlocks on drop
3. `Arc<T>` = atomic `Rc<T>` — safe to share across threads
4. `Arc<Mutex<T>>` = the standard pattern for shared mutable state
5. Rust prevents data races at compile time — `Rc` won't compile in threads!
