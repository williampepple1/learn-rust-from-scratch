## DESIGN.md Template (RustKV Example)

```markdown
# RustKV — Key-Value Database Design

## Goals
- In-memory key-value store with optional persistence
- Simple text protocol (Redis-like)
- TTL support for key expiration

## Features
- GET key, SET key value [TTL], DELETE key
- KEYS pattern (glob-style)
- PING, QUIT

## Architecture
- storage/ — in-memory engine, persistence
- protocol/ — command parsing, response formatting
- server/ — TCP listener, connection handling
- client/ — CLI client (optional)

## Data Model
- Value: String, Integer, Float, List, Hash
- Entry: value + created_at + expires_at

## Non-Functional Requirements
- Performance: 10k ops/sec target
- Concurrency: multi-threaded, Arc<RwLock>
- Durability: WAL + snapshots
```

````rust
// Define the core data types for RustKV

use std::collections::HashMap;
use std::time::Duration;

/// Commands the client can send
#[derive(Debug, Clone, PartialEq)]
pub enum Command {
    Get(String),
    Set(String, Value, Option<Duration>),
    Delete(String),
    Keys(String),
    Ping,
    Quit,
}

/// Supported value types
#[derive(Debug, Clone, PartialEq)]
pub enum Value {
    String(String),
    Integer(i64),
    Float(f64),
    List(Vec<String>),
    Hash(HashMap<String, String>),
}

/// Stored entry with metadata
#[derive(Debug, Clone)]
pub struct Entry {
    pub value: Value,
    pub created_at: std::time::Instant,
    pub expires_at: Option<std::time::Instant>,
}

// Example usage
let cmd = Command::Set("name".to_string(), Value::String("RustKV".to_string()), Some(Duration::from_secs(60)));
println!("Command: {:?}", cmd);
````
