# 🦀 Day 3: Set Up CI/CD — GitHub Actions for Rust
---

Today we set up continuous integration and deployment for our Rust project.

- **What CI/CD is**: Automate build, test, and lint on every push/PR
- **GitHub Actions basics**: Workflows, jobs, steps
- **Rust workflow**: `cargo fmt`, `cargo clippy`, `cargo test`, `cargo build --release`
- **Matrix builds**: Test on multiple OS and Rust versions
- **Caching**: Speed up builds with `actions/cache`

## Complete `.github/workflows/rust.yml`

```yaml
name: CI

on:
  push:
    branches: [main, master]
  pull_request:
    branches: [main, master]

jobs:
  build:
    runs-on: ${{ matrix.os }}
    strategy:
      matrix:
        os: [ubuntu-latest, windows-latest, macos-latest]
        rust: [stable]
    steps:
      - uses: actions/checkout@v4
      - uses: dtolnay/rust-toolchain@stable
        with:
          components: clippy, rustfmt
      - uses: Swatinem/rust-cache@v2
      - run: cargo fmt -- --check
      - run: cargo clippy -- -D warnings
      - run: cargo test
      - run: cargo build --release
```

````rust
// Generate the YAML content programmatically (for learning)

let workflow = r#"
name: CI
on:
  push:
    branches: [main, master]
  pull_request:
    branches: [main, master]
jobs:
  build:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: dtolnay/rust-toolchain@stable
        with:
          components: clippy, rustfmt
      - uses: Swatinem/rust-cache@v2
      - run: cargo fmt -- --check
      - run: cargo clippy -- -D warnings
      - run: cargo test
      - run: cargo build --release
"#;

println!("Workflow steps:");
for (i, line) in workflow.trim().lines().enumerate() {
    if line.starts_with("-") && line.contains("run:") {
        println!("  {}. {}", (i+1)/2, line.trim());
    }
}
````

## Branch Protection & Required Checks

In GitHub: **Settings → Branches → Add rule** for `main`:
- Require status checks: `build` (or your job name)
- Require branches to be up to date before merging

This ensures no PR merges without passing CI.

## 🎯 Key Takeaways

1. CI/CD automates build, test, and lint on every push/PR
2. GitHub Actions: workflows → jobs → steps
3. Rust workflow: checkout → rust-toolchain → cache → fmt → clippy → test → build
4. Matrix builds test on multiple OS/Rust versions
5. `Swatinem/rust-cache` caches `~/.cargo` and `target/` for faster builds
6. Branch protection + required checks prevent merging broken code
