# 🦀 Day 4: Set Up clippy Lints and rustfmt Config
---

Today we configure Clippy (Rust's linter) and rustfmt for consistent, idiomatic code.

- **Clippy**: Catches common mistakes, enforces idioms, suggests improvements
- **Lint levels**: allow, warn, deny, forbid
- **clippy.toml**: cognitive-complexity-threshold, too-many-arguments-threshold
- **rustfmt.toml**: max_width, tab_spaces, imports_granularity
- **Useful lints**: unwrap_used, expect_used, dbg_macro, todo

## Clippy vs Pedantic

| Command | What it does |
|---------|--------------|
| `cargo clippy` | Default lints (sensible) |
| `cargo clippy -- -W clippy::pedantic` | Extra strict lints |
| `cargo clippy -- -D warnings` | Fail build on any lint (CI) |

**Lint levels**: `#[allow(clippy::xyz)]`, `#[warn(...)]`, `#[deny(...)]`, `#[forbid(...)]`

````rust
// Clippy catches issues — example of what it would flag

// BAD: unwrap can panic (clippy::unwrap_used)
// let x: Option<i32> = Some(42);
// let _ = x.unwrap();

// GOOD: use match or if let
let x: Option<i32> = Some(42);
let result = match x {
    Some(v) => v,
    None => 0,
};
println!("Result: {}", result);

// BAD: redundant clone (clippy::redundant_clone)
// let s = String::from("hello");
// let t = s.clone();

// GOOD: use reference if you don't need ownership
let s = String::from("hello");
let len = s.len();
println!("Length: {}", len);
````

````rust
// clippy.toml and rustfmt.toml configuration

let clippy_toml = r#"
# clippy.toml
cognitive-complexity-threshold = 15
too-many-arguments-threshold = 6
type-complexity-threshold = 250
"#;

let rustfmt_toml = r#"
# rustfmt.toml
max_width = 100
tab_spaces = 4
imports_granularity = "Crate"
group_imports = "StdExternalCrate"
"#;

println!("clippy.toml:");
println!("{}", clippy_toml);
println!("rustfmt.toml:");
println!("{}", rustfmt_toml);
````

````rust
// #[allow(clippy::...)] for justified exceptions

#[allow(clippy::too_many_arguments)]
fn complex_function(a: i32, b: i32, c: i32, d: i32, e: i32, f: i32, g: i32) -> i32 {
    a + b + c + d + e + f + g
}

// Sometimes unwrap is acceptable (e.g., in tests)
#[allow(clippy::unwrap_used)]
fn test_helper() -> String {
    Some("test".to_string()).unwrap()
}

println!("Sum: {}", complex_function(1, 2, 3, 4, 5, 6, 7));
println!("Helper: {}", test_helper());
````

## Useful Clippy Lints to Enable

Add to `lib.rs` or `main.rs` (or in `Cargo.toml` under `[lints.clippy]`):

```rust
#![warn(clippy::unwrap_used)]
#![warn(clippy::expect_used)]
#![warn(clippy::dbg_macro)]
#![warn(clippy::todo)]
```

`cargo fmt` formats code; `cargo fmt -- --check` fails if not formatted (for CI).
