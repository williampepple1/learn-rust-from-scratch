## Doc comments: `///` and `//!`

- **`///`** documents the *next* item (function, struct, enum, etc.)
- **`//!`** documents the *enclosing* item (module, crate) — usually at top of `lib.rs`
- Both support **Markdown** (headings, code blocks, lists)
- Code blocks in `# Examples` are run as tests with `cargo test`

````rust
/// A simple key-value store backed by an in-memory HashMap.
///
/// # Examples
/// ```
/// let mut store = KeyValueStore::new();
/// store.set("foo".to_string(), "bar".to_string());
/// assert_eq!(store.get("foo"), Some("bar".to_string()));
/// ```
pub struct KeyValueStore {
    data: std::collections::HashMap<String, String>,
}

impl KeyValueStore {
    /// Creates a new empty store.
    pub fn new() -> Self {
        Self { data: std::collections::HashMap::new() }
    }

    /// Sets a key to a value.
    pub fn set(&mut self, key: String, value: String) {
        self.data.insert(key, value);
    }

    /// Gets a value by key, or `None` if not found.
    pub fn get(&self, key: &str) -> Option<String> {
        self.data.get(key).cloned()
    }
}

mod kv_store {
    use super::*;
    pub struct KeyValueStore { pub(super) data: std::collections::HashMap<String, String> }
    impl KeyValueStore {
        pub fn new() -> Self { Self { data: std::collections::HashMap::new() } }
        pub fn set(&mut self, k: String, v: String) { self.data.insert(k, v); }
        pub fn get(&self, k: &str) -> Option<String> { self.data.get(k).cloned() }
    }
}

let mut store = kv_store::KeyValueStore::new();
store.set("foo".to_string(), "bar".to_string());
assert_eq!(store.get("foo"), Some("bar".to_string()));
println!("Documented API works! ✅");
````

````rust
/// Parses a simple GET/SET command string.
///
/// # Examples
/// ```
/// assert_eq!(parse_command("GET foo"), Some(("GET", "foo", None)));
/// assert_eq!(parse_command("SET foo bar"), Some(("SET", "foo", Some("bar"))));
/// assert_eq!(parse_command("PING"), Some(("PING", "", None)));
/// ```
fn parse_command(cmd: &str) -> Option<(&str, &str, Option<&str>)> {
    let parts: Vec<&str> = cmd.trim().split_whitespace().collect();
    match parts.as_slice() {
        ["GET", key] => Some(("GET", key, None)),
        ["SET", key, value] => Some(("SET", key, Some(value))),
        ["PING"] => Some(("PING", "", None)),
        _ => None,
    }
}

assert_eq!(parse_command("GET foo"), Some(("GET", "foo", None)));
assert_eq!(parse_command("SET foo bar"), Some(("SET", "foo", Some("bar"))));
println!("Doc tests would run with: cargo test --doc");
````

## 📝 Exercise: Write complete docs for your capstone's public API

1. Add `//!` crate-level docs at the top of `lib.rs`
2. Add `///` docs to every `pub` function, struct, and enum
3. Include `# Examples` with runnable code for main APIs
4. Run `cargo doc --open` to preview
5. Run `cargo test` to ensure doc examples pass

## 🎯 Key Takeaways

1. README.md: badges, description, features, installation, usage, contributing, license
2. `///` documents the next item; `//!` documents the enclosing module/crate
3. Doc examples in `# Examples` run as tests with `cargo test`
4. `cargo doc --open` generates and opens HTML API docs
5. Good docs = future-proof code and happy contributors
