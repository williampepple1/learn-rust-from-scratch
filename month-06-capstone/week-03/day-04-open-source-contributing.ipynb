# 🦀 Day 4: Open Source — Fork, Fix, Submit PR
---

Today we walk through the **contribution workflow**.

- **Workflow**: Fork → clone → branch → fix → test → PR
- **Git**: `git remote add upstream`, `git fetch upstream`
- **Commit messages**: Clear, imperative style
- **PR description**: What, why, how, testing
- **Responding to code review** and rebasing

## The contribution workflow

1. **Fork** the repo on GitHub
2. **Clone** your fork: `git clone https://github.com/YOU/repo.git`
3. **Add upstream**: `git remote add upstream https://github.com/ORIGINAL/repo.git`
4. **Branch**: `git checkout -b fix/issue-123-description`
5. **Fix** the code
6. **Test**: `cargo test`, `cargo clippy`, `cargo fmt`
7. **Commit**: `git add .` then `git commit -m "fix: resolve issue #123"`
8. **Push**: `git push origin fix/issue-123-description`
9. **Open PR** on GitHub against `ORIGINAL/main`

````rust
// Git commands for the workflow (run in terminal, not EvCxR)
let git_workflow = r#"
git remote add upstream https://github.com/ORIGINAL/repo.git
git fetch upstream
git checkout -b fix/issue-123
# ... make changes ...
cargo test && cargo clippy && cargo fmt
git add .
git commit -m "fix: resolve issue #123 - add missing error handling"
git push origin fix/issue-123
"#;

println!("Git workflow:");
println!("{}", git_workflow);
````

## Writing good commit messages

- **Format**: `type: short description` (e.g., `fix:`, `feat:`, `docs:`)
- **Imperative mood**: "add" not "added"
- **Reference issue**: "fix: resolve #123 - add validation"

## PR description template

```markdown
## What
Brief description of changes.

## Why
Fixes #123

## How
- Added X
- Updated Y

## Testing
- Ran `cargo test`
- Manual verification: ...
```

## Common PR mistakes to avoid

- Pushing to main instead of a branch
- PRs that change unrelated files (run `cargo fmt` only on changed code)
- Ignoring CI failures
- Not responding to review feedback
- Rebasing after review (use `git rebase upstream/main` before pushing new commits)

## 📝 Exercise: Practice the fork-branch-PR workflow

1. Fork your own capstone repo (or use a throwaway repo)
2. Clone, add upstream, create a branch
3. Make a small change (e.g., fix a typo in README)
4. Run `cargo test`, `cargo clippy`, `cargo fmt`
5. Commit, push, open a PR against your main branch

## 🎯 Key Takeaways

1. Fork → clone → upstream → branch → fix → test → commit → push → PR
2. Use `git fetch upstream` and `git rebase upstream/main` to stay in sync
3. Good commit messages: type + imperative + issue reference
4. PR: what, why, how, testing
5. Respond to review; fix and push; avoid force-pushing after review
