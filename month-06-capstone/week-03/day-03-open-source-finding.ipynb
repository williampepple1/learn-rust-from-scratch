## Popular Rust projects welcoming contributions

- **tokio** — async runtime
- **serde** — serialization
- **clap** — CLI parsing
- **thiserror / anyhow** — error handling
- **ripgrep, fd, bat** — CLI tools
- **rust-lang/rust** — the compiler itself (advanced)

Check each repo's CONTRIBUTING.md and issue labels before diving in.

````rust
// Checklist for evaluating a project
let checklist = [
    "Active: recent commits and issue responses",
    "CONTRIBUTING.md exists and is clear",
    "CODE_OF_CONDUCT.md present",
    "good-first-issue or help-wanted labels used",
    "Maintainers respond to PRs within a few weeks",
];

for (i, item) in checklist.iter().enumerate() {
    println!("{}. {}", i + 1, item);
}
println!("\nHow to claim: Comment on the issue: 'I'd like to work on this.'");
````

## Setting up the development environment

1. Fork the repo on GitHub
2. Clone your fork: `git clone https://github.com/YOUR_USERNAME/repo.git`
3. Add upstream: `git remote add upstream https://github.com/ORIGINAL/repo.git`
4. Run `cargo test`, `cargo clippy`, `cargo fmt` to ensure you can build
5. Create a branch: `git checkout -b fix/issue-123`

## 🎯 Key Takeaways

1. Use `good-first-issue` and `help-wanted` labels to find entry points
2. Evaluate: activity, CONTRIBUTING.md, maintainer responsiveness
3. Read CONTRIBUTING.md and CODE_OF_CONDUCT.md before contributing
4. Claim issues by commenting; set up fork + upstream remote
