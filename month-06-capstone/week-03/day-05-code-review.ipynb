## Common Month 1 patterns to improve

| Pattern | Problem | Better approach |
|---------|---------|-----------------|
| `.unwrap()` everywhere | Panics on error | `?` operator, `match`, proper error types |
| `String` where `&str` works | Unnecessary allocations | Use `&str` for input/output |
| Manual `for` loops | Verbose | Iterator chains (`.map()`, `.filter()`, `.collect()`) |
| No tests | Hard to refactor | Add `#[test]` and doc tests |
| Missing `impl Display` | Poor output | Implement `Display` for user-facing types |

```rust
// Month 1 style: unwrap, manual loop, String allocations
fn sum_squares_old(nums: Vec<i32>) -> i32 {
    let mut sum = 0;
    for n in nums {
        sum += n * n;
    }
    sum
}

// Month 6 idiomatic: iterator chain
fn sum_squares_new(nums: &[i32]) -> i32 {
    nums.iter().map(|n| n * n).sum()
}

let nums = vec![1, 2, 3, 4, 5];
assert_eq!(sum_squares_old(nums.clone()), 55);
assert_eq!(sum_squares_new(&nums), 55);
println!("Both work; new version: no clone, no mut, iterator style");
```

```rust
// Before: unwrap everywhere
// let n: i32 = s.parse().unwrap();

// After: proper error handling
fn parse_positive(s: &str) -> Result<i32, String> {
    let n: i32 = s.parse().map_err(|_| format!("invalid number: {}", s))?;
    if n > 0 { Ok(n) } else { Err("must be positive".to_string()) }
}

assert!(parse_positive("42").is_ok());
assert!(parse_positive("abc").is_err());
assert!(parse_positive("-1").is_err());
println!("Error handling: use Result and ? instead of unwrap");
```

```rust
// Before: fn greet(name: String) -> String
// After: fn greet(name: &str) -> String  — avoids allocation when caller has &str

fn greet(name: &str) -> String {
    format!("Hello, {}!", name)
}

let s = "Rust";
println!("{}", greet(s));  // No .to_string() needed
println!("{}", greet("World"));  // Works with string literal
```
