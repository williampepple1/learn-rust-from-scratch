# 🦀 Day 7: Capstone v1.0 Release
---

Today we **tag a release** on GitHub — your capstone's v1.0 milestone!

- **What a release means**: Stable, documented, tested milestone
- **Pre-release checklist**: Tests pass, docs complete, README, CHANGELOG
- **Git tagging**: `git tag -a v1.0.0 -m "Initial release"`
- **GitHub Releases**: Create release with notes
- **CHANGELOG.md**: Keep a Changelog convention

## Pre-release checklist

- [ ] `cargo test` — all tests pass
- [ ] `cargo clippy` — no warnings
- [ ] `cargo fmt` — code formatted
- [ ] README.md up to date (installation, usage, examples)
- [ ] API docs: `cargo doc` builds
- [ ] CHANGELOG.md written for this release

````rust
// CHANGELOG.md format (Keep a Changelog)
let changelog = r#"
# Changelog

## [1.0.0] - 2025-03-08

### Added
- GET, SET, DEL commands
- TCP server with async support
- Basic persistence (optional)
- README and API documentation

### Changed
- Initial stable release

[1.0.0]: https://github.com/you/rustkv/releases/tag/v1.0.0
"#;

println!("CHANGELOG.md template:");
println!("{}", changelog);
````

````rust
// Git commands for release (run in terminal)
let release_commands = r##"
git add .
git commit -m "chore: prepare v1.0.0 release"
git tag -a v1.0.0 -m "Initial release"
git push origin main
git push origin v1.0.0

Then: GitHub -> Releases -> Create new release -> choose tag v1.0.0
Add release notes (copy from CHANGELOG.md)

Optional: cargo build --release, attach target/release/rustkv to release
"##;

println!("Release workflow:");
println!("{}", release_commands);
````

## Release notes template

```markdown
## RustKV v1.0.0

First stable release! 🎉

### Features
- GET, SET, DEL key-value operations
- TCP server (Redis-like protocol)
- Async with Tokio

### Installation
```bash
cargo install rustkv
```

### Acknowledgments
Built as part of the Learn Rust From Scratch course.
```

## GitHub Actions for automated releases (conceptual)

Many projects use a workflow that:
1. Triggers on tag push (`v*`)
2. Runs `cargo build --release`
3. Cross-compiles for Linux, macOS, Windows
4. Attaches binaries to the GitHub Release

See [release-plz](https://github.com/MarcoIeni/release-plz) or [cargo-release](https://github.com/crate-ci/cargo-release) for automation.
