## Repository Audit Checklist

| Check | Tool/Command |
|-------|--------------|
| Dead code | `#[warn(dead_code)]` in lib.rs |
| Unused deps | `cargo udeps` (install: `cargo install cargo-udeps`) |
| License audit | `cargo deny check` |
| Outdated deps | `cargo outdated` |
| Security | `cargo audit` |
| Format | `cargo fmt -- --check` |
| Lint | `cargo clippy` |

````rust
// Audit dependencies and check for issues
// Run these in your project directory (not in EvCxR):
//   cargo build --release
//   cargo clippy
//   cargo fmt -- --check
//   cargo audit (if installed)
//   cargo udeps (if installed)

// In EvCxR: simulate checking for common issues
let audit_commands = [
    "cargo build --release",
    "cargo clippy -- -D warnings",
    "cargo fmt -- --check",
    "cargo audit",
    "cargo udeps",
];

println!("Repository audit checklist:");
for (i, cmd) in audit_commands.iter().enumerate() {
    println!("  {}. {}", i + 1, cmd);
}
println!("\nRun these in your capstone/project root.");
````

## 🎯 Key Takeaways

1. A clean repo improves maintainability and collaboration
2. Use `cargo udeps` for unused dependencies, `cargo deny` for licenses
3. `cargo audit` checks for known vulnerabilities
4. Consistent READMEs and folder structure across months
5. Git tags mark milestones: `month-01-complete`, etc.
6. Be cautious with history cleanup — prefer incremental fixes
