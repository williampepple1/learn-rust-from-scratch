````rust
// Rust patterns cheat sheet — common idioms

// 1. Option/Result chaining
let maybe: Option<i32> = Some(42);
let doubled = maybe.map(|x| x * 2);

// 2. Iterator chain
let sum: i32 = [1, 2, 3, 4, 5].iter().filter(|&x| x > 2).sum();

// 3. Builder pattern
struct Config { timeout: u64 }
impl Config {
    fn new() -> Self { Config { timeout: 0 } }
    fn timeout(mut self, t: u64) -> Self { self.timeout = t; self }
}

// 4. Default + override
let config = Config::new().timeout(30);

// 5. Clone-on-write (Cow)
use std::borrow::Cow;
fn process(s: &str) -> Cow<str> {
    if s.contains(' ') { Cow::Owned(s.replace(' ', "_")) } else { Cow::Borrowed(s) }
}

println!("Patterns: Option map, Iterator, Builder, Cow");
println!("sum of [1..5] where x>2: {}", sum);
````
