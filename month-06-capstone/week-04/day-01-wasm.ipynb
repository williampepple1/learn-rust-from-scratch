# 🦀 Day 1: WebAssembly (WASM) with Rust
---

Welcome to **Week 4: Advanced Topics & Beyond**! Today we explore Rust's powerful WebAssembly story.

- **What WASM is**: Portable binary format that runs in browsers and beyond
- **Why Rust + WASM**: No garbage collector, small binaries, memory safety
- **wasm-pack & wasm-bindgen**: The toolchain for compiling Rust to WASM
- **#[wasm_bindgen]**: Exposing Rust functions to JavaScript
- **WASI**: WebAssembly System Interface for server-side WASM
- **Use cases**: Games, image processing, crypto, data processing

## WASM Module Setup

A typical `Cargo.toml` for a WASM crate:

```toml
[package]
name = "my-wasm-crate"
version = "0.1.0"

[lib]
crate-type = ["cdylib"]

[dependencies]
wasm-bindgen = "0.2"
```

Build with: `wasm-pack build --target web`

````rust
// Pure Rust algorithms that would compile to WASM
// These run in EvCxR; the same logic works in a WASM module

/// Fast Fibonacci (iterative) — great for WASM: no heap, pure computation
fn fib(n: u32) -> u64 {
    match n {
        0 => 0,
        1 => 1,
        _ => {
            let (mut a, mut b) = (0u64, 1u64);
            for _ in 2..=n {
                let next = a + b;
                a = b;
                b = next;
            }
            b
        }
    }
}

/// String processing: count words — typical WASM use case
fn count_words(s: &str) -> usize {
    s.split_whitespace().count()
}

// In a real WASM crate, you'd add:
// #[wasm_bindgen]
// pub fn fib_wasm(n: u32) -> u64 { fib(n) }

println!("fib(20) = {}", fib(20));
println!("count_words('hello world from rust') = {}", count_words("hello world from rust"));
````

## JavaScript Integration

After building with `wasm-pack build --target web`, you get:

```javascript
import init, { fib_wasm } from './pkg/my_wasm_crate.js';
await init();
console.log(fib_wasm(20)); // 6765
```

## 🎯 Key Takeaways

1. WebAssembly is a portable binary format for browsers and beyond
2. Rust + WASM: no GC, small binaries, memory safety — ideal for performance-critical web code
3. `wasm-pack` and `wasm-bindgen` are the standard toolchain
4. Use `#[wasm_bindgen]` to expose Rust functions to JavaScript
5. WASI enables server-side WASM with file I/O and networking
6. Use cases: games, image processing, crypto, data processing
