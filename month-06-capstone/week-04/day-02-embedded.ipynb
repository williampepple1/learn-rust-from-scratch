# 🦀 Day 2: Embedded Rust — no_std, embedded-hal Concepts
---

Today we explore Rust for microcontrollers and bare-metal systems.

- **What embedded Rust is**: Rust for microcontrollers without an OS
- **#![no_std]**: Rust without the standard library
- **What you lose**: Heap allocation, I/O, threads, networking
- **What you keep**: Core language, iterators, Option, Result, traits
- **embedded-hal**: Trait abstractions for GPIO, SPI, I2C, UART
- **PAC → HAL → Board Support**: The embedded crate hierarchy

## core vs alloc vs std

| Crate | Heap | I/O | Threads | Use Case |
|-------|------|-----|---------|----------|
| `core` | ❌ | ❌ | ❌ | Bare metal, no allocator |
| `alloc` | ✅ | ❌ | ❌ | With allocator, no I/O |
| `std` | ✅ | ✅ | ✅ | Full hosted environment |

Bare-metal programs use `#![no_main]` and `#[entry]` instead of `main`.

```rust
// no_std-style code: pure computation, no heap, no I/O
// This compiles and runs in EvCxR (which has std), but the logic is no_std compatible

/// Fixed-point multiply (no floats on many embedded targets)
fn fixed_mul(a: i32, b: i32, shift: u32) -> i32 {
    ((a as i64 * b as i64) >> shift) as i32
}

/// Bit manipulation — common in embedded (registers, flags)
fn set_bit(value: u32, bit: u8) -> u32 {
    value | (1 << bit)
}

fn clear_bit(value: u32, bit: u8) -> u32 {
    value & !(1 << bit)
}

// embedded-hal concepts: traits like OutputPin, InputPin, SpiBus
// You implement these for your specific hardware

println!("fixed_mul(100, 200, 8) = {}", fixed_mul(100, 200, 8));
println!("set_bit(0, 3) = {}", set_bit(0, 3));
println!("clear_bit(0b1111, 2) = {}", clear_bit(0b1111, 2));
```

## Popular Embedded Targets

- **ARM Cortex-M**: STM32, nRF52, RP2040 (Raspberry Pi Pico)
- **RISC-V**: ESP32-C3, SiFive
- **ESP32**: Xtensa and RISC-V variants

Use **QEMU** or **probe-rs** for development without physical hardware.

## 🎯 Key Takeaways

1. Embedded Rust uses `#![no_std]` to run without the standard library
2. `core` = no heap; `alloc` = heap; `std` = full environment
3. `#![no_main]` and `#[entry]` for bare-metal entry points
4. `embedded-hal` provides trait abstractions (GPIO, SPI, I2C, UART)
5. PAC (Peripheral Access Crate) → HAL → Board Support is the crate hierarchy
6. Use QEMU or simulators for development without hardware
