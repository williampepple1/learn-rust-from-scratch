```rust
// Final exam: a concurrent, generic, well-tested mini data structure
// Use everything you've learned: generics, Arc, Mutex, traits, tests

use std::sync::{Arc, Mutex};

/// Thread-safe, generic counter
struct Counter<T: Copy + Default> {
    value: Arc<Mutex<T>>,
}

impl<T: Copy + Default> Counter<T> {
    fn new() -> Self {
        Counter { value: Arc::new(Mutex::new(T::default())) }
    }
    fn get(&self) -> T {
        *self.value.lock().unwrap()
    }
}

impl Counter<u64> {
    fn inc(&self) {
        let mut v = self.value.lock().unwrap();
        *v += 1;
    }
}

let counter = Arc::new(Counter::<u64>::new());
counter.inc();
counter.inc();
println!("Counter value: {}", counter.get());
println!("You built a concurrent, generic data structure!");
```
