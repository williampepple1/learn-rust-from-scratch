Today we dive into Rust's advanced type system features.

## Generic Associated Types (GATs)

GATs allow associated types to have generic parameters (including lifetimes):

```rust
trait LendingIterator {
    type Item<'a> where Self: 'a;
    fn next(&mut self) -> Option<Self::Item<'_>>;
}
```

This enables **lending iterators** — iterators that borrow from `self`.

````rust
// Type-state pattern with PhantomData
// Compile-time state machine: invalid transitions don't compile

use std::marker::PhantomData;

struct Locked;
struct Unlocked;

struct Door<State> {
    _state: PhantomData<State>,
}

impl Door<Locked> {
    fn new() -> Self {
        Door { _state: PhantomData }
    }
    fn unlock(self) -> Door<Unlocked> {
        Door { _state: PhantomData }
    }
}

impl Door<Unlocked> {
    fn lock(self) -> Door<Locked> {
        Door { _state: PhantomData }
    }
    fn open(&self) {
        println!("Door opened!");
    }
}

let door = Door::<Locked>::new();
let door = door.unlock();
door.open();
let _door = door.lock();
// door.open(); // ERROR: door is locked!
println!("Type-state pattern: invalid transitions don't compile");
````

## Sealed Traits & impl Trait

- **Sealed traits**: Prevent external crates from implementing your trait
- **impl Trait**: Existential types — "some type that implements Trait"
- **Type aliases**: `type Result<T> = std::result::Result<T, MyError>;`

## 🎯 Key Takeaways

1. GATs (`type Item<'a>`) enable lending iterators and async traits
2. PhantomData lets you carry type-level state without runtime cost
3. Type-state pattern: invalid state transitions don't compile
4. Sealed traits prevent external implementations
5. `impl Trait` provides existential types in return position
6. Type aliases simplify complex type signatures
