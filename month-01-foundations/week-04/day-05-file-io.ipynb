## 📖 Reading Files

```rust
use std::fs;

// First, let's create a file to work with
fs::write("example.txt", "Hello, Rust!\nThis is line 2.\nAnd line 3.")
    .expect("Failed to write file");
println!("File created! ✅");

// Method 1: Read entire file to String (simplest)
let contents = fs::read_to_string("example.txt")
    .expect("Failed to read file");
println!("--- File contents ---");
println!("{}", contents);
println!("--- End ---");
```

```rust
// Method 2: Read as bytes
let bytes = fs::read("example.txt")
    .expect("Failed to read");
println!("File size: {} bytes", bytes.len());
println!("First 10 bytes: {:?}", &bytes[..10]);
```

```rust
use std::io::{self, BufRead};
use std::fs::File;

// Method 3: Read line by line (memory efficient for large files)
let file = File::open("example.txt").expect("Failed to open");
let reader = io::BufReader::new(file);

println!("Reading line by line:");
for (i, line) in reader.lines().enumerate() {
    let line = line.expect("Failed to read line");
    println!("  Line {}: {}", i + 1, line);
}
```

## ✏️ Writing Files

```rust
use std::fs;
use std::io::Write;

// Method 1: Write entire content at once (overwrites!)
fs::write("output.txt", "This is the entire file content.")
    .expect("Failed to write");
println!("Written with fs::write ✅");

// Method 2: Using File and write!/writeln!
let mut file = File::create("report.txt").expect("Failed to create");
writeln!(file, "=== Report ===").unwrap();
writeln!(file, "Date: 2026-03-08").unwrap();
writeln!(file, "").unwrap();
for i in 1..=5 {
    writeln!(file, "Item {}: value = {}", i, i * 10).unwrap();
}
writeln!(file, "\n=== End ===").unwrap();
println!("Report written ✅");

// Verify
println!("{}", fs::read_to_string("report.txt").unwrap());
```

```rust
use std::fs::OpenOptions;

// Method 3: Append to existing file
let mut file = OpenOptions::new()
    .append(true)
    .open("report.txt")
    .expect("Failed to open for append");

writeln!(file, "\n--- Appended Section ---").unwrap();
writeln!(file, "This was added later!").unwrap();

println!("Appended ✅");
println!("{}", fs::read_to_string("report.txt").unwrap());
```

## ⚠️ Proper Error Handling

```rust
use std::fs;
use std::io;

fn read_file_safe(path: &str) -> Result<String, io::Error> {
    fs::read_to_string(path)
}

// Handle gracefully
match read_file_safe("example.txt") {
    Ok(contents) => println!("Read {} bytes", contents.len()),
    Err(e) => println!("Error reading: {}", e),
}

match read_file_safe("nonexistent.txt") {
    Ok(contents) => println!("Contents: {}", contents),
    Err(e) => println!("Expected error: {} (kind: {:?})", e, e.kind()),
}
```

## 📂 Filesystem Operations

```rust
use std::fs;
use std::path::Path;

// Check if file/dir exists
println!("example.txt exists: {}", Path::new("example.txt").exists());
println!("fake.txt exists: {}", Path::new("fake.txt").exists());

// File metadata
let metadata = fs::metadata("example.txt").unwrap();
println!("Size: {} bytes", metadata.len());
println!("Is file: {}", metadata.is_file());
println!("Is dir: {}", metadata.is_dir());

// Create directory
fs::create_dir_all("test_dir/sub_dir").unwrap();
println!("Directory created ✅");

// Copy file
fs::copy("example.txt", "test_dir/copy.txt").unwrap();
println!("File copied ✅");

// Rename/move
fs::rename("test_dir/copy.txt", "test_dir/renamed.txt").unwrap();
println!("File renamed ✅");

// Clean up
fs::remove_file("test_dir/renamed.txt").unwrap();
fs::remove_dir_all("test_dir").unwrap();
println!("Cleaned up ✅");
```

## 📝 Practical: Note-Taking App

```rust
use std::fs::{self, OpenOptions};
use std::io::Write;

const NOTES_FILE: &str = "notes.txt";

fn add_note(note: &str) {
    let mut file = OpenOptions::new()
        .create(true)
        .append(true)
        .open(NOTES_FILE)
        .expect("Cannot open notes");
    writeln!(file, "- {}", note).expect("Cannot write");
    println!("  📝 Added: {}", note);
}

fn list_notes() {
    match fs::read_to_string(NOTES_FILE) {
        Ok(contents) => {
            println!("\n📋 Your Notes:");
            println!("{}", contents);
        },
        Err(_) => println!("No notes yet!"),
    }
}

fn note_count() -> usize {
    fs::read_to_string(NOTES_FILE)
        .map(|c| c.lines().count())
        .unwrap_or(0)
}

// Use the app
add_note("Learn Rust file I/O");
add_note("Practice with std::fs");
add_note("Build a note-taking CLI");
add_note("Push to GitHub");

list_notes();
println!("Total notes: {}", note_count());

// Clean up
fs::remove_file(NOTES_FILE).ok();
fs::remove_file("example.txt").ok();
fs::remove_file("output.txt").ok();
fs::remove_file("report.txt").ok();
```

## 🎯 Key Takeaways

1. `fs::read_to_string()` / `fs::write()` for simple operations
2. `BufReader` for line-by-line reading (memory efficient)
3. `OpenOptions` for fine-grained control (append, create, etc.)
4. File operations return `Result` — always handle errors!
5. Use `Path` and `PathBuf` for cross-platform paths
