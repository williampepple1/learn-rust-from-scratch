## 📦 What is crates.io?

[crates.io](https://crates.io) is Rust's package registry — like npm for Node.js or PyPI for Python.

As of 2026, it hosts **150,000+** crates!

### Adding a Dependency

**In your `Cargo.toml`:**
```toml
[package]
name = "my_project"
version = "0.1.0"
edition = "2021"

[dependencies]
rand = "0.8"          # Random number generation
colored = "2"          # Colored terminal output
serde = { version = "1", features = ["derive"] }  # With features
```

Or use the CLI:
```bash
cargo add rand
cargo add colored
cargo add serde --features derive
```

## 🔢 Semantic Versioning

Rust uses **SemVer**: `MAJOR.MINOR.PATCH`

| Version string | Meaning |
|---|---|
| `"0.8.5"` | Exactly 0.8.5 |
| `"0.8"` | Any 0.8.x (compatible) |
| `"^0.8"` | Same as `"0.8"` |
| `"~0.8"` | 0.8.x only (more restrictive) |
| `">=1.0, <2.0"` | Range |
| `"*"` | Any version (avoid!) |

## 🎲 Using the `rand` Crate

In EvCxR Jupyter, we can load crates with the `:dep` command:

````rust
:dep rand = "0.8"
````

````rust
use rand::Rng;

let mut rng = rand::thread_rng();

// Random integer in range
let n: i32 = rng.gen_range(1..=100);
println!("Random number (1-100): {}", n);

// Random float
let f: f64 = rng.gen_range(0.0..1.0);
println!("Random float (0-1): {:.4}", f);

// Random bool
let coin: bool = rng.gen();
println!("Coin flip: {}", if coin { "Heads" } else { "Tails" });
````

````rust
use rand::seq::SliceRandom;

let mut rng = rand::thread_rng();

// Shuffle a vector
let mut cards = vec!["Ace", "King", "Queen", "Jack", "10"];
cards.shuffle(&mut rng);
println!("Shuffled: {:?}", cards);

// Pick a random element
let chosen = cards.choose(&mut rng).unwrap();
println!("Random card: {}", chosen);

// Generate a random password
let chars: Vec<char> = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%"
    .chars().collect();
let password: String = (0..16)
    .map(|_| *chars.choose(&mut rng).unwrap())
    .collect();
println!("Random password: {}", password);
````

## 🎯 Mini Dice Game

````rust
use rand::Rng;

fn roll_dice() -> u8 {
    rand::thread_rng().gen_range(1..=6)
}

fn play_round(player: &str) -> u8 {
    let d1 = roll_dice();
    let d2 = roll_dice();
    let total = d1 + d2;
    println!("  {} rolled [{} + {}] = {}", player, d1, d2, total);
    total
}

println!("🎲 Dice Game!\n");
let mut p1_total = 0u32;
let mut p2_total = 0u32;

for round in 1..=3 {
    println!("Round {}:", round);
    p1_total += play_round("Player 1") as u32;
    p2_total += play_round("Player 2") as u32;
    println!();
}

println!("Final: P1={}, P2={}", p1_total, p2_total);
match p1_total.cmp(&p2_total) {
    std::cmp::Ordering::Greater => println!("🏆 Player 1 wins!"),
    std::cmp::Ordering::Less => println!("🏆 Player 2 wins!"),
    std::cmp::Ordering::Equal => println!("🤝 It's a tie!"),
}
````
