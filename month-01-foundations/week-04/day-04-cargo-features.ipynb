## 📝 Documentation Comments

Rust has first-class documentation support. Write docs with `///` and generate HTML with `cargo doc`.

````rust
/// A 2D point in Cartesian coordinates.
///
/// # Examples
///
/// ```
/// let p = Point::new(3.0, 4.0);
/// assert_eq!(p.distance_to_origin(), 5.0);
/// ```
#[derive(Debug, Clone, Copy)]
pub struct Point {
    /// The x-coordinate
    pub x: f64,
    /// The y-coordinate
    pub y: f64,
}

impl Point {
    /// Creates a new `Point` at the given coordinates.
    ///
    /// # Arguments
    ///
    /// * `x` - The x-coordinate
    /// * `y` - The y-coordinate
    ///
    /// # Examples
    ///
    /// ```
    /// let origin = Point::new(0.0, 0.0);
    /// ```
    pub fn new(x: f64, y: f64) -> Self {
        Point { x, y }
    }

    /// Calculates the distance from this point to the origin `(0, 0)`.
    ///
    /// Uses the Pythagorean theorem: `√(x² + y²)`
    pub fn distance_to_origin(&self) -> f64 {
        (self.x.powi(2) + self.y.powi(2)).sqrt()
    }

    /// Calculates the distance between this point and another.
    ///
    /// # Panics
    ///
    /// This function will not panic.
    pub fn distance_to(&self, other: &Point) -> f64 {
        ((self.x - other.x).powi(2) + (self.y - other.y).powi(2)).sqrt()
    }
}

let p = Point::new(3.0, 4.0);
println!("Distance to origin: {}", p.distance_to_origin());
println!("Docs sections: Examples, Arguments, Panics, Errors, Safety");
````

### Documentation Sections

Standard doc comment sections:

| Section | When to use |
|---------|-------------|
| `# Examples` | Show usage (these are also **tested** by `cargo test`!) |
| `# Arguments` | Describe parameters |
| `# Returns` | Describe return value |
| `# Errors` | When can it return `Err`? |
| `# Panics` | When can it panic? |
| `# Safety` | For unsafe functions |

## 🧪 Testing Preview

Rust has built-in testing. Here's a quick preview (we'll dive deep in Month 3).

````rust
// In a real project, tests go in the same file or in tests/ directory
// Here we'll just show the syntax

fn add(a: i32, b: i32) -> i32 {
    a + b
}

fn is_even(n: i32) -> bool {
    n % 2 == 0
}

// Simulate what tests look like:
// In a real file you'd write:
//
// #[cfg(test)]
// mod tests {
//     use super::*;
//
//     #[test]
//     fn test_add() {
//         assert_eq!(add(2, 3), 5);
//         assert_eq!(add(-1, 1), 0);
//         assert_eq!(add(0, 0), 0);
//     }
//
//     #[test]
//     fn test_is_even() {
//         assert!(is_even(4));
//         assert!(!is_even(7));
//     }
// }

// Let's manually verify:
assert_eq!(add(2, 3), 5);
assert_eq!(add(-1, 1), 0);
assert!(is_even(4));
assert!(!is_even(7));
println!("All assertions passed! ✅");
````

### Test Macros

| Macro | Purpose |
|-------|--------|
| `assert!(expr)` | Assert expression is true |
| `assert_eq!(a, b)` | Assert a equals b |
| `assert_ne!(a, b)` | Assert a does NOT equal b |
| `#[should_panic]` | Test that should panic |

Run tests: `cargo test`

## 📦 Cargo Workspaces

Workspaces let you manage multiple related crates together.

```
my_workspace/
├── Cargo.toml            # Workspace root
├── core/
│   ├── Cargo.toml
│   └── src/lib.rs
├── cli/
│   ├── Cargo.toml
│   └── src/main.rs
└── web/
    ├── Cargo.toml
    └── src/main.rs
```

**Root `Cargo.toml`:**
```toml
[workspace]
members = [
    "core",
    "cli",
    "web",
]
```

**`cli/Cargo.toml`:**
```toml
[dependencies]
core = { path = "../core" }
```

Benefits:
- Shared `Cargo.lock` and `target/` directory
- Build all crates with one `cargo build`
- Test everything with `cargo test --workspace`

## 🧹 Cargo Clippy & Fmt

### `cargo fmt` — Auto-format code
```bash
cargo fmt           # Format all code
cargo fmt -- --check  # Check without modifying
```

Configure with `rustfmt.toml`:
```toml
max_width = 100
tab_spaces = 4
use_small_heuristics = "Max"
```

### `cargo clippy` — Lint your code
```bash
cargo clippy          # Show warnings
cargo clippy --fix    # Auto-fix what it can
```

Clippy catches hundreds of common mistakes and suggests idiomatic code!

## 🎯 Key Takeaways

1. `cargo check` is faster than `cargo build` — use it often
2. `///` doc comments support Markdown and generate HTML docs
3. Code examples in `///` are **tested** by `cargo test`!
4. **Always** run `cargo clippy` and `cargo fmt` before committing
5. Workspaces manage related crates together
