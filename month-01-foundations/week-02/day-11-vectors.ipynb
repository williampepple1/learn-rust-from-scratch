## 📋 Creating Vectors

```rust
// Method 1: Vec::new()
let mut v1: Vec<i32> = Vec::new();
v1.push(1);
v1.push(2);
v1.push(3);

// Method 2: vec! macro (most common)
let v2 = vec![1, 2, 3];

// Method 3: Repeated value
let v3 = vec![0; 5]; // [0, 0, 0, 0, 0]

// Method 4: With capacity (optimization)
let v4: Vec<i32> = Vec::with_capacity(100);

println!("v1: {:?}", v1);
println!("v2: {:?}", v2);
println!("v3: {:?}", v3);
println!("v4: len={}, capacity={}", v4.len(), v4.capacity());
```

## 📖 Reading Elements

```rust
let v = vec![10, 20, 30, 40, 50];

// Method 1: Indexing (panics if out of bounds!)
println!("v[0] = {}", v[0]);
println!("v[2] = {}", v[2]);

// Method 2: .get() (returns Option — safe!)
println!("v.get(2) = {:?}", v.get(2));
println!("v.get(99) = {:?}", v.get(99)); // None, no panic!

// Useful methods
println!("len: {}", v.len());
println!("is_empty: {}", v.is_empty());
println!("first: {:?}", v.first());
println!("last: {:?}", v.last());
println!("contains 30: {}", v.contains(&30));
```

## ✏️ Modifying Vectors

```rust
let mut v = vec![1, 2, 3];
println!("Start: {:?}", v);

v.push(4);                    // Add to end
println!("push(4): {:?}", v);

v.insert(1, 10);              // Insert at index 1
println!("insert(1,10): {:?}", v);

let popped = v.pop();         // Remove last
println!("pop: {:?}, vec: {:?}", popped, v);

let removed = v.remove(1);    // Remove at index
println!("remove(1): {}, vec: {:?}", removed, v);

v.extend([7, 8, 9]);         // Append multiple
println!("extend: {:?}", v);

v.retain(|&x| x > 2);       // Keep only elements > 2
println!("retain(>2): {:?}", v);

v.sort();
println!("sort: {:?}", v);

v.reverse();
println!("reverse: {:?}", v);

v.dedup();                    // Remove consecutive duplicates
println!("dedup: {:?}", v);
```

## 🔄 Iterating

```rust
let v = vec![10, 20, 30, 40, 50];

// Immutable iteration
for val in &v {
    print!("{} ", val);
}
println!();

// With index
for (i, val) in v.iter().enumerate() {
    println!("  v[{}] = {}", i, val);
}

// Mutable iteration
let mut v2 = vec![1, 2, 3, 4, 5];
for val in &mut v2 {
    *val *= 10;
}
println!("Multiplied: {:?}", v2);
```

## 📊 Aggregation

```rust
let scores = vec![85, 92, 78, 95, 88, 73, 91];

let sum: i32 = scores.iter().sum();
let count = scores.len();
let avg = sum as f64 / count as f64;
let max = scores.iter().max().unwrap();
let min = scores.iter().min().unwrap();

println!("Scores: {:?}", scores);
println!("Sum: {}, Count: {}, Avg: {:.1}", sum, count, avg);
println!("Max: {}, Min: {}", max, min);
```

## 📝 Mini-Project: In-Memory To-Do List

```rust
let mut todos: Vec<(String, bool)> = Vec::new();

// Add tasks
fn add_todo(list: &mut Vec<(String, bool)>, task: &str) {
    list.push((task.to_string(), false));
    println!("  ✅ Added: {}", task);
}

fn complete_todo(list: &mut Vec<(String, bool)>, index: usize) {
    if let Some(todo) = list.get_mut(index) {
        todo.1 = true;
        println!("  ✅ Completed: {}", todo.0);
    }
}

fn display_todos(list: &Vec<(String, bool)>) {
    println!("\n📋 To-Do List:");
    println!("  {}", "-".repeat(30));
    for (i, (task, done)) in list.iter().enumerate() {
        let icon = if *done { "✅" } else { "⬜" };
        println!("  {} [{}] {}", icon, i, task);
    }
    let done_count = list.iter().filter(|(_, d)| *d).count();
    println!("  {}/{} completed", done_count, list.len());
}

add_todo(&mut todos, "Learn Rust vectors");
add_todo(&mut todos, "Practice pattern matching");
add_todo(&mut todos, "Build a project");
add_todo(&mut todos, "Push to GitHub");

complete_todo(&mut todos, 0);
complete_todo(&mut todos, 1);

display_todos(&todos);
```

## 🎯 Key Takeaways

1. `Vec<T>` is a **growable** array on the heap
2. Use `vec![]` macro or `Vec::new()`
3. `.get()` is safe (returns `Option`); `[i]` panics on out-of-bounds
4. Iterate with `&v`, `&mut v`, or `.iter()`
