```rust
use std::collections::HashMap;
```

## 🗺️ Creating HashMaps

```rust
// Method 1: new + insert
let mut scores = HashMap::new();
scores.insert("Alice", 95);
scores.insert("Bob", 87);
scores.insert("Charlie", 92);
println!("{:?}", scores);

// Method 2: From arrays of tuples
let colors: HashMap<&str, &str> = HashMap::from([
    ("red", "#FF0000"),
    ("green", "#00FF00"),
    ("blue", "#0000FF"),
]);
println!("{:?}", colors);

// Method 3: collect from iterator
let names = vec!["Alice", "Bob"];
let ages = vec![30, 25];
let people: HashMap<_, _> = names.into_iter().zip(ages.into_iter()).collect();
println!("{:?}", people);
```

## 📖 Accessing Values

```rust
let mut scores = HashMap::new();
scores.insert("Alice", 95);
scores.insert("Bob", 87);

// .get() returns Option<&V> — safe!
println!("Alice: {:?}", scores.get("Alice"));
println!("Eve: {:?}", scores.get("Eve")); // None

// Indexing with [] — panics if key not found!
println!("Bob: {}", scores["Bob"]);

// Check existence
println!("Contains Alice: {}", scores.contains_key("Alice"));
println!("Len: {}", scores.len());
```

## ✏️ Updating Values

```rust
let mut map = HashMap::new();
map.insert("key", 1);
println!("Before: {:?}", map);

// Overwrite
map.insert("key", 2);
println!("Overwrite: {:?}", map);

// Insert only if key doesn't exist
map.entry("key").or_insert(99);      // Won't insert (key exists)
map.entry("new_key").or_insert(99);  // Will insert
println!("Entry: {:?}", map);

// Update based on old value
let counter = map.entry("key").or_insert(0);
*counter += 10;
println!("Updated: {:?}", map);
```

```rust
// Remove
let mut map = HashMap::from([("a", 1), ("b", 2), ("c", 3)]);
let removed = map.remove("b");
println!("Removed: {:?}, Map: {:?}", removed, map);
```

## 🔄 Iterating

```rust
let scores = HashMap::from([
    ("Alice", 95), ("Bob", 87), ("Charlie", 92)
]);

// Key-value pairs
for (name, score) in &scores {
    println!("{}: {}", name, score);
}

// Keys only
let names: Vec<&&str> = scores.keys().collect();
println!("Names: {:?}", names);

// Values only
let vals: Vec<&i32> = scores.values().collect();
println!("Scores: {:?}", vals);
```

## 📊 Word Frequency Counter

```rust
let text = "the quick brown fox jumps over the lazy dog the fox the dog";

let mut freq: HashMap<&str, i32> = HashMap::new();
for word in text.split_whitespace() {
    let count = freq.entry(word).or_insert(0);
    *count += 1;
}

// Sort by frequency
let mut sorted: Vec<(&&str, &i32)> = freq.iter().collect();
sorted.sort_by(|a, b| b.1.cmp(a.1));

println!("Word frequencies:");
for (word, count) in &sorted {
    let bar = "█".repeat(**count as usize);
    println!("  {:>8} [{}] {}", word, count, bar);
}
```

## 🎯 Key Takeaways

1. `HashMap<K, V>` stores key-value pairs (import from `std::collections`)
2. `.get()` returns `Option` (safe); `[]` panics on missing key
3. `.entry().or_insert()` is the idiomatic way to update-or-insert
4. HashMap order is **not guaranteed**
