# 🦀 Day 2: Loops — `loop`, `while`, `for`

Today: all three loop types in Rust, plus `break`, `continue`, and loop labels.

---

## 🔁 `loop` — Infinite Loop

```rust
// loop runs forever until you break
let mut count = 0;
loop {
    count += 1;
    if count == 5 {
        break;
    }
}
println!("Counted to {}", count);

// loop can RETURN a value via break!
let mut n = 0;
let result = loop {
    n += 1;
    if n * n > 100 {
        break n;  // return n from the loop
    }
};
println!("First n where n²>100: {} ({}²={})", result, result, result * result);
```

## 🔄 `while` Loop

```rust
let mut countdown = 5;
while countdown > 0 {
    println!("{}...", countdown);
    countdown -= 1;
}
println!("Liftoff! 🚀");
```

## 🎯 `for` Loop

```rust
// Range: start..end (exclusive) or start..=end (inclusive)
print!("1..5:  ");
for i in 1..5 {
    print!("{} ", i);
}
println!();

print!("1..=5: ");
for i in 1..=5 {
    print!("{} ", i);
}
println!();

// Reverse
print!("Rev:   ");
for i in (1..=5).rev() {
    print!("{} ", i);
}
println!();

// Step by
print!("Step2: ");
for i in (0..=10).step_by(2) {
    print!("{} ", i);
}
println!();
```

```rust
// Iterating over collections
let fruits = ["apple", "banana", "cherry"];

for fruit in &fruits {
    println!("I like {}!", fruit);
}

// With index using enumerate
for (i, fruit) in fruits.iter().enumerate() {
    println!("  [{}] {}", i, fruit);
}
```

## 🏷️ Loop Labels & Nested Loops

```rust
// Labels let you break/continue specific outer loops
'outer: for i in 0..5 {
    for j in 0..5 {
        if i + j == 6 {
            println!("Breaking outer at i={}, j={}", i, j);
            break 'outer;
        }
    }
}

// continue with label
println!("\nSkipping inner loop:");
'outer: for i in 0..3 {
    for j in 0..3 {
        if j == 1 {
            continue 'outer;
        }
        println!("  i={}, j={}", i, j);
    }
}
```

## 🎵 Practical: Multiplication Table

```rust
println!("Multiplication Table (1-9):\n");
print!("     ");
for i in 1..=9 {
    print!("{:>4}", i);
}
println!("\n    {}", "----".repeat(9));

for i in 1..=9 {
    print!("{:>2} | ", i);
    for j in 1..=9 {
        print!("{:>4}", i * j);
    }
    println!();
}
```
