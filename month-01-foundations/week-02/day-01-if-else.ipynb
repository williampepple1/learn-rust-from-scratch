## 🔀 Basic `if`/`else`

```rust
let temperature = 25;

if temperature > 30 {
    println!("It's hot! 🔥");
} else if temperature > 20 {
    println!("It's pleasant! 😊");
} else if temperature > 10 {
    println!("It's cool! 🧥");
} else {
    println!("It's cold! 🥶");
}

// ⚠️ Condition MUST be a bool — no truthy/falsy values!
// if 1 { }     // ❌ ERROR: expected bool, found integer
// if "yes" { } // ❌ ERROR
```

## 💡 `if` as an Expression

In Rust, `if` is an expression — it returns a value!

```rust
let age = 20;
let status = if age >= 18 { "adult" } else { "minor" };
println!("{} is a {}", age, status);

let x = 10;
let abs_x = if x >= 0 { x } else { -x };
println!("abs({}) = {}", x, abs_x);

// Both branches must return the SAME type!
// let bad = if true { 5 } else { "hello" }; // ❌ ERROR
```

```rust
// Nested if expressions
let score = 85;
let grade = if score >= 90 {
    'A'
} else if score >= 80 {
    'B'
} else if score >= 70 {
    'C'
} else if score >= 60 {
    'D'
} else {
    'F'
};
println!("Score {} = Grade {}", score, grade);
```

## 🔗 Logical Operators

```rust
let age = 25;
let has_license = true;

// AND (&&), OR (||), NOT (!)
if age >= 18 && has_license {
    println!("Can drive! 🚗");
}

let is_weekend = true;
let is_holiday = false;
if is_weekend || is_holiday {
    println!("Day off! 🎉");
}

if !is_holiday {
    println!("Not a holiday");
}

// Range check
let x = 15;
if x >= 10 && x <= 20 {
    println!("{} is between 10 and 20", x);
}
```

## 🎯 FizzBuzz!

```rust
for n in 1..=30 {
    if n % 15 == 0 {
        print!("FizzBuzz ");
    } else if n % 3 == 0 {
        print!("Fizz ");
    } else if n % 5 == 0 {
        print!("Buzz ");
    } else {
        print!("{} ", n);
    }
}
println!();
```

## 🎯 Key Takeaways

1. `if` conditions **must** be `bool` — no truthy/falsy
2. `if`/`else` is an **expression** — can assign to variables
3. Both branches must return the **same type**
4. Use `&&`, `||`, `!` for logical operations
