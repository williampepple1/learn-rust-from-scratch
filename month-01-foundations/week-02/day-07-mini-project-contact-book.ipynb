# 🦀 Day 7: Mini-Project — Contact Book

Let's build a contact book combining HashMaps, Vectors, and iterators!

Features:
- Add/remove contacts
- Search contacts
- List and filter contacts
- Contact statistics

---

```rust
use std::collections::HashMap;

// Contact data structure
#[derive(Debug, Clone)]
struct Contact {
    name: String,
    phone: String,
    email: String,
    category: String,
}

impl Contact {
    fn new(name: &str, phone: &str, email: &str, category: &str) -> Self {
        Contact {
            name: name.to_string(),
            phone: phone.to_string(),
            email: email.to_string(),
            category: category.to_string(),
        }
    }
}

println!("Contact struct defined! ✅");
```

```rust
// Contact Book using HashMap
struct ContactBook {
    contacts: HashMap<String, Contact>,
}

impl ContactBook {
    fn new() -> Self {
        ContactBook { contacts: HashMap::new() }
    }

    fn add(&mut self, contact: Contact) {
        println!("  ✅ Added: {}", contact.name);
        self.contacts.insert(contact.name.clone(), contact);
    }

    fn remove(&mut self, name: &str) -> bool {
        if self.contacts.remove(name).is_some() {
            println!("  🗑️ Removed: {}", name);
            true
        } else {
            println!("  ❌ Not found: {}", name);
            false
        }
    }

    fn find(&self, name: &str) -> Option<&Contact> {
        self.contacts.get(name)
    }

    fn search(&self, query: &str) -> Vec<&Contact> {
        let q = query.to_lowercase();
        self.contacts.values()
            .filter(|c| {
                c.name.to_lowercase().contains(&q) ||
                c.email.to_lowercase().contains(&q) ||
                c.phone.contains(&q)
            })
            .collect()
    }

    fn by_category(&self, category: &str) -> Vec<&Contact> {
        self.contacts.values()
            .filter(|c| c.category == category)
            .collect()
    }

    fn list_all(&self) {
        let border = "═".repeat(60);
        println!("\n╔{}╗", border);
        println!("║{:^60}║", "📒 CONTACT BOOK");
        println!("╠{}╣", border);

        let mut names: Vec<&String> = self.contacts.keys().collect();
        names.sort();

        for name in names {
            let c = &self.contacts[name];
            println!("║ {:20} │ {:15} │ {:>18} ║", c.name, c.phone, c.category);
        }

        println!("╠{}╣", border);
        println!("║{:^60}║", format!("{} contacts total", self.contacts.len()));
        println!("╚{}╝", border);
    }

    fn stats(&self) {
        let mut categories: HashMap<&str, usize> = HashMap::new();
        for c in self.contacts.values() {
            *categories.entry(c.category.as_str()).or_insert(0) += 1;
        }
        println!("\n📊 Contact Statistics:");
        println!("  Total contacts: {}", self.contacts.len());
        for (cat, count) in &categories {
            println!("  {}: {} {}", cat, count, "█".repeat(*count));
        }
    }
}

println!("ContactBook defined! ✅");
```

```rust
// Create and populate the contact book
let mut book = ContactBook::new();

println!("📥 Adding contacts:");
book.add(Contact::new("Alice Johnson", "+1-555-0101", "alice@email.com", "Friend"));
book.add(Contact::new("Bob Smith", "+1-555-0102", "bob@work.com", "Work"));
book.add(Contact::new("Charlie Brown", "+1-555-0103", "charlie@email.com", "Friend"));
book.add(Contact::new("Diana Prince", "+1-555-0104", "diana@work.com", "Work"));
book.add(Contact::new("Eve Wilson", "+1-555-0105", "eve@family.com", "Family"));
book.add(Contact::new("Frank Castle", "+1-555-0106", "frank@email.com", "Work"));

book.list_all();
```

```rust
// Search functionality
println!("🔍 Searching for 'alice':");
for c in book.search("alice") {
    println!("  Found: {} — {} ({})", c.name, c.email, c.phone);
}

println!("\n🔍 Work contacts:");
for c in book.by_category("Work") {
    println!("  {} — {}", c.name, c.email);
}

// Look up specific contact
println!("\n📇 Looking up Bob Smith:");
if let Some(c) = book.find("Bob Smith") {
    println!("  Name:  {}", c.name);
    println!("  Phone: {}", c.phone);
    println!("  Email: {}", c.email);
    println!("  Group: {}", c.category);
}
```

```rust
// Remove and stats
println!("\n🗑️ Removing contacts:");
book.remove("Charlie Brown");
book.remove("Unknown Person");

book.stats();
book.list_all();
```
