## 🎯 Basic `match`

```rust
let day = 3;
let day_name = match day {
    1 => "Monday",
    2 => "Tuesday",
    3 => "Wednesday",
    4 => "Thursday",
    5 => "Friday",
    6 => "Saturday",
    7 => "Sunday",
    _ => "Invalid day",  // _ is the catch-all (required!)
};
println!("Day {} is {}", day, day_name);
```

```rust
// match MUST be exhaustive — cover all cases!
// Multiple patterns with |
let day = 6;
let day_type = match day {
    1 | 2 | 3 | 4 | 5 => "Weekday",
    6 | 7 => "Weekend",
    _ => "Invalid",
};
println!("Day {} is a {}", day, day_type);
```

```rust
// Ranges in match
let score = 85;
let grade = match score {
    90..=100 => 'A',
    80..=89  => 'B',
    70..=79  => 'C',
    60..=69  => 'D',
    0..=59   => 'F',
    _        => '?',
};
println!("Score {} = Grade {}", score, grade);
```

```rust
// Match with char
let c = 'x';
let kind = match c {
    'a'..='z' => "lowercase letter",
    'A'..='Z' => "uppercase letter",
    '0'..='9' => "digit",
    ' '       => "space",
    _         => "other",
};
println!("'{}' is a {}", c, kind);
```

## 🛡️ Match Guards

```rust
let num = 4;
let desc = match num {
    n if n < 0  => "negative",
    0           => "zero",
    n if n % 2 == 0 => "positive even",
    _           => "positive odd",
};
println!("{} is {}", num, desc);
```

## 📦 Destructuring in Match

```rust
// Matching tuples
let point = (0, 5);
let location = match point {
    (0, 0)     => "origin",
    (x, 0)     => "on x-axis",
    (0, y)     => "on y-axis",
    (x, y) if x == y => "on diagonal",
    _          => "somewhere else",
};
println!("{:?} is {}", point, location);
```

```rust
// Matching with binding (@)
let age = 25;
match age {
    n @ 0..=12   => println!("{} — child", n),
    n @ 13..=17  => println!("{} — teenager", n),
    n @ 18..=64  => println!("{} — adult", n),
    n @ 65..=150 => println!("{} — senior", n),
    _ => println!("invalid age"),
}
```

## 🆚 `if let` — Concise Matching

```rust
// When you only care about ONE pattern, use if let
let favorite_color: Option<&str> = Some("blue");

// Instead of:
match favorite_color {
    Some(color) => println!("Favorite: {}", color),
    None => {},
}

// Use if let:
if let Some(color) = favorite_color {
    println!("Favorite: {}", color);
}
```

## 🎯 Key Takeaways

1. `match` must be **exhaustive** — `_` is the catch-all
2. `match` is an **expression** — returns a value
3. Use `|` for multiple patterns, `..=` for ranges
4. **Guards** add `if` conditions to arms
5. `if let` is a shorthand for single-pattern matching
