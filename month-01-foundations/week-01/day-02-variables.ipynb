## 📦 Variables with `let`

In Rust, variables are **immutable by default**. This is a key design decision that helps prevent bugs.

```rust
// Declaring an immutable variable
let x = 5;
println!("The value of x is: {}", x);

// Rust infers the type — x is i32 (32-bit integer) by default
// You can also specify the type explicitly:
let y: i32 = 10;
println!("The value of y is: {}", y);
```

```rust
// ❌ This would cause a compile error!
// Uncomment the line below to see the error:

// let x = 5;
// x = 6;  // ERROR: cannot assign twice to immutable variable `x`
```

```rust
// Rust has powerful type inference — it figures out types for you
let a = 42;          // i32 (default integer type)
let b = 3.14;        // f64 (default float type)
let c = true;        // bool
let d = 'Z';         // char (single Unicode character, uses single quotes)
let e = "hello";     // &str (string slice)

println!("a: {} (integer)", a);
println!("b: {} (float)", b);
println!("c: {} (boolean)", c);
println!("d: {} (character)", d);
println!("e: {} (string slice)", e);

// You can see the type Rust inferred using a trick:
// let _: () = a;  // Uncomment to see the error revealing a's type!
```

### Type Annotations

Sometimes you need to tell Rust the type explicitly — especially when multiple types are possible.

```rust
// Explicit type annotations — the type goes after a colon
let age: u32 = 25;           // unsigned 32-bit integer
let temperature: f64 = -3.5; // 64-bit float
let initial: char = 'R';     // Unicode character
let active: bool = true;     // boolean

println!("Age: {}, Temp: {}, Initial: {}, Active: {}", age, temperature, initial, active);

// Type annotations are REQUIRED when Rust can't infer the type
let guess: i32 = "42".parse().unwrap();  // parse() could return many types!
println!("Parsed: {}", guess);

// Type suffix — another way to specify type
let x = 42u8;       // u8 (unsigned 8-bit)
let y = 3.14f32;    // f32 (32-bit float)
let z = 1_000i64;   // i64 (64-bit integer)
println!("x={}, y={}, z={}", x, y, z);
```

### Why immutable by default?

1. **Safety**: Prevents accidental modification of data
2. **Clarity**: When reading code, you know a value won't change
3. **Concurrency**: Immutable data is safe to share between threads
4. **Optimization**: The compiler can optimize immutable values more aggressively

## 🔄 Mutable Variables with `let mut`

When you *do* need to change a value, use `mut`:

```rust
let mut score = 0;
println!("Initial score: {}", score);

score = 10;
println!("After first round: {}", score);

score = score + 5;
println!("After bonus: {}", score);

score += 3;  // Shorthand for score = score + 3
println!("Final score: {}", score);
```

```rust
// mut works with all types
let mut name = String::from("Alice");
println!("Name: {}", name);

name = String::from("Bob");
println!("Name changed to: {}", name);

let mut is_active = true;
println!("Active: {}", is_active);

is_active = false;
println!("Active: {}", is_active);
```

### ⚠️ Important: You can't change the TYPE of a mutable variable

### Compound Assignment Operators

Rust supports shorthand operators for updating mutable variables.

```rust
let mut n = 100;

n += 10;   // n = n + 10  → 110
println!("After +=  : {}", n);

n -= 20;   // n = n - 20  → 90
println!("After -=  : {}", n);

n *= 3;    // n = n * 3   → 270
println!("After *=  : {}", n);

n /= 9;    // n = n / 9   → 30
println!("After /=  : {}", n);

n %= 7;    // n = n % 7   → 2
println!("After %%=  : {}", n);

// Bitwise operators work too
let mut flags: u8 = 0b0000_0000;
flags |= 0b0000_0101;   // Set bits 0 and 2
println!("Flags: {:08b}", flags);
```

```rust
// ❌ This would fail:
// let mut x = 5;
// x = "hello";  // ERROR: expected integer, found &str

// mut only allows changing the VALUE, not the TYPE
let mut x = 5;
x = 10;  // ✅ Same type (i32), different value
println!("x = {}", x);
```

## 🔒 Constants with `const`

Constants are values that are **always immutable** and must be known at compile time.

```rust
// Constants MUST have a type annotation
const MAX_SCORE: u32 = 100;
const PI: f64 = 3.14159265358979;
const APP_NAME: &str = "Rust Learning App";

println!("Max score: {}", MAX_SCORE);
println!("Pi: {}", PI);
println!("App: {}", APP_NAME);
```

```rust
// Constants can be computed from other constants at compile time
const SECONDS_IN_MINUTE: u32 = 60;
const MINUTES_IN_HOUR: u32 = 60;
const HOURS_IN_DAY: u32 = 24;
const SECONDS_IN_DAY: u32 = SECONDS_IN_MINUTE * MINUTES_IN_HOUR * HOURS_IN_DAY;

println!("There are {} seconds in a day", SECONDS_IN_DAY);
```

### `const` vs `let` — Key Differences

| Feature | `const` | `let` |
|---------|---------|-------|
| Mutability | Always immutable | Immutable by default, `mut` optional |
| Type annotation | **Required** | Optional (type inference) |
| Compile-time value | **Required** | Can be runtime value |
| Scope | Any scope (including global) | Block scope only |
| Naming convention | `SCREAMING_SNAKE_CASE` | `snake_case` |
| Shadowing | Cannot be shadowed | Can be shadowed |

### Numeric Literals & Underscores

Rust lets you write numbers in various formats and use underscores for readability.

```rust
// Underscores in numbers — for readability (compiler ignores them)
let million = 1_000_000;
let pi = 3.141_592_653;
let bytes = 0xFF_FF;
let bits = 0b1111_0000;

println!("Million: {}", million);
println!("Pi:      {}", pi);
println!("Bytes:   {} (hex: {:#X})", bytes, bytes);
println!("Bits:    {} (binary: {:08b})", bits, bits);

// Different numeric bases
let decimal = 255;
let hex = 0xFF;        // prefix 0x
let octal = 0o377;     // prefix 0o
let binary = 0b1111_1111;  // prefix 0b
let byte = b'A';       // byte literal (u8 value of ASCII char)

println!("\nAll represent the same value:");
println!("  Decimal: {}", decimal);
println!("  Hex:     {}", hex);
println!("  Octal:   {}", octal);
println!("  Binary:  {}", binary);
println!("  Byte 'A': {}", byte);
```

## 👻 Shadowing

Shadowing lets you declare a new variable with the same name as a previous one. The new variable **shadows** the old one.

```rust
let x = 5;
println!("x = {}", x);  // 5

let x = x + 1;  // Shadows the previous x
println!("x = {}", x);  // 6

let x = x * 2;  // Shadows again
println!("x = {}", x);  // 12
```

### Shadowing vs. `mut` — The Key Difference

With shadowing, you can **change the type**!

```rust
// Shadowing allows type changes
let spaces = "   ";           // &str
println!("spaces (str): '{}'", spaces);

let spaces = spaces.len();     // usize — different type, same name!
println!("spaces (num): {}", spaces);

// This is very useful for transforming data while keeping
// a meaningful variable name
```

```rust
// Shadowing in inner scopes
let x = 5;
println!("Outer x: {}", x);  // 5

{
    let x = x * 2;  // This x only exists in this block
    println!("Inner x: {}", x);  // 10
}

println!("Outer x again: {}", x);  // 5 (unchanged!)
```

```rust
// Practical example: parsing user input
let input = "42";                              // Start as &str
let input: i32 = input.parse().unwrap();       // Shadow as i32
let input = input * 2;                          // Shadow with computation
println!("Result: {}", input);                  // 84
```

## 📛 Variable Naming Conventions

Rust has strong naming conventions enforced by the compiler (warnings if you break them).

```rust
// ✅ snake_case for variables and functions (Rust convention)
let user_name = "Alice";
let max_retries = 3;
let is_valid = true;

// ✅ SCREAMING_SNAKE_CASE for constants and statics
const MAX_CONNECTIONS: u32 = 100;
static APP_VERSION: &str = "1.0.0";

// ⚠️ The compiler warns about non-standard names:
// let UserName = "Bob";  // warning: variable should be snake_case

// Prefix with _ to suppress "unused variable" warnings
let _temporary = 42;  // no warning even if unused
let _ = "discarded";  // completely discards the value

println!("User: {}, Retries: {}, Valid: {}", user_name, max_retries, is_valid);
println!("Max connections: {}, Version: {}", MAX_CONNECTIONS, APP_VERSION);
```

## 🧩 Destructuring — Multiple Variables at Once

You can unpack tuples and other structures into multiple variables in a single `let` statement.

```rust
// Destructure a tuple into separate variables
let (x, y, z) = (10, 20, 30);
println!("x={}, y={}, z={}", x, y, z);

// Useful for returning multiple values from expressions
let (width, height) = (1920, 1080);
println!("Resolution: {}x{}", width, height);

// Use _ to ignore values you don't need
let (name, _, age) = ("Alice", "ignored", 30);
println!("{} is {} years old", name, age);

// Destructuring with type annotations
let (a, b): (f64, f64) = (3.14, 2.71);
println!("a={:.2}, b={:.2}", a, b);
```

## 📐 Static Variables

`static` variables have a fixed memory address and live for the entire program.

```rust
static GREETING: &str = "Hello, Rustacean!";
println!("{}", GREETING);

// static mut exists but is unsafe (we'll cover unsafe much later)
// For now, prefer const over static in most cases
```

## 🧰 Practical Examples

Let's put everything together with real-world-style scenarios.

```rust
// Example 1: Shopping cart total
const TAX_RATE: f64 = 0.08;  // 8% sales tax

let item1_price = 29.99;
let item2_price = 49.99;
let item3_price = 9.99;

let subtotal = item1_price + item2_price + item3_price;
let tax = subtotal * TAX_RATE;
let total = subtotal + tax;

println!("╔════════════════════════════╗");
println!("║    SHOPPING CART           ║");
println!("╠════════════════════════════╣");
println!("║ Item 1:       ${:>8.2}    ║", item1_price);
println!("║ Item 2:       ${:>8.2}    ║", item2_price);
println!("║ Item 3:       ${:>8.2}    ║", item3_price);
println!("╠════════════════════════════╣");
println!("║ Subtotal:     ${:>8.2}    ║", subtotal);
println!("║ Tax (8%):     ${:>8.2}    ║", tax);
println!("║ TOTAL:        ${:>8.2}    ║", total);
println!("╚════════════════════════════╝");
```

```rust
// Example 2: Unit converter using shadowing
let distance = "42.195";                          // marathon distance as string
let distance: f64 = distance.parse().unwrap();    // shadow as float (km)
let distance_miles = distance * 0.621371;         // convert to miles

println!("Marathon distance:");
println!("  {:.3} km", distance);
println!("  {:.3} miles", distance_miles);

// Example 3: Swap two variables
let mut a = 10;
let mut b = 20;
println!("\nBefore swap: a={}, b={}", a, b);

let temp = a;
a = b;
b = temp;
println!("After swap:  a={}, b={}", a, b);

// Rust's tuple destructuring makes swapping elegant:
let (a, b) = (b, a);
println!("Swapped back: a={}, b={}", a, b);
```

```rust
// Example 4: Fibonacci using mutable variables
let mut a = 0u64;
let mut b = 1u64;
let n = 15;

print!("First {} Fibonacci numbers: {}", n, a);
for _ in 1..n {
    print!(", {}", b);
    let next = a + b;
    a = b;
    b = next;
}
println!();
```

## ⚠️ Common Mistakes

Here are mistakes every Rust beginner makes — learn them now to save debugging later!

```rust
// Mistake 1: Forgetting `mut` when you need to modify a variable
// let count = 0;
// count += 1;  // ERROR: cannot assign twice to immutable variable
// Fix: let mut count = 0;

// Mistake 2: Using `mut` when shadowing would be cleaner
// ❌ Mutable but only assigned once:
let mut result = 0;
result = 5 * 10;
println!("Mutable (unnecessary): {}", result);

// ✅ Better — just shadow:
let result = 5 * 10;
println!("Shadowed (cleaner): {}", result);

// Mistake 3: Thinking shadowing modifies the original
let x = 5;
{
    let x = 99;  // this is a NEW variable, not the outer x
    println!("Inner x: {}", x);
}
println!("Outer x is still: {}", x);

// Mistake 4: Constants without type annotations
// const X = 5;  // ERROR: missing type for `const` item
// Fix: const X: i32 = 5;

// Mistake 5: Trying to use a variable before declaring it
// println!("{}", z);  // ERROR: cannot find value `z`
// let z = 42;
// Rust reads top-to-bottom — declare before use!

println!("\nAll examples compiled successfully!");
```

## 🎯 Key Takeaways

1. Variables are **immutable by default** — use `let mut` when you need to change values
2. Rust has powerful **type inference**, but you can add explicit annotations: `let x: i32 = 5`
3. **Constants** (`const`) require type annotations, compile-time values, and `SCREAMING_SNAKE_CASE`
4. **Shadowing** (`let x = ...` again) creates a new variable and can change the type
5. `mut` only changes the value, **not** the type — shadowing can change both
6. **Compound assignment** operators: `+=`, `-=`, `*=`, `/=`, `%=`
7. Use **underscores** in numbers for readability: `1_000_000`, `0xFF_FF`, `0b1111_0000`
8. **Naming**: `snake_case` for variables, `SCREAMING_SNAKE_CASE` for constants
9. Prefix unused variables with `_` to suppress compiler warnings
10. **Destructuring** lets you unpack tuples: `let (x, y) = (10, 20)`
11. Blocks `{ }` create scopes — inner variables don't affect outer ones
