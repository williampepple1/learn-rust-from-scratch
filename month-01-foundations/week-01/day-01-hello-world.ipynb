## 🤔 What is Rust?

Rust is a systems programming language focused on:

- **Safety**: Memory safety without garbage collection
- **Speed**: Performance comparable to C and C++
- **Concurrency**: Fearless concurrency through ownership

Rust has been voted the **most loved programming language** on Stack Overflow for multiple years running!

## 🚀 Your First Rust Program

The classic "Hello, World!" — every programmer's first step in a new language.

````rust
// Your very first Rust program!
println!("Hello, World!");
````

### Breaking it down:

- `println!` is a **macro** (note the `!`), not a function
- Macros in Rust are indicated by `!` after the name
- Strings use **double quotes** `""`
- Statements end with a **semicolon** `;`

````rust
// Print your name!
println!("Hello, my name is Rust Learner!");
println!("I'm starting my 6-month Rust journey today!");
````

## 📝 The `println!` Macro in Detail

### Basic printing

````rust
// print! does NOT add a newline at the end
print!("Hello ");
print!("World");
println!(); // Just prints a newline

// println! DOES add a newline
println!("This is on its own line");
println!("And so is this");
````

### String Interpolation with `{}`

Use `{}` as placeholders for values in your format string.

````rust
// Positional placeholders
println!("My name is {} and I am {} years old.", "Alice", 30);

// Multiple values
println!("{} + {} = {}", 2, 3, 2 + 3);

// Repeating a value by index
println!("{0} is {1} and {0} is also great!", "Rust", "fast");
````

````rust
// Named parameters
println!("{language} is {adjective}!", language = "Rust", adjective = "awesome");

// Inline variables (Rust 1.58+)
let language = "Rust";
let year = 2026;
println!("Learning {language} in {year}!");
````

### Formatting Numbers

````rust
// Decimal places
let pi = 3.14159265;
println!("Pi is approximately {:.2}", pi);      // 2 decimal places
println!("Pi is approximately {:.4}", pi);      // 4 decimal places

// Padding and alignment
println!("|{:>10}|", "right");     // Right-align in 10 chars
println!("|{:<10}|", "left");      // Left-align in 10 chars
println!("|{:^10}|", "center");    // Center in 10 chars
println!("|{:*^10}|", "hi");       // Center with * fill
````

````rust
// Number bases
let num = 255;
println!("Decimal:     {}", num);
println!("Binary:      {:b}", num);
println!("Octal:       {:o}", num);
println!("Hexadecimal: {:x}", num);
println!("Hex (upper): {:X}", num);
println!("With prefix: {:#x}", num);  // 0xff
````

### Debug Printing with `{:?}`

````rust
// {:?} uses the Debug trait - useful for debugging
println!("Debug: {:?}", (1, 2, 3));
println!("Debug: {:?}", [1, 2, 3, 4, 5]);

// {:#?} pretty-prints the debug output
println!("Pretty debug: {:#?}", ("hello", 42, true));
````

## 🎯 Other Print Macros

````rust
// eprintln! prints to stderr (standard error)
eprintln!("This is an error message!");

// format! returns a String instead of printing
let greeting = format!("Hello, {}!", "World");
println!("The greeting is: {}", greeting);
````
