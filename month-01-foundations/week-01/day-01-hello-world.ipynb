## 🤔 What is Rust?

Rust is a systems programming language focused on:

- **Safety**: Memory safety without garbage collection
- **Speed**: Performance comparable to C and C++
- **Concurrency**: Fearless concurrency through ownership

Rust has been voted the **most loved programming language** on Stack Overflow for multiple years running!

## 📦 Your First Cargo Project

Cargo is Rust's build system and package manager. Here's how you'd create a project outside of Jupyter:

```bash
# Create a new project
cargo new hello-rust
cd hello-rust

# Project structure:
# hello-rust/
# ├── Cargo.toml    ← project metadata and dependencies
# └── src/
#     └── main.rs   ← your code goes here

# Build and run
cargo run

# Or just build (output goes to target/debug/)
cargo build

# Build with optimizations (for release)
cargo build --release

# Check for errors without building (faster!)
cargo check
```

The generated `src/main.rs` contains:
```rust
fn main() {
    println!("Hello, world!");
}
```

`fn main()` is the **entry point** of every Rust program. In these EvCxR notebooks, we can run code directly without `fn main()` — the kernel handles that for us.

### Cargo.toml — Your Project's Identity

Every Cargo project has a `Cargo.toml` file:

```toml
[package]
name = "hello-rust"
version = "0.1.0"
edition = "2021"          # Rust edition (2015, 2018, 2021, 2024)

[dependencies]
# External crates (libraries) go here
# rand = "0.8"            # Example: add the `rand` crate
```

Key concepts:
- **Crate** = a Rust package (library or binary)
- **crates.io** = Rust's package registry (like npm or pip)
- **Edition** = a set of language features; `2021` is the current default

## 📝 The `println!` Macro in Detail

### Basic printing

````rust
// print! does NOT add a newline at the end
print!("Hello ");
print!("World");
println!(); // Just prints a newline

// println! DOES add a newline
println!("This is on its own line");
println!("And so is this");
````

### String Interpolation with `{}`

Use `{}` as placeholders for values in your format string.

````rust
// Positional placeholders
println!("My name is {} and I am {} years old.", "Alice", 30);

// Multiple values
println!("{} + {} = {}", 2, 3, 2 + 3);

// Repeating a value by index
println!("{0} is {1} and {0} is also great!", "Rust", "fast");
````

````rust
// Named parameters
println!("{language} is {adjective}!", language = "Rust", adjective = "awesome");

// Inline variables (Rust 1.58+)
let language = "Rust";
let year = 2026;
println!("Learning {language} in {year}!");
````

### Formatting Numbers

````rust
// Decimal places
let pi = 3.14159265;
println!("Pi is approximately {:.2}", pi);      // 2 decimal places
println!("Pi is approximately {:.4}", pi);      // 4 decimal places

// Padding and alignment
println!("|{:>10}|", "right");     // Right-align in 10 chars
println!("|{:<10}|", "left");      // Left-align in 10 chars
println!("|{:^10}|", "center");    // Center in 10 chars
println!("|{:*^10}|", "hi");       // Center with * fill
````

````rust
// Number bases
let num = 255;
println!("Decimal:     {}", num);
println!("Binary:      {:b}", num);
println!("Octal:       {:o}", num);
println!("Hexadecimal: {:x}", num);
println!("Hex (upper): {:X}", num);
println!("With prefix: {:#x}", num);  // 0xff
````

### Debug Printing with `{:?}`

````rust
// {:?} uses the Debug trait - useful for debugging
println!("Debug: {:?}", (1, 2, 3));
println!("Debug: {:?}", [1, 2, 3, 4, 5]);

// {:#?} pretty-prints the debug output
println!("Pretty debug: {:#?}", ("hello", 42, true));
````

## 🔤 Escape Characters

Special characters in strings use the backslash `\` prefix.

````rust
// Common escape sequences
println!("Newline: First line\nSecond line");
println!("Tab: Column1\tColumn2\tColumn3");
println!("Backslash: C:\\Users\\Documents");
println!("Double quote: She said \"hello!\"");
println!("Null character: \\0 is the null char");

// Unicode escape
println!("Unicode: \u{1F980}");  // 🦀
println!("Unicode: \u{2764}");   // ❤

// Raw strings — no escaping needed! Prefix with r
println!(r"Raw string: no \n escaping here, C:\Users\path works");
println!(r#"Raw string with "quotes" inside"#);
````

## 🐛 The `dbg!` Macro — Your Debugging Best Friend

`dbg!` prints the expression, its value, and the file/line — perfect for quick debugging.

````rust
let x = 5;
let y = 10;

// dbg! prints file:line, the expression, and its value
dbg!(x + y);
dbg!(x * y);

// dbg! returns the value, so you can use it inline
let result = dbg!(x + y) * 2;
println!("Result: {}", result);

// Compare: println! vs dbg!
println!("println: {}", x + y);  // just the value
dbg!(x + y);                      // expression = value + location
````

## 🔢 Basic Arithmetic in Rust

You can do math right away — Rust handles the usual operators.

````rust
// Arithmetic operators
println!("Addition:       5 + 3 = {}", 5 + 3);
println!("Subtraction:    10 - 4 = {}", 10 - 4);
println!("Multiplication: 6 * 7 = {}", 6 * 7);
println!("Division:       15 / 4 = {}", 15 / 4);       // integer division!
println!("Float division: 15.0 / 4.0 = {}", 15.0 / 4.0);
println!("Remainder:      17 % 5 = {}", 17 % 5);

// Watch out: integer division truncates!
println!("\n⚠️  Integer division truncates:");
println!("  7 / 2 = {} (not 3.5!)", 7 / 2);
println!("  -7 / 2 = {}", -7 / 2);
````

## 🌍 Rust vs Other Languages — Quick Comparison

If you're coming from another language, here's how "Hello, World!" looks:

| Language | Code | Key Difference |
|----------|------|----------------|
| **Rust** | `println!("Hello!");` | `!` means it's a macro, not a function |
| Python | `print("Hello!")` | No semicolons, dynamic typing |
| JavaScript | `console.log("Hello!")` | No semicolons required |
| C | `printf("Hello!\n");` | Manual `\n`, no memory safety |
| C++ | `std::cout << "Hello!" << std::endl;` | Stream-based output |
| Java | `System.out.println("Hello!");` | Verbose, needs a class |
| Go | `fmt.Println("Hello!")` | Similar simplicity |

### Why `println!` is a macro, not a function

Functions in Rust must have a fixed number of arguments with known types. `println!` accepts a **variable number** of arguments with **different types** — that's only possible with a macro. The `!` is Rust's way of being explicit about this.

## 🧰 Practical Examples

Let's combine what we've learned into some real-world output patterns.

````rust
// Example 1: Receipt printer
println!("╔══════════════════════════════╗");
println!("║       RUST COFFEE SHOP       ║");
println!("╠══════════════════════════════╣");
println!("║ {:<20} {:>7} ║", "Espresso", "$3.50");
println!("║ {:<20} {:>7} ║", "Latte", "$4.75");
println!("║ {:<20} {:>7} ║", "Cappuccino", "$4.25");
println!("╠══════════════════════════════╣");
println!("║ {:<20} {:>7} ║", "TOTAL", "$12.50");
println!("╚══════════════════════════════╝");
````

````rust
// Example 2: Progress bar
let total = 10;
for i in 0..=total {
    let filled = "█".repeat(i);
    let empty = "░".repeat(total - i);
    let percent = (i as f64 / total as f64) * 100.0;
    println!("[{}{}] {:.0}%", filled, empty, percent);
}
println!("Done!");
````

````rust
// Example 3: Color table (using format specifiers)
println!("{:─^40}", " RGB Color Values ");
println!("{:<12} {:>5} {:>5} {:>5}  {}", "Color", "R", "G", "B", "Hex");
println!("{}", "─".repeat(40));
println!("{:<12} {:>5} {:>5} {:>5}  #{:02X}{:02X}{:02X}", "Red",     255, 0,   0,   255, 0, 0);
println!("{:<12} {:>5} {:>5} {:>5}  #{:02X}{:02X}{:02X}", "Green",   0,   255, 0,   0, 255, 0);
println!("{:<12} {:>5} {:>5} {:>5}  #{:02X}{:02X}{:02X}", "Blue",    0,   0,   255, 0, 0, 255);
println!("{:<12} {:>5} {:>5} {:>5}  #{:02X}{:02X}{:02X}", "Yellow",  255, 255, 0,   255, 255, 0);
println!("{:<12} {:>5} {:>5} {:>5}  #{:02X}{:02X}{:02X}", "Cyan",    0,   255, 255, 0, 255, 255);
println!("{:<12} {:>5} {:>5} {:>5}  #{:02X}{:02X}{:02X}", "Magenta", 255, 0,   255, 255, 0, 255);
````

## ⚠️ Common Mistakes (and What the Compiler Tells You)

Rust's compiler gives incredibly helpful error messages. Let's see some common beginner mistakes and understand the errors. These cells show what **wrong** code looks like — read the comments to understand why.

````rust
// ✅ Correct: strings use double quotes
println!("Hello!");

// ❌ Common mistake: single quotes are for CHARACTERS, not strings
// println!('Hello!');  // ERROR: expected `char`, found `str`
// In Rust, 'a' is a char, "a" is a string

// ✅ Single character with single quotes:
let letter = 'A';
println!("Character: {}", letter);

// ❌ Common mistake: forgetting the ! in println
// println("Hello!");  // ERROR: use of undeclared function
// It's println! (with the bang), because it's a macro

// ❌ Common mistake: mismatched {} placeholders
// println!("{} and {}", 1);  // ERROR: 2 positional arguments expected, 1 given

// ✅ Always match the number of {} with the number of arguments
println!("{} and {}", 1, 2);
````

## 📖 Formatting Cheat Sheet

Keep this handy as you go through the course!

| Specifier | What it does | Example |
|-----------|-------------|---------|
| `{}` | Display format | `println!("{}", 42)` → `42` |
| `{:?}` | Debug format | `println!("{:?}", (1,2))` → `(1, 2)` |
| `{:#?}` | Pretty debug | Multi-line, indented output |
| `{:.2}` | 2 decimal places | `3.14159` → `3.14` |
| `{:>10}` | Right-align, width 10 | `"hi"` → `"        hi"` |
| `{:<10}` | Left-align, width 10 | `"hi"` → `"hi        "` |
| `{:^10}` | Center, width 10 | `"hi"` → `"    hi    "` |
| `{:0>5}` | Zero-pad, width 5 | `42` → `00042` |
| `{:*^10}` | Fill with `*`, center | `"hi"` → `"****hi****"` |
| `{:b}` | Binary | `255` → `11111111` |
| `{:o}` | Octal | `255` → `377` |
| `{:x}` | Hex (lowercase) | `255` → `ff` |
| `{:X}` | Hex (uppercase) | `255` → `FF` |
| `{:#x}` | Hex with `0x` prefix | `255` → `0xff` |
| `{:e}` | Scientific notation | `1000.0` → `1e3` |

## 🎯 Key Takeaways

1. **Rust** is a systems language focused on safety, speed, and concurrency
2. **Cargo** is your build tool and package manager — `cargo new`, `cargo run`, `cargo build`
3. `println!` is a **macro** (note the `!`) — not a function — for formatted output
4. Use `{}` for Display formatting and `{:?}` for Debug formatting
5. Rust has powerful string formatting: alignment (`<`, `>`, `^`), padding, number bases (`b`, `o`, `x`)
6. `format!` returns a `String` instead of printing; `eprintln!` prints to stderr
7. `dbg!` is your debugging best friend — shows expression, value, and location
8. **Escape characters**: `\n`, `\t`, `\\`, `\"`, `\u{...}` for Unicode
9. **Raw strings** (`r"..."` or `r#"..."#`) skip all escaping
10. Integer division **truncates** in Rust — use floats for decimal results
