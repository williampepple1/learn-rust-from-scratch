## 🔧 Defining Functions

Rust uses `fn` keyword. Convention is `snake_case` naming.

```rust
fn greet() {
    println!("Hello from a function!");
}

greet();
greet(); // Can call multiple times
```

## 📥 Parameters

```rust
// Parameters MUST have type annotations
fn greet_person(name: &str) {
    println!("Hello, {}!", name);
}

fn add(a: i32, b: i32) {
    println!("{} + {} = {}", a, b, a + b);
}

greet_person("Alice");
greet_person("Bob");
add(3, 7);
add(100, 200);
```

## 📤 Return Values

Functions return values using `->` syntax. The **last expression** (without `;`) is the return value.

```rust
// The last expression IS the return value (no semicolon!)
fn square(x: i32) -> i32 {
    x * x  // No semicolon = this is the return value
}

fn add(a: i32, b: i32) -> i32 {
    a + b
}

println!("5² = {}", square(5));
println!("3 + 4 = {}", add(3, 4));

// You can also use `return` keyword for early returns
fn absolute(x: i32) -> i32 {
    if x < 0 {
        return -x;  // Early return
    }
    x  // Normal return (last expression)
}

println!("abs(-5) = {}", absolute(-5));
println!("abs(3) = {}", absolute(3));
```

## 💡 Expressions vs Statements

This is a **critical** Rust concept!

- **Statements** perform an action, do NOT return a value (end with `;`)
- **Expressions** evaluate to a value (do NOT end with `;`)

```rust
// Statement: let binding (doesn't produce a value)
let x = 5;  // this is a statement

// Expression: block that returns a value
let y = {
    let a = 3;
    let b = 4;
    a + b  // No semicolon = expression = return value of block
};
println!("y = {}", y); // 7

// if-else is an expression!
let max = if x > y { x } else { y };
println!("max = {}", max);
```

```rust
// ⚠️ Adding a semicolon turns an expression into a statement!
fn oops() -> i32 {
    5  // ✅ Returns 5
    // 5; ❌ Would return () — type mismatch!
}

println!("oops() = {}", oops());
```

## 🔄 Returning Multiple Values (Tuples)

```rust
fn divide(a: f64, b: f64) -> (f64, f64) {
    let quotient = a / b;
    let remainder = a % b;
    (quotient, remainder)
}

let (q, r) = divide(17.0, 5.0);
println!("17 / 5 = {} remainder {}", q, r);
```

## 🌡️ Temperature Converter Project

```rust
fn celsius_to_fahrenheit(c: f64) -> f64 {
    c * 9.0 / 5.0 + 32.0
}

fn fahrenheit_to_celsius(f: f64) -> f64 {
    (f - 32.0) * 5.0 / 9.0
}

fn celsius_to_kelvin(c: f64) -> f64 {
    c + 273.15
}

fn kelvin_to_celsius(k: f64) -> f64 {
    k - 273.15
}

// Test all conversions
let temps_c = [0.0, 20.0, 37.0, 100.0];

println!("{:>10} {:>10} {:>10}", "°C", "°F", "K");
println!("{}", "-".repeat(32));
for &c in &temps_c {
    println!("{:>10.1} {:>10.1} {:>10.2}",
        c, celsius_to_fahrenheit(c), celsius_to_kelvin(c));
}
```

## 🔁 Nested Functions & Closures Preview

```rust
// Functions can be nested
fn outer() {
    fn inner(x: i32) -> i32 {
        x * 2
    }
    println!("inner(5) = {}", inner(5));
}
outer();

// Functions as values (function pointers)
fn apply(f: fn(i32) -> i32, value: i32) -> i32 {
    f(value)
}

fn double(x: i32) -> i32 { x * 2 }
fn triple(x: i32) -> i32 { x * 3 }

println!("double(5) = {}", apply(double, 5));
println!("triple(5) = {}", apply(triple, 5));
```

## 🎯 Key Takeaways

1. Functions declared with `fn`, use `snake_case`
2. Parameters **must** have type annotations
3. Return type specified with `->`, last expression (no `;`) is the return value
4. **Expressions** return values; **statements** don't
5. `if/else` and blocks `{}` are expressions in Rust
