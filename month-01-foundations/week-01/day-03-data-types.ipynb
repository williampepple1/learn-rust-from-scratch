## 🔢 Integer Types

Rust has integer types for every size and signedness:

| Size | Signed | Unsigned |
|------|--------|----------|
| 8-bit | `i8` | `u8` |
| 16-bit | `i16` | `u16` |
| 32-bit | `i32` (default) | `u32` |
| 64-bit | `i64` | `u64` |
| 128-bit | `i128` | `u128` |
| arch | `isize` | `usize` |

- **Signed** (`i`): can hold negative values (range: -2^(n-1) to 2^(n-1) - 1)
- **Unsigned** (`u`): only positive values (range: 0 to 2^n - 1)
- **arch**: depends on architecture (32-bit on 32-bit systems, 64-bit on 64-bit)

```rust
// Default integer type is i32
let a = 42;          // i32 by default
let b: i64 = 42;     // Explicitly i64
let c: u8 = 255;     // Unsigned 8-bit (max value)
let d: i8 = -128;    // Signed 8-bit (min value)

println!("a (i32): {}", a);
println!("b (i64): {}", b);
println!("c (u8):  {}", c);
println!("d (i8):  {}", d);
```

```rust
// Integer literals can be written in multiple formats
let decimal = 98_222;        // Underscores for readability
let hex = 0xff;              // Hexadecimal
let octal = 0o77;            // Octal
let binary = 0b1111_0000;    // Binary
let byte = b'A';             // Byte (u8 only)

println!("Decimal:  {}", decimal);
println!("Hex:      {}", hex);
println!("Octal:    {}", octal);
println!("Binary:   {}", binary);
println!("Byte 'A': {}", byte);
```

```rust
// Type suffix notation
let x = 42u32;     // x is u32
let y = 100_i64;   // y is i64
let z = 255u8;     // z is u8

println!("x: {} (u32)", x);
println!("y: {} (i64)", y);
println!("z: {} (u8)", z);
```

### Integer Ranges

```rust
// Explore the min and max values of each type
println!("i8:   {} to {}", i8::MIN, i8::MAX);
println!("i16:  {} to {}", i16::MIN, i16::MAX);
println!("i32:  {} to {}", i32::MIN, i32::MAX);
println!("i64:  {} to {}", i64::MIN, i64::MAX);
println!("i128: {} to {}", i128::MIN, i128::MAX);
println!();
println!("u8:   {} to {}", u8::MIN, u8::MAX);
println!("u16:  {} to {}", u16::MIN, u16::MAX);
println!("u32:  {} to {}", u32::MIN, u32::MAX);
println!("u64:  {} to {}", u64::MIN, u64::MAX);
```

### Integer Arithmetic & Overflow

```rust
// Basic arithmetic
let sum = 5 + 10;
let difference = 95.0 - 4.3;
let product = 4 * 30;
let quotient = 56.7 / 32.2;
let remainder = 43 % 5;

println!("Sum:        {}", sum);
println!("Difference: {}", difference);
println!("Product:    {}", product);
println!("Quotient:   {:.4}", quotient);
println!("Remainder:  {}", remainder);
```

```rust
// Integer division truncates toward zero
println!("7 / 2 = {}", 7 / 2);       // 3, not 3.5!
println!("-7 / 2 = {}", -7 / 2);     // -3
println!("7 / -2 = {}", 7 / -2);     // -3
```

```rust
// Overflow safety: Rust provides checked, wrapping, saturating operations
let a: u8 = 250;

// checked_add returns Option<u8> — None if overflow
println!("250 + 5 (checked):     {:?}", a.checked_add(5));
println!("250 + 10 (checked):    {:?}", a.checked_add(10));

// wrapping_add wraps around (like C)
println!("250 + 10 (wrapping):   {}", a.wrapping_add(10));

// saturating_add caps at MAX
println!("250 + 10 (saturating): {}", a.saturating_add(10));

// overflowing_add returns (result, did_overflow)
println!("250 + 10 (overflowing): {:?}", a.overflowing_add(10));
```

## 🔢 Floating-Point Types

Rust has two floating-point types:
- `f32` — 32-bit (single precision)
- `f64` — 64-bit (double precision, **default**)

```rust
let x = 2.0;         // f64 (default)
let y: f32 = 3.14;   // f32

println!("x (f64): {}", x);
println!("y (f32): {}", y);

// Special float values
let infinity = f64::INFINITY;
let neg_infinity = f64::NEG_INFINITY;
let nan = f64::NAN;

println!("Infinity:     {}", infinity);
println!("Neg Infinity: {}", neg_infinity);
println!("NaN:          {}", nan);
println!("Is NaN?       {}", nan.is_nan());
```

```rust
// Float precision matters!
println!("0.1 + 0.2 = {}", 0.1 + 0.2);  // Not exactly 0.3!
println!("0.1 + 0.2 = {:.20}", 0.1 + 0.2);  // See the full precision

// Useful float methods
let num = -3.7_f64;
println!("abs(-3.7):   {}", num.abs());
println!("ceil(-3.7):  {}", num.ceil());
println!("floor(-3.7): {}", num.floor());
println!("round(-3.7): {}", num.round());
println!("sqrt(16.0):  {}", 16.0_f64.sqrt());
println!("pow(2, 10):  {}", 2.0_f64.powi(10));
```

### ⚠️ You Can't Mix Integer and Float Types

```rust
// ❌ This won't compile:
// let x: i32 = 5;
// let y: f64 = 2.0;
// let z = x + y;  // ERROR: can't add i32 and f64

// ✅ You must explicitly convert:
let x: i32 = 5;
let y: f64 = 2.0;
let z = x as f64 + y;   // Convert i32 to f64 first
println!("5 + 2.0 = {}", z);

// Or convert the other way
let w = x + y as i32;   // Convert f64 to i32 (truncates!)
println!("5 + 2 = {}", w);
```

### Type Casting with `as`

```rust
// Casting between numeric types
let big: i64 = 1_000_000;
let small: i32 = big as i32;   // May lose data if value too large!
println!("i64 {} as i32 = {}", big, small);

let float_val: f64 = 3.99;
let int_val: i32 = float_val as i32;  // Truncates (doesn't round!)
println!("f64 {} as i32 = {}", float_val, int_val);

let negative: i32 = -5;
let unsigned: u32 = negative as u32;  // Wraps around!
println!("i32 {} as u32 = {}", negative, unsigned);
```

## ✅ Boolean Type (`bool`)

Booleans are one byte in size and have exactly two values: `true` and `false`.

```rust
let is_active = true;
let is_deleted: bool = false;

println!("Active:  {}", is_active);
println!("Deleted: {}", is_deleted);

// Booleans from comparisons
let x = 10;
let y = 20;

println!("x == y: {}", x == y);
println!("x != y: {}", x != y);
println!("x < y:  {}", x < y);
println!("x >= y: {}", x >= y);
```

```rust
// Boolean operators
let a = true;
let b = false;

println!("AND: true && false  = {}", a && b);
println!("OR:  true || false  = {}", a || b);
println!("NOT: !true          = {}", !a);

// Short-circuit evaluation
// && stops if first is false, || stops if first is true
let x = 5;
let result = x > 0 && x < 10;  // Both conditions checked
println!("0 < {} < 10? {}", x, result);
```

## 🔤 Character Type (`char`)

Rust's `char` is 4 bytes — it represents a **Unicode Scalar Value**, not just ASCII!

```rust
// char uses SINGLE quotes
let letter = 'A';
let digit = '7';
let emoji = '🦀';
let chinese = '中';
let heart = '❤';

println!("Letter:  {}", letter);
println!("Digit:   {}", digit);
println!("Emoji:   {}", emoji);
println!("Chinese: {}", chinese);
println!("Heart:   {}", heart);
println!("Size of char: {} bytes", std::mem::size_of::<char>());
```

```rust
// Useful char methods
let c = 'A';
println!("'{}' is alphabetic:  {}", c, c.is_alphabetic());
println!("'{}' is numeric:     {}", c, c.is_numeric());
println!("'{}' is uppercase:   {}", c, c.is_uppercase());
println!("'{}' is lowercase:   {}", c, c.is_lowercase());
println!("'{}' is whitespace:  {}", c, c.is_whitespace());
println!("'{}' to lowercase:   {}", c, c.to_lowercase().next().unwrap());

// char to/from integer
println!("'A' as u32: {}", 'A' as u32);  // 65
println!("65 as char: {}", 65u8 as char);  // 'A'
```

## 📏 Checking Type Sizes

```rust
use std::mem;

println!("=== Type Sizes ===");
println!("bool:  {} byte",  mem::size_of::<bool>());
println!("char:  {} bytes", mem::size_of::<char>());
println!("i8:    {} byte",  mem::size_of::<i8>());
println!("i16:   {} bytes", mem::size_of::<i16>());
println!("i32:   {} bytes", mem::size_of::<i32>());
println!("i64:   {} bytes", mem::size_of::<i64>());
println!("i128:  {} bytes", mem::size_of::<i128>());
println!("f32:   {} bytes", mem::size_of::<f32>());
println!("f64:   {} bytes", mem::size_of::<f64>());
println!("usize: {} bytes", mem::size_of::<usize>());
```

## 🎯 Key Takeaways

1. Rust has **specific-size** integer types — choose the right one for your data
2. Default types: `i32` for integers, `f64` for floats
3. **No implicit type conversion** — you must use `as` to cast
4. `char` is 4 bytes and supports full **Unicode**
5. Rust provides safe overflow operations: `checked_*`, `wrapping_*`, `saturating_*`
