# 🦀 Day 5: Tuples and Arrays

Today we learn Rust's **compound types** that group multiple values:
- **Tuples** — fixed-size, mixed types
- **Arrays** — fixed-size, same type

---

## 📦 Tuples

A tuple groups values of **different types** into one compound type. Fixed length — cannot grow or shrink.

```rust
// Creating tuples
let point = (3, 4);                    // (i32, i32)
let person: (&str, i32, bool) = ("Alice", 30, true);
let mixed = (1, 2.5, 'a', true, "hello");

println!("Point: {:?}", point);
println!("Person: {:?}", person);
println!("Mixed: {:?}", mixed);
```

```rust
// Accessing elements with dot notation (0-indexed)
let person = ("Alice", 30, "alice@example.com");

println!("Name:  {}", person.0);
println!("Age:   {}", person.1);
println!("Email: {}", person.2);
```

```rust
// Destructuring — unpack a tuple into variables
let (x, y, z) = (10, 20, 30);
println!("x={}, y={}, z={}", x, y, z);

let (name, age, active) = ("Bob", 25, true);
println!("{} is {} and active={}", name, age, active);

// Use _ to ignore values
let (first, _, third) = (1, 2, 3);
println!("first={}, third={}", first, third);
```

```rust
// Mutable tuples
let mut color = (255, 0, 128);
println!("Before: {:?}", color);
color.1 = 128;
println!("After:  {:?}", color);
```

```rust
// Functions can return tuples (return multiple values!)
fn min_max(a: i32, b: i32) -> (i32, i32) {
    if a < b { (a, b) } else { (b, a) }
}

let (min, max) = min_max(42, 17);
println!("min={}, max={}", min, max);
```

```rust
// Practical example: RGB colors as tuples
let red: (u8, u8, u8) = (255, 0, 0);
let green: (u8, u8, u8) = (0, 255, 0);
let blue: (u8, u8, u8) = (0, 0, 255);
let white: (u8, u8, u8) = (255, 255, 255);

fn describe_color(name: &str, rgb: (u8, u8, u8)) {
    let (r, g, b) = rgb;
    println!("{}: rgb({}, {}, {}) → #{:02X}{:02X}{:02X}", name, r, g, b, r, g, b);
}

describe_color("Red", red);
describe_color("Green", green);
describe_color("Blue", blue);
describe_color("White", white);
```

### Unit Tuple
The empty tuple `()` is called the **unit type**. Functions that don't return a value implicitly return `()`.

```rust
let unit = ();
println!("Unit: {:?}", unit);
println!("Size of (): {} bytes", std::mem::size_of::<()>());
```

## 📚 Arrays

Arrays hold multiple values of the **same type** with a **fixed length** known at compile time.

```rust
// Creating arrays
let numbers = [1, 2, 3, 4, 5];           // [i32; 5]
let names: [&str; 3] = ["Alice", "Bob", "Charlie"];
let zeros = [0; 10];                      // [0, 0, 0, ...] (10 zeros)
let flags = [true; 5];                    // [true, true, true, true, true]

println!("Numbers: {:?}", numbers);
println!("Names:   {:?}", names);
println!("Zeros:   {:?}", zeros);
println!("Flags:   {:?}", flags);
```

```rust
// Accessing elements (0-indexed)
let fruits = ["apple", "banana", "cherry", "date"];
println!("First:  {}", fruits[0]);
println!("Second: {}", fruits[1]);
println!("Last:   {}", fruits[fruits.len() - 1]);
println!("Length: {}", fruits.len());
```

```rust
// Mutable arrays
let mut scores = [0; 5];
println!("Before: {:?}", scores);

scores[0] = 95;
scores[1] = 87;
scores[2] = 92;
scores[3] = 78;
scores[4] = 88;
println!("After:  {:?}", scores);
```

```rust
// Iterating over arrays
let numbers = [10, 20, 30, 40, 50];

// for loop
for n in &numbers {
    print!("{} ", n);
}
println!();

// with index
for (i, n) in numbers.iter().enumerate() {
    println!("  [{}] = {}", i, n);
}
```

```rust
// Array slices
let arr = [1, 2, 3, 4, 5, 6, 7, 8];
let slice = &arr[2..5];  // [3, 4, 5]
println!("Slice [2..5]: {:?}", slice);

let first_three = &arr[..3];
let last_three = &arr[5..];
println!("First 3: {:?}", first_three);
println!("Last 3:  {:?}", last_three);
```

```rust
// Useful array methods
let nums = [3, 1, 4, 1, 5, 9, 2, 6];

println!("Contains 5: {}", nums.contains(&5));
println!("Contains 7: {}", nums.contains(&7));

// iter() methods
let sum: i32 = nums.iter().sum();
let max = nums.iter().max().unwrap();
let min = nums.iter().min().unwrap();
println!("Sum: {}, Max: {}, Min: {}", sum, max, min);

// Sorting (need a mutable copy)
let mut sorted = nums.clone();
sorted.sort();
println!("Sorted: {:?}", sorted);
```

## 🆚 Tuples vs Arrays

| Feature | Tuple | Array |
|---------|-------|-------|
| Types | Mixed | Same type only |
| Access | `.0`, `.1`... | `[0]`, `[1]`... |
| Iterate | Not directly | Yes |
| Destructure | Yes | Yes (patterns) |
| Use case | Group related different data | Collection of same-type data |

## 🎯 Key Takeaways

1. **Tuples** hold mixed types, accessed with `.0`, `.1`, destructuring
2. **Arrays** hold same type, fixed size, accessed with `[index]`
3. Both are **stack-allocated** and fixed-size
4. Use **slices** (`&arr[1..3]`) to reference parts of an array
5. For growable collections, use `Vec<T>` (coming in Week 2!)
