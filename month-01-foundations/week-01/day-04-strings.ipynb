# 🦀 Day 4: Strings — `String` vs `&str`

Today we tackle Rust's string types:
- `String` — owned, heap-allocated, growable
- `&str` — borrowed string slice
- String methods and operations

---

## 📖 Two String Types

| Feature | `String` | `&str` |
|---------|----------|--------|
| Storage | Heap | Anywhere |
| Ownership | Owned | Borrowed |
| Mutable | Yes (if `mut`) | No |
| Growable | Yes | No |

## 📝 String Literals (`&str`)

```rust
let greeting: &str = "Hello, World!";
println!("{}", greeting);

let name = "Rustacean"; // type inferred as &str
println!("Hello, {}!", name);
```

## 🏗️ Creating `String`

```rust
let s1 = String::from("Hello");
let s2 = "World".to_string();
let s3 = String::new(); // empty
let s4 = format!("{} {}", s1, s2);

println!("s1: {}", s1);
println!("s2: {}", s2);
println!("s3: '{}' (empty)", s3);
println!("s4: {}", s4);
```

## ✏️ Modifying Strings

```rust
let mut s = String::from("Hello");
s.push_str(", World"); // append &str
s.push('!');            // append single char
println!("{}", s);

// Removing
let last = s.pop();  // removes last char
println!("Popped: {:?}, Remaining: '{}'", last, s);

s.truncate(5);
println!("After truncate(5): '{}'", s);

s.clear();
println!("After clear: empty={}", s.is_empty());
```

## 🔗 Concatenation

```rust
// + operator (consumes left operand!)
let a = String::from("Hello");
let b = String::from(" World");
let c = a + &b; // a is MOVED
println!("{}", c);
// println!("{}", a); // ❌ a was moved

// format! (preferred — doesn't move anything)
let x = String::from("Hello");
let y = String::from("World");
let z = format!("{}, {}!", x, y);
println!("{}", z);
println!("x still works: {}", x); // ✅
```

## 🔍 Useful String Methods

```rust
let s = "Hello, Rust World!";

println!("len: {}", s.len());
println!("contains 'Rust': {}", s.contains("Rust"));
println!("starts_with 'Hello': {}", s.starts_with("Hello"));
println!("find 'Rust': {:?}", s.find("Rust"));
println!("to_uppercase: {}", s.to_uppercase());
println!("replace: {}", s.replace("Rust", "Beautiful"));
```

```rust
// Trimming
let padded = "  Hello  ";
println!("trim: '{}'", padded.trim());

// Splitting
let csv = "apple,banana,cherry";
let fruits: Vec<&str> = csv.split(',').collect();
println!("Fruits: {:?}", fruits);

let sentence = "  Rust   is    great  ";
let words: Vec<&str> = sentence.split_whitespace().collect();
println!("Words: {:?}", words);

// Joining
let parts = vec!["Rust", "is", "fast"];
println!("Joined: {}", parts.join(" "));
```

## 🔠 Strings Are UTF-8!

```rust
let emoji = "🦀";
let chinese = "你好";

println!("'{}' — {} bytes, {} chars", emoji, emoji.len(), emoji.chars().count());
println!("'{}' — {} bytes, {} chars", chinese, chinese.len(), chinese.chars().count());

// Can't index: let c = emoji[0]; ❌
// Use .chars().nth() instead:
let s = "Hello";
println!("3rd char: {:?}", s.chars().nth(2));
```

```rust
// String slicing (byte boundaries only!)
let s = "Hello, World!";
let hello = &s[0..5];
let world = &s[7..12];
println!("{} {}", hello, world);
```

## 🔄 Converting Between Types

```rust
// &str → String
let slice: &str = "hello";
let owned = slice.to_string();
let owned2 = String::from(slice);

// String → &str
let owned = String::from("hello");
let borrowed: &str = &owned;
let borrowed2 = owned.as_str();

println!("{} {} {}", owned, borrowed, borrowed2);
```

## 🎯 Key Takeaways

1. `String` is **owned**; `&str` is **borrowed**
2. Use `&str` for function params, `String` when you need ownership
3. Strings are **UTF-8** — no indexing with `s[0]`
4. Prefer `format!()` for concatenation
