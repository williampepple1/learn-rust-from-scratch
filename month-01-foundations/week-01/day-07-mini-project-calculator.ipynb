## 🧮 Step 1: Basic Operations

```rust
fn add(a: f64, b: f64) -> f64 { a + b }
fn subtract(a: f64, b: f64) -> f64 { a - b }
fn multiply(a: f64, b: f64) -> f64 { a * b }

fn divide(a: f64, b: f64) -> (bool, f64) {
    if b == 0.0 {
        (false, 0.0)  // Error: division by zero
    } else {
        (true, a / b)
    }
}

fn modulus(a: f64, b: f64) -> (bool, f64) {
    if b == 0.0 {
        (false, 0.0)
    } else {
        (true, a % b)
    }
}

fn power(base: f64, exp: f64) -> f64 { base.powf(exp) }
fn sqrt(x: f64) -> (bool, f64) {
    if x < 0.0 {
        (false, 0.0)
    } else {
        (true, x.sqrt())
    }
}

println!("Basic operations defined! ✅");
```

## 🖥️ Step 2: Display Engine

```rust
fn format_number(n: f64) -> String {
    if n == n.floor() {
        format!("{:.0}", n)
    } else {
        format!("{:.4}", n)
    }
}

fn display_result(a: f64, op: &str, b: f64, result: f64) {
    println!("  {} {} {} = {}",
        format_number(a), op, format_number(b), format_number(result));
}

fn display_header(title: &str) {
    let border = "═".repeat(40);
    println!("\n╔{}╗", border);
    println!("║{:^40}║", title);
    println!("╚{}╝", border);
}

display_header("🧮 Rust Calculator");
```

## 🚀 Step 3: Run the Calculator

```rust
fn calculate(a: f64, op: char, b: f64) {
    match op {
        '+' => display_result(a, "+", b, add(a, b)),
        '-' => display_result(a, "-", b, subtract(a, b)),
        '*' => display_result(a, "×", b, multiply(a, b)),
        '/' => {
            let (ok, result) = divide(a, b);
            if ok { display_result(a, "÷", b, result); }
            else  { println!("  ❌ Error: Division by zero!"); }
        }
        '%' => {
            let (ok, result) = modulus(a, b);
            if ok { display_result(a, "mod", b, result); }
            else  { println!("  ❌ Error: Modulo by zero!"); }
        }
        '^' => display_result(a, "^", b, power(a, b)),
        _ => println!("  ❌ Unknown operator: {}", op),
    }
}

display_header("🧮 Rust Calculator");
println!("\n📐 Basic Arithmetic:");
calculate(10.0, '+', 5.0);
calculate(10.0, '-', 5.0);
calculate(10.0, '*', 5.0);
calculate(10.0, '/', 5.0);
calculate(10.0, '%', 3.0);

println!("\n🔢 Powers:");
calculate(2.0, '^', 10.0);
calculate(3.0, '^', 3.0);

println!("\n⚠️ Edge Cases:");
calculate(10.0, '/', 0.0);
calculate(10.0, '%', 0.0);

println!("\n📏 Square Roots:");
for &val in &[4.0, 16.0, 144.0, 2.0] {
    let (ok, result) = sqrt(val);
    if ok {
        println!("  √{} = {}", format_number(val), format_number(result));
    }
}
```
