## ⚠️ The Result Type

```rust
enum Result<T, E> {
    Ok(T),   // Success with value T
    Err(E),  // Failure with error E
}
```

````rust
// Basic Result usage
fn divide(a: f64, b: f64) -> Result<f64, String> {
    if b == 0.0 {
        Err(String::from("Division by zero!"))
    } else {
        Ok(a / b)
    }
}

// Handle with match
match divide(10.0, 3.0) {
    Ok(result) => println!("10 / 3 = {:.4}", result),
    Err(e) => println!("Error: {}", e),
}

match divide(10.0, 0.0) {
    Ok(result) => println!("Result: {}", result),
    Err(e) => println!("Error: {}", e),
}
````

## 🧰 Result Methods

````rust
let ok: Result<i32, String> = Ok(42);
let err: Result<i32, String> = Err("error".into());

println!("is_ok: {}, is_err: {}", ok.is_ok(), err.is_err());

// unwrap variants
println!("unwrap_or: {}", err.unwrap_or(0));
println!("unwrap_or_else: {}", err.unwrap_or_else(|_| -1));

// map — transform the Ok value
let doubled = ok.map(|v| v * 2);
println!("map: {:?}", doubled);

// map_err — transform the Err value
let mapped_err = err.map_err(|e| format!("FATAL: {}", e));
println!("map_err: {:?}", mapped_err);
````

## 📝 Parsing & Real-World Results

````rust
// String parsing returns Result
let good: Result<i32, _> = "42".parse();
let bad: Result<i32, _> = "hello".parse();

println!("Parse '42': {:?}", good);
println!("Parse 'hello': {:?}", bad);

// Practical: parse multiple values safely
fn parse_point(input: &str) -> Result<(f64, f64), String> {
    let parts: Vec<&str> = input.split(',').collect();
    if parts.len() != 2 {
        return Err(format!("Expected 'x,y', got: {}", input));
    }
    let x: f64 = parts[0].trim().parse()
        .map_err(|_| format!("Invalid x: {}", parts[0]))?;
    let y: f64 = parts[1].trim().parse()
        .map_err(|_| format!("Invalid y: {}", parts[1]))?;
    Ok((x, y))
}

println!("{:?}", parse_point("3.5, 7.2"));
println!("{:?}", parse_point("abc, 7"));
println!("{:?}", parse_point("1,2,3"));
````

## ❓ The `?` Operator

Shorthand for: if `Err`, return it; if `Ok`, unwrap it.

````rust
fn calculate(input: &str) -> Result<f64, String> {
    let num: f64 = input.parse()
        .map_err(|_| format!("'{}' is not a number", input))?; // ? propagates error

    if num < 0.0 {
        return Err("Cannot sqrt negative number".into());
    }

    Ok(num.sqrt())
}

println!("calc('25'): {:?}", calculate("25"));
println!("calc('-4'): {:?}", calculate("-4"));
println!("calc('abc'): {:?}", calculate("abc"));
````

## 🆚 Option vs Result

| | `Option<T>` | `Result<T, E>` |
|---|---|---|
| Meaning | Value or nothing | Success or failure |
| Variants | `Some(T)` / `None` | `Ok(T)` / `Err(E)` |
| Error info | No | Yes (the `E` type) |
| Use when | Absence is normal | Failure needs explaining |

````rust
// Converting between them
let opt: Option<i32> = Some(42);
let res: Result<i32, &str> = opt.ok_or("was None");
println!("Option→Result: {:?}", res);

let res: Result<i32, &str> = Ok(42);
let opt: Option<i32> = res.ok();
println!("Result→Option: {:?}", opt);
````

## 🎯 Key Takeaways

1. `Result<T, E>` = `Ok(T)` for success, `Err(E)` for failure
2. The `?` operator propagates errors concisely
3. Use `Option` when absence is normal; `Result` when failure needs explanation
4. **Never use `.unwrap()` in production** — handle errors properly!
