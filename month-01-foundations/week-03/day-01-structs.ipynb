## 🏗️ Defining Structs

```rust
#[derive(Debug)]
struct Book {
    title: String,
    author: String,
    pages: u32,
    rating: f64,
    available: bool,
}

// Create an instance
let book = Book {
    title: String::from("The Rust Programming Language"),
    author: String::from("Steve Klabnik"),
    pages: 560,
    rating: 4.8,
    available: true,
};

println!("{:?}", book);
println!("Title: {}", book.title);
println!("Author: {}", book.author);
println!("Rating: {:.1}/5.0", book.rating);
```

```rust
// Mutable struct
let mut book = Book {
    title: String::from("Rust in Action"),
    author: String::from("Tim McNamara"),
    pages: 456,
    rating: 4.5,
    available: true,
};

// Modify fields
book.rating = 4.7;
book.available = false;
println!("Updated: {} — rating {}, available: {}", book.title, book.rating, book.available);
```

## ✨ Field Init Shorthand & Update Syntax

```rust
// Field init shorthand — when variable name matches field name
fn create_book(title: String, author: String, pages: u32) -> Book {
    Book {
        title,       // shorthand for title: title
        author,      // shorthand for author: author
        pages,       // shorthand for pages: pages
        rating: 0.0,
        available: true,
    }
}

let b = create_book("Zero to Production".into(), "Luca Palmieri".into(), 485);
println!("{} by {}", b.title, b.author);
```

```rust
// Struct update syntax (like spread in JS)
let book1 = Book {
    title: String::from("Programming Rust"),
    author: String::from("Jim Blandy"),
    pages: 738,
    rating: 4.6,
    available: true,
};

// Create book2 with some fields from book1
let book2 = Book {
    title: String::from("Programming Rust, 2nd Ed"),
    pages: 800,
    ..book1  // Take remaining fields from book1
};

println!("{} — {} pages", book2.title, book2.pages);
// Note: book1.title and book1.author were MOVED to book2!
// book1.title is no longer valid, but book1.pages, book1.rating still are
```

## 📚 Practical: Book Collection

```rust
let library = vec![
    Book { title: "The Rust Book".into(), author: "Klabnik".into(), pages: 560, rating: 4.8, available: true },
    Book { title: "Rust in Action".into(), author: "McNamara".into(), pages: 456, rating: 4.5, available: false },
    Book { title: "Programming Rust".into(), author: "Blandy".into(), pages: 738, rating: 4.6, available: true },
    Book { title: "Rust for Rustaceans".into(), author: "Gjengset".into(), pages: 282, rating: 4.9, available: true },
];

println!("📚 Library ({} books)\n", library.len());
for book in &library {
    let status = if book.available { "📗" } else { "📕" };
    println!("{} {} by {} ({} pp, ⭐{:.1})",
        status, book.title, book.author, book.pages, book.rating);
}

// Stats using iterators
let avg_rating: f64 = library.iter().map(|b| b.rating).sum::<f64>() / library.len() as f64;
let longest = library.iter().max_by_key(|b| b.pages).unwrap();
let available_count = library.iter().filter(|b| b.available).count();

println!("\nAvg rating: {:.1}", avg_rating);
println!("Longest: {} ({} pp)", longest.title, longest.pages);
println!("Available: {}/{}", available_count, library.len());
```

## 🎯 Key Takeaways

1. Structs group related data with named fields
2. Use `#[derive(Debug)]` for easy printing
3. **Field init shorthand** when variable name = field name
4. **Struct update syntax** (`..other`) copies remaining fields
5. The entire struct must be `mut` — can't have individual mutable fields
