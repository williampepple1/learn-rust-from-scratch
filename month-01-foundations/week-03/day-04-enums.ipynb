## 🏷️ Basic Enums

```rust
#[derive(Debug)]
enum Direction {
    North,
    South,
    East,
    West,
}

let dir = Direction::North;
println!("Direction: {:?}", dir);

// Match on enum variants
fn describe_direction(d: &Direction) -> &str {
    match d {
        Direction::North => "Going up ⬆️",
        Direction::South => "Going down ⬇️",
        Direction::East  => "Going right ➡️",
        Direction::West  => "Going left ⬅️",
    }
}

println!("{}", describe_direction(&dir));
println!("{}", describe_direction(&Direction::East));
```

## 📦 Enums with Data

Each variant can hold different data — this is what makes Rust enums so powerful!

```rust
#[derive(Debug)]
enum Shape {
    Circle(f64),                    // radius
    Rectangle(f64, f64),            // width, height
    Triangle(f64, f64, f64),        // three sides
}

impl Shape {
    fn area(&self) -> f64 {
        match self {
            Shape::Circle(r) => std::f64::consts::PI * r * r,
            Shape::Rectangle(w, h) => w * h,
            Shape::Triangle(a, b, c) => {
                let s = (a + b + c) / 2.0;
                (s * (s - a) * (s - b) * (s - c)).sqrt()
            }
        }
    }

    fn name(&self) -> &str {
        match self {
            Shape::Circle(_) => "Circle",
            Shape::Rectangle(_, _) => "Rectangle",
            Shape::Triangle(_, _, _) => "Triangle",
        }
    }
}

let shapes = vec![
    Shape::Circle(5.0),
    Shape::Rectangle(4.0, 6.0),
    Shape::Triangle(3.0, 4.0, 5.0),
];

for shape in &shapes {
    println!("{:>10}: area = {:.2}", shape.name(), shape.area());
}
```

```rust
// Variants with named fields (struct-like)
#[derive(Debug)]
enum Message {
    Quit,                                    // no data
    Echo(String),                            // one String
    Move { x: i32, y: i32 },                 // named fields
    Color(u8, u8, u8),                       // tuple data
}

fn process(msg: &Message) {
    match msg {
        Message::Quit => println!("Quitting!"),
        Message::Echo(text) => println!("Echo: {}", text),
        Message::Move { x, y } => println!("Moving to ({}, {})", x, y),
        Message::Color(r, g, b) => println!("Color: rgb({},{},{})", r, g, b),
    }
}

let messages = vec![
    Message::Echo("Hello!".to_string()),
    Message::Move { x: 10, y: 20 },
    Message::Color(255, 128, 0),
    Message::Quit,
];

for msg in &messages {
    process(msg);
}
```

## 🃏 Practical: Playing Cards

```rust
#[derive(Debug, Clone, Copy)]
enum Suit { Hearts, Diamonds, Clubs, Spades }

#[derive(Debug, Clone, Copy)]
enum Rank {
    Ace, Two, Three, Four, Five, Six, Seven,
    Eight, Nine, Ten, Jack, Queen, King,
}

#[derive(Debug)]
struct Card { rank: Rank, suit: Suit }

impl Card {
    fn value(&self) -> u8 {
        match self.rank {
            Rank::Ace => 11,
            Rank::Two => 2,   Rank::Three => 3,
            Rank::Four => 4,  Rank::Five => 5,
            Rank::Six => 6,   Rank::Seven => 7,
            Rank::Eight => 8, Rank::Nine => 9,
            Rank::Ten | Rank::Jack | Rank::Queen | Rank::King => 10,
        }
    }

    fn display(&self) -> String {
        let suit_symbol = match self.suit {
            Suit::Hearts => "♥",   Suit::Diamonds => "♦",
            Suit::Clubs => "♣",    Suit::Spades => "♠",
        };
        format!("{:?}{}", self.rank, suit_symbol)
    }
}

let hand = vec![
    Card { rank: Rank::Ace, suit: Suit::Spades },
    Card { rank: Rank::King, suit: Suit::Hearts },
    Card { rank: Rank::Seven, suit: Suit::Diamonds },
];

println!("🃏 Hand:");
let total: u8 = hand.iter().map(|c| {
    println!("  {} (value: {})", c.display(), c.value());
    c.value()
}).sum();
println!("  Total: {}", total);
```

## 🎯 Key Takeaways

1. Enums define a type with a fixed set of **variants**
2. Variants can hold **no data**, **tuple data**, or **named fields**
3. `match` on enums must be **exhaustive**
4. Enums + match = powerful, type-safe algebraic data types
