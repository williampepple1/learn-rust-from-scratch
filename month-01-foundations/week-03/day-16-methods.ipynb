# 🦀 Day 16: Methods & Associated Functions (`impl` blocks)

Today: adding behavior to your structs!
- `impl` blocks
- Methods (`&self`, `&mut self`, `self`)
- Associated functions (constructors)

---

```rust
#[derive(Debug)]
struct Rectangle {
    width: f64,
    height: f64,
}

impl Rectangle {
    // Associated function (constructor) — no self parameter
    fn new(width: f64, height: f64) -> Self {
        Rectangle { width, height }
    }

    fn square(size: f64) -> Self {
        Rectangle { width: size, height: size }
    }

    // Methods — take &self (immutable reference)
    fn area(&self) -> f64 {
        self.width * self.height
    }

    fn perimeter(&self) -> f64 {
        2.0 * (self.width + self.height)
    }

    fn is_square(&self) -> bool {
        (self.width - self.height).abs() < f64::EPSILON
    }

    fn can_hold(&self, other: &Rectangle) -> bool {
        self.width > other.width && self.height > other.height
    }

    // Method with &mut self — can modify the struct
    fn scale(&mut self, factor: f64) {
        self.width *= factor;
        self.height *= factor;
    }

    fn display(&self) {
        println!("Rectangle({:.1} x {:.1}) — area: {:.1}, perimeter: {:.1}",
            self.width, self.height, self.area(), self.perimeter());
    }
}

// Associated function called with ::  (like static method)
let r1 = Rectangle::new(10.0, 5.0);
let r2 = Rectangle::square(7.0);

r1.display();
r2.display();
println!("r1 is square: {}", r1.is_square());
println!("r2 is square: {}", r2.is_square());
println!("r1 can hold r2: {}", r1.can_hold(&r2));
```

```rust
// Mutable methods
let mut r = Rectangle::new(10.0, 5.0);
println!("Before scale:");
r.display();

r.scale(2.0);
println!("After scale(2.0):");
r.display();
```

## 📖 Method Receiver Types

| Receiver | Usage |
|----------|-------|
| `&self` | Read-only access (most common) |
| `&mut self` | Can modify the struct |
| `self` | Takes ownership (consumes the struct) |

## 📚 Practical: Book with Methods

```rust
#[derive(Debug)]
struct Book {
    title: String,
    author: String,
    pages: u32,
    rating: f64,
}

impl Book {
    fn new(title: &str, author: &str, pages: u32) -> Self {
        Book { title: title.into(), author: author.into(), pages, rating: 0.0 }
    }

    fn summary(&self) -> String {
        format!("'{}' by {} ({} pages, ⭐{:.1})", self.title, self.author, self.pages, self.rating)
    }

    fn is_long(&self) -> bool {
        self.pages > 500
    }

    fn rate(&mut self, rating: f64) {
        self.rating = rating.clamp(0.0, 5.0);
    }

    fn reading_time_hours(&self, pages_per_hour: u32) -> f64 {
        self.pages as f64 / pages_per_hour as f64
    }
}

let mut book = Book::new("The Rust Programming Language", "Steve Klabnik", 560);
book.rate(4.8);
println!("{}", book.summary());
println!("Long book: {}", book.is_long());
println!("Reading time: {:.1} hours", book.reading_time_hours(30));
```

## 🎯 Key Takeaways

1. `impl` blocks add methods and associated functions to structs
2. **Methods** take `self` as first param; **associated functions** don't
3. Use `Self` as return type for constructors
4. `&self` for reading, `&mut self` for modifying
