## 🛡️ The Option Type

Rust has **no null**. Instead, it uses `Option<T>`:
```rust
enum Option<T> {
    Some(T),  // There IS a value
    None,     // There is NO value
}
```

````rust
// Basic usage
let some_number: Option<i32> = Some(42);
let no_number: Option<i32> = None;

println!("some: {:?}", some_number);
println!("none: {:?}", no_number);
println!("is_some: {}, is_none: {}", some_number.is_some(), no_number.is_none());
````

````rust
// Matching on Option
fn describe(opt: Option<i32>) {
    match opt {
        Some(value) => println!("Got value: {}", value),
        None => println!("Got nothing!"),
    }
}

describe(Some(42));
describe(None);
````

## 🧰 Option Methods

````rust
let x: Option<i32> = Some(42);
let y: Option<i32> = None;

// unwrap — gets the value or PANICS
println!("unwrap: {}", x.unwrap());
// y.unwrap(); // ❌ PANIC!

// unwrap_or — provides a default
println!("unwrap_or: {}", y.unwrap_or(0));

// unwrap_or_else — lazy default
println!("unwrap_or_else: {}", y.unwrap_or_else(|| 100));

// expect — like unwrap but with custom message
println!("expect: {}", x.expect("should have a value"));
````

````rust
// map — transform the inner value
let x: Option<i32> = Some(5);
let doubled = x.map(|v| v * 2);
println!("map: {:?}", doubled); // Some(10)

let none: Option<i32> = None;
println!("map None: {:?}", none.map(|v| v * 2)); // None

// and_then (flatmap) — chain operations that return Option
fn try_divide(a: f64, b: f64) -> Option<f64> {
    if b == 0.0 { None } else { Some(a / b) }
}

let result = Some(10.0)
    .and_then(|x| try_divide(x, 3.0))
    .and_then(|x| try_divide(x, 0.0));  // Stops here
println!("and_then chain: {:?}", result); // None

// filter
let x = Some(4);
println!("filter even: {:?}", x.filter(|v| v % 2 == 0)); // Some(4)
println!("filter odd: {:?}", x.filter(|v| v % 2 != 0));  // None
````

## 📋 Practical Examples

````rust
// Safe division
fn safe_divide(a: f64, b: f64) -> Option<f64> {
    if b == 0.0 { None } else { Some(a / b) }
}

// Finding in a collection
fn find_user(users: &[&str], name: &str) -> Option<usize> {
    users.iter().position(|&u| u == name)
}

let users = ["Alice", "Bob", "Charlie"];

match find_user(&users, "Bob") {
    Some(index) => println!("Found Bob at index {}", index),
    None => println!("Bob not found"),
}

// if let for concise handling
if let Some(result) = safe_divide(10.0, 3.0) {
    println!("10 / 3 = {:.4}", result);
}
````

````rust
// Option in structs
#[derive(Debug)]
struct UserProfile {
    name: String,
    email: String,
    phone: Option<String>,      // Optional!
    nickname: Option<String>,   // Optional!
}

let user1 = UserProfile {
    name: "Alice".into(),
    email: "alice@mail.com".into(),
    phone: Some("+1-555-0101".into()),
    nickname: None,
};

println!("Name: {}", user1.name);
println!("Phone: {}", user1.phone.as_deref().unwrap_or("Not provided"));
println!("Nick: {}", user1.nickname.as_deref().unwrap_or("None"));
````

## 🎯 Key Takeaways

1. Rust has **no null** — use `Option<T>` instead
2. The compiler **forces** you to handle the `None` case
3. Use `.unwrap_or()` for defaults, `.map()` for transforms
4. `if let Some(x) = ...` for concise single-pattern matching
5. Use `Option<T>` in struct fields for optional data
