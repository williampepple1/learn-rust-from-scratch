## 📦 Tuple Structs

Like tuples but with a name — gives meaning to the type.

```rust
#[derive(Debug, Clone, Copy)]
struct Color(u8, u8, u8);

#[derive(Debug, Clone, Copy)]
struct Point2D(f64, f64);

#[derive(Debug, Clone, Copy)]
struct Point3D(f64, f64, f64);

let red = Color(255, 0, 0);
let origin = Point2D(0.0, 0.0);
let point = Point3D(1.0, 2.0, 3.0);

// Access with .0, .1, .2
println!("Red: rgb({}, {}, {})", red.0, red.1, red.2);
println!("Origin: ({}, {})", origin.0, origin.1);
println!("3D Point: ({}, {}, {})", point.0, point.1, point.2);
```

```rust
// Add methods to tuple structs
impl Point2D {
    fn distance_to(&self, other: &Point2D) -> f64 {
        ((self.0 - other.0).powi(2) + (self.1 - other.1).powi(2)).sqrt()
    }

    fn midpoint(&self, other: &Point2D) -> Point2D {
        Point2D((self.0 + other.0) / 2.0, (self.1 + other.1) / 2.0)
    }
}

let a = Point2D(0.0, 0.0);
let b = Point2D(3.0, 4.0);
println!("Distance: {}", a.distance_to(&b));
println!("Midpoint: {:?}", a.midpoint(&b));
```

### Type Safety with Tuple Structs

Even if two tuple structs have the same internal types, they're **different types**!

```rust
struct Meters(f64);
struct Kilograms(f64);

let distance = Meters(100.0);
let weight = Kilograms(75.0);

// Can't mix them up!
// let wrong: Meters = weight;  // ❌ ERROR: type mismatch

println!("Distance: {} m", distance.0);
println!("Weight: {} kg", weight.0);
```

## 📍 Unit Structs

Structs with no fields — used as markers or for trait implementations.

```rust
#[derive(Debug)]
struct Marker;

let m = Marker;
println!("Unit struct: {:?}", m);
println!("Size: {} bytes", std::mem::size_of::<Marker>()); // 0 bytes!
```

## 🎯 Project: 3D Vector Math

```rust
#[derive(Debug, Clone, Copy)]
struct Vec3 {
    x: f64,
    y: f64,
    z: f64,
}

impl Vec3 {
    fn new(x: f64, y: f64, z: f64) -> Self { Vec3 { x, y, z } }
    fn zero() -> Self { Vec3 { x: 0.0, y: 0.0, z: 0.0 } }

    fn magnitude(&self) -> f64 {
        (self.x * self.x + self.y * self.y + self.z * self.z).sqrt()
    }

    fn add(&self, other: &Vec3) -> Vec3 {
        Vec3::new(self.x + other.x, self.y + other.y, self.z + other.z)
    }

    fn scale(&self, factor: f64) -> Vec3 {
        Vec3::new(self.x * factor, self.y * factor, self.z * factor)
    }

    fn dot(&self, other: &Vec3) -> f64 {
        self.x * other.x + self.y * other.y + self.z * other.z
    }

    fn cross(&self, other: &Vec3) -> Vec3 {
        Vec3::new(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x,
        )
    }

    fn normalize(&self) -> Vec3 {
        let mag = self.magnitude();
        if mag > 0.0 { self.scale(1.0 / mag) } else { Vec3::zero() }
    }
}

let a = Vec3::new(1.0, 2.0, 3.0);
let b = Vec3::new(4.0, 5.0, 6.0);

println!("a = {:?}", a);
println!("b = {:?}", b);
println!("a + b = {:?}", a.add(&b));
println!("a · b = {}", a.dot(&b));
println!("a × b = {:?}", a.cross(&b));
println!("|a| = {:.4}", a.magnitude());
println!("â = {:?}", a.normalize());
```
