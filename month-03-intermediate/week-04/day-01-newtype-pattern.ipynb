# 🦀 Day 1: The Newtype Pattern
---

```rust
// Problem: raw primitives are ambiguous
fn send_email_bad(to: &str, subject: &str, body: &str) {
    println!("To: {}, Subject: {}", to, subject);
}
// send_email_bad("subject", "email@x.com", "body");  // Wrong order! Compiles fine!

// Solution: Newtype wrapper
struct Email(String);
struct Subject(String);
struct Body(String);

fn send_email(to: &Email, subject: &Subject, body: &Body) {
    println!("To: {}, Subject: {}", to.0, subject.0);
}

let email = Email("user@example.com".into());
let subject = Subject("Hello".into());
let body = Body("World".into());
send_email(&email, &subject, &body);
// send_email(&subject, &email, &body);  // ❌ Compile error! Type safety!
```

```rust
// Newtype for units
#[derive(Debug, Clone, Copy, PartialEq, PartialOrd)]
struct Meters(f64);
#[derive(Debug, Clone, Copy, PartialEq, PartialOrd)]
struct Kilometers(f64);

impl From<Kilometers> for Meters {
    fn from(km: Kilometers) -> Self { Meters(km.0 * 1000.0) }
}
impl From<Meters> for Kilometers {
    fn from(m: Meters) -> Self { Kilometers(m.0 / 1000.0) }
}
impl std::fmt::Display for Meters {
    fn fmt(&self, f: &mut std::fmt::Formatter) -> std::fmt::Result { write!(f, "{:.1}m", self.0) }
}
impl std::fmt::Display for Kilometers {
    fn fmt(&self, f: &mut std::fmt::Formatter) -> std::fmt::Result { write!(f, "{:.1}km", self.0) }
}

let distance = Kilometers(5.0);
let in_meters: Meters = distance.into();
println!("{} = {}", distance, in_meters);
// Can't accidentally add meters and kilometers without converting!
```

## 🎯 Key Takeaways
1. Newtype = `struct Foo(T)` — zero-cost wrapper
2. Prevents mixing up same-typed arguments
3. Add meaning to primitives (Email vs String, Meters vs f64)
4. Implement `From`/`Into` for conversions
5. No runtime overhead — compiler optimizes away the wrapper
