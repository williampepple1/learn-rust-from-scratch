# 🦀 Day 3: Type State Pattern
---

```rust
// Compile-time state machine — invalid states are unrepresentable!
use std::marker::PhantomData;

// State markers (zero-sized types)
struct Draft;
struct PendingReview;
struct Published;

struct Post<State> {
    content: String,
    _state: PhantomData<State>,
}

// Can only add content in Draft state
impl Post<Draft> {
    fn new() -> Self { Post { content: String::new(), _state: PhantomData } }
    fn add_text(&mut self, text: &str) { self.content.push_str(text); }
    fn submit(self) -> Post<PendingReview> {
        println!("📨 Submitted for review");
        Post { content: self.content, _state: PhantomData }
    }
}

// Can only approve in PendingReview state
impl Post<PendingReview> {
    fn approve(self) -> Post<Published> {
        println!("✅ Approved!");
        Post { content: self.content, _state: PhantomData }
    }
    fn reject(self) -> Post<Draft> {
        println!("❌ Rejected, back to draft");
        Post { content: self.content, _state: PhantomData }
    }
}

// Can only read content when Published
impl Post<Published> {
    fn content(&self) -> &str { &self.content }
}

let mut post = Post::new();
post.add_text("Hello, World!");
// post.content(); // ❌ Compile error! Can't read Draft content

let pending = post.submit();
// pending.add_text("more"); // ❌ Can't edit while pending!

let published = pending.approve();
println!("Content: {}", published.content());
```

## 🎯 Key Takeaways
1. Type states make **invalid states unrepresentable** at compile time
2. State transitions consume `self` and return new state
3. Methods only available in appropriate states
4. `PhantomData` holds the zero-sized state marker
5. Zero runtime cost — all checked at compile time
