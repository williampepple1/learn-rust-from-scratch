```rust
use std::fmt;

/// Abstract Syntax Tree for math expressions
#[derive(Debug, Clone)]
enum Expr {
    Num(f64),
    Add(Box<Expr>, Box<Expr>),
    Sub(Box<Expr>, Box<Expr>),
    Mul(Box<Expr>, Box<Expr>),
    Div(Box<Expr>, Box<Expr>),
    Neg(Box<Expr>),
}

impl Expr {
    fn num(n: f64) -> Self { Expr::Num(n) }
    fn add(a: Expr, b: Expr) -> Self { Expr::Add(Box::new(a), Box::new(b)) }
    fn sub(a: Expr, b: Expr) -> Self { Expr::Sub(Box::new(a), Box::new(b)) }
    fn mul(a: Expr, b: Expr) -> Self { Expr::Mul(Box::new(a), Box::new(b)) }
    fn div(a: Expr, b: Expr) -> Self { Expr::Div(Box::new(a), Box::new(b)) }
    fn neg(a: Expr) -> Self { Expr::Neg(Box::new(a)) }

    fn eval(&self) -> Result<f64, String> {
        match self {
            Expr::Num(n) => Ok(*n),
            Expr::Add(a, b) => Ok(a.eval()? + b.eval()?),
            Expr::Sub(a, b) => Ok(a.eval()? - b.eval()?),
            Expr::Mul(a, b) => Ok(a.eval()? * b.eval()?),
            Expr::Div(a, b) => {
                let divisor = b.eval()?;
                if divisor == 0.0 { Err("Division by zero".into()) }
                else { Ok(a.eval()? / divisor) }
            }
            Expr::Neg(a) => Ok(-a.eval()?),
        }
    }
}

impl fmt::Display for Expr {
    fn fmt(&self, f: &mut fmt::Formatter) -> fmt::Result {
        match self {
            Expr::Num(n) => write!(f, "{}", n),
            Expr::Add(a, b) => write!(f, "({} + {})", a, b),
            Expr::Sub(a, b) => write!(f, "({} - {})", a, b),
            Expr::Mul(a, b) => write!(f, "({} × {})", a, b),
            Expr::Div(a, b) => write!(f, "({} ÷ {})", a, b),
            Expr::Neg(a) => write!(f, "(-{})", a),
        }
    }
}

println!("Expression AST defined! ✅");
```

```rust
// Build and evaluate expressions
// (3 + 4) × 2
let expr1 = Expr::mul(
    Expr::add(Expr::num(3.0), Expr::num(4.0)),
    Expr::num(2.0),
);
println!("{} = {:?}", expr1, expr1.eval());

// (10 - 3) ÷ (1 + 1)
let expr2 = Expr::div(
    Expr::sub(Expr::num(10.0), Expr::num(3.0)),
    Expr::add(Expr::num(1.0), Expr::num(1.0)),
);
println!("{} = {:?}", expr2, expr2.eval());

// Division by zero
let bad = Expr::div(Expr::num(1.0), Expr::num(0.0));
println!("{} = {:?}", bad, bad.eval());
```

```rust
// Simple string parser: "3 + 4 * 2" → Expr
fn parse_simple(input: &str) -> Result<Expr, String> {
    let tokens: Vec<&str> = input.split_whitespace().collect();
    if tokens.is_empty() { return Err("Empty expression".into()); }

    // Parse first number
    let mut result = Expr::num(tokens[0].parse::<f64>()
        .map_err(|_| format!("Invalid number: {}", tokens[0]))?);

    // Process operator-number pairs (left-to-right, no precedence)
    let mut i = 1;
    while i < tokens.len() {
        if i + 1 >= tokens.len() { return Err("Missing operand".into()); }
        let op = tokens[i];
        let num = Expr::num(tokens[i + 1].parse::<f64>()
            .map_err(|_| format!("Invalid number: {}", tokens[i + 1]))?);
        result = match op {
            "+" => Expr::add(result, num),
            "-" => Expr::sub(result, num),
            "*" => Expr::mul(result, num),
            "/" => Expr::div(result, num),
            _ => return Err(format!("Unknown operator: {}", op)),
        };
        i += 2;
    }
    Ok(result)
}

for input in ["3 + 4", "10 - 3 * 2", "100 / 4 + 5", "abc"] {
    match parse_simple(input) {
        Ok(expr) => println!("'{}' → {} = {:?}", input, expr, expr.eval()),
        Err(e) => println!("'{}' → Error: {}", input, e),
    }
}
```

```rust
// Tests
assert_eq!(Expr::num(42.0).eval().unwrap(), 42.0);
assert_eq!(Expr::add(Expr::num(2.0), Expr::num(3.0)).eval().unwrap(), 5.0);
assert_eq!(Expr::neg(Expr::num(5.0)).eval().unwrap(), -5.0);
assert!(Expr::div(Expr::num(1.0), Expr::num(0.0)).eval().is_err());
assert!(parse_simple("").is_err());
assert!(parse_simple("abc").is_err());
assert_eq!(parse_simple("3 + 4").unwrap().eval().unwrap(), 7.0);

println!("\n🎉 All tests passed!");
println!("🎉 Month 3 Complete! Next: Month 4 — Async, Networking & Web!");
```

## 🎯 Key Takeaways
1. **Recursive enums** (`Box<Expr>`) naturally represent tree structures
2. **Pattern matching** elegantly handles evaluation
3. **Display trait** provides human-readable output
4. **Result propagation** (`?`) handles errors throughout parsing/eval
5. Combines: generics, traits, enums, Box, error handling, pattern matching
