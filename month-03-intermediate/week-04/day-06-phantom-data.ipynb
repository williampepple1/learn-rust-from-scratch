# 🦀 Day 6: PhantomData & Zero-Cost Abstractions
---

```rust
use std::marker::PhantomData;

// PhantomData — tells compiler a type is "used" even when it has no field
// Common use: type-level tags

struct Meters;
struct Feet;

#[derive(Debug, Clone, Copy)]
struct Distance<Unit> {
    value: f64,
    _unit: PhantomData<Unit>,
}

impl<Unit> Distance<Unit> {
    fn new(value: f64) -> Self {
        Distance { value, _unit: PhantomData }
    }
}

// Conversion only between specific units
impl Distance<Meters> {
    fn to_feet(self) -> Distance<Feet> {
        Distance::new(self.value * 3.28084)
    }
}

impl Distance<Feet> {
    fn to_meters(self) -> Distance<Meters> {
        Distance::new(self.value / 3.28084)
    }
}

let height = Distance::<Meters>::new(1.8);
let in_feet = height.to_feet();
println!("{:.1}m = {:.1}ft", height.value, in_feet.value);

// Can't add Meters + Feet!
// let bad = height + in_feet;  // ❌ Different types!

println!("\n💡 PhantomData has ZERO runtime cost!");
println!("   size_of::<Distance<Meters>>() = {}", std::mem::size_of::<Distance<Meters>>());
println!("   size_of::<f64>() = {}", std::mem::size_of::<f64>());
println!("   Same! The marker is zero-sized.");
```

```rust
// PhantomData for lifetime tracking
struct Slice<'a, T: 'a> {
    ptr: *const T,
    len: usize,
    _marker: PhantomData<&'a T>, // Tells compiler: this borrows data with lifetime 'a
}

println!("PhantomData used for:");
println!("  1. Type-level markers (units, states)");
println!("  2. Lifetime tracking in unsafe code");
println!("  3. Variance control");
println!("  4. All at zero runtime cost!");
```

## 🎯 Key Takeaways
1. `PhantomData<T>` marks a type parameter as "used" without storing data
2. **Zero-cost**: no runtime overhead, same memory layout
3. Use for unit systems, type-level tags, state machines
4. Essential for safe abstractions over unsafe code
5. Rust's zero-cost abstractions: safety without performance penalty
