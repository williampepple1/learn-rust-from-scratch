# 🦀 Day 2: Builder Pattern
---

```rust
#[derive(Debug)]
struct Server {
    host: String,
    port: u16,
    max_connections: usize,
    timeout_secs: u64,
    tls: bool,
}

struct ServerBuilder {
    host: String,
    port: u16,
    max_connections: usize,
    timeout_secs: u64,
    tls: bool,
}

impl ServerBuilder {
    fn new(host: &str, port: u16) -> Self {
        ServerBuilder {
            host: host.to_string(),
            port,
            max_connections: 100, // sensible defaults
            timeout_secs: 30,
            tls: false,
        }
    }
    fn max_connections(mut self, n: usize) -> Self { self.max_connections = n; self }
    fn timeout(mut self, secs: u64) -> Self { self.timeout_secs = secs; self }
    fn tls(mut self, enabled: bool) -> Self { self.tls = enabled; self }

    fn build(self) -> Result<Server, String> {
        if self.host.is_empty() { return Err("Host required".into()); }
        Ok(Server {
            host: self.host, port: self.port,
            max_connections: self.max_connections,
            timeout_secs: self.timeout_secs, tls: self.tls,
        })
    }
}

let server = ServerBuilder::new("localhost", 8080)
    .max_connections(1000)
    .timeout(60)
    .tls(true)
    .build()
    .unwrap();
println!("{:#?}", server);

// Minimal with defaults
let simple = ServerBuilder::new("0.0.0.0", 3000).build().unwrap();
println!("{:#?}", simple);
```

## 🎯 Key Takeaways
1. Builder separates construction from representation
2. Method chaining: each setter returns `self`
3. `build()` validates and produces the final struct
4. Sensible defaults reduce boilerplate for common cases
5. Better than many constructor parameters
