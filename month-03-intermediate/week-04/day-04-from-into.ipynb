# 🦀 Day 4: `From`/`Into` Conversions
---

```rust
// From/Into for ergonomic APIs
#[derive(Debug)]
struct Color { r: u8, g: u8, b: u8 }

impl From<(u8, u8, u8)> for Color {
    fn from(t: (u8, u8, u8)) -> Self { Color { r: t.0, g: t.1, b: t.2 } }
}
impl From<u32> for Color {
    fn from(hex: u32) -> Self {
        Color { r: ((hex >> 16) & 0xFF) as u8, g: ((hex >> 8) & 0xFF) as u8, b: (hex & 0xFF) as u8 }
    }
}
impl From<&str> for Color {
    fn from(s: &str) -> Self {
        match s {
            "red" => Color { r: 255, g: 0, b: 0 },
            "green" => Color { r: 0, g: 255, b: 0 },
            "blue" => Color { r: 0, g: 0, b: 255 },
            _ => Color { r: 0, g: 0, b: 0 },
        }
    }
}

// All these work!
let c1 = Color::from((255, 128, 0));
let c2: Color = 0xFF00FFu32.into();
let c3: Color = "red".into();
println!("{:?}, {:?}, {:?}", c1, c2, c3);

// Accept anything convertible
fn paint(color: impl Into<Color>) {
    let c = color.into();
    println!("Painting with {:?}", c);
}
paint((0, 255, 0));
paint(0x0000FFu32);
paint("green");
```

## 🎯 Key Takeaways
1. Implement `From<T>` and get `Into<U>` for free
2. `impl Into<T>` in function params accepts any convertible type
3. Multiple `From` impls = many ways to construct
4. Enables intuitive, Python-like flexibility with Rust's type safety
5. Common: `From<&str>`, `From<String>`, `From<(T,T)>`
