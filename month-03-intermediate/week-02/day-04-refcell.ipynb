# 🦀 Day 4: `RefCell<T>` — Interior Mutability
---

```rust
use std::cell::RefCell;

// RefCell checks borrowing rules at RUNTIME, not compile time
let data = RefCell::new(vec![1, 2, 3]);

// Immutable borrow
{
    let borrowed = data.borrow();
    println!("Borrowed: {:?}", *borrowed);
}

// Mutable borrow
{
    let mut borrowed = data.borrow_mut();
    borrowed.push(4);
}

println!("After mutation: {:?}", data.borrow());
```

```rust
// Interior mutability — mutate through an immutable reference!
struct Logger {
    messages: RefCell<Vec<String>>,
}

impl Logger {
    fn new() -> Self { Logger { messages: RefCell::new(Vec::new()) } }

    fn log(&self, msg: &str) {  // &self, not &mut self!
        self.messages.borrow_mut().push(msg.to_string());
    }

    fn print_all(&self) {
        for msg in self.messages.borrow().iter() {
            println!("  📝 {}", msg);
        }
    }
}

let logger = Logger::new();
logger.log("Started");
logger.log("Processing");
logger.log("Done");
logger.print_all();
```

```rust
// Runtime panic if borrow rules violated!
let data = RefCell::new(42);

// This would PANIC at runtime:
// let borrow1 = data.borrow();
// let borrow2 = data.borrow_mut();  // 💥 Already borrowed immutably!

// Safe: use try_borrow/try_borrow_mut
let borrow1 = data.borrow();
match data.try_borrow_mut() {
    Ok(_) => println!("Got mutable borrow"),
    Err(e) => println!("Can't borrow mutably: {}", e),
}
std::mem::drop(borrow1);
println!("After dropping immutable borrow, mutable works: {}", data.borrow());
```

## 🎯 Key Takeaways
1. `RefCell<T>` moves borrow checking to **runtime**
2. `.borrow()` → `Ref<T>` (immutable), `.borrow_mut()` → `RefMut<T>` (mutable)
3. **Panics** if rules violated at runtime
4. `try_borrow()`/`try_borrow_mut()` return `Result` instead
5. Enables the **interior mutability** pattern
