# 🦀 Day 2: `Deref` & `Drop` Traits
---

```rust
// Deref trait — allows using * operator
use std::ops::Deref;

struct MyBox<T>(T);

impl<T> MyBox<T> {
    fn new(x: T) -> MyBox<T> { MyBox(x) }
}

impl<T> Deref for MyBox<T> {
    type Target = T;
    fn deref(&self) -> &T { &self.0 }
}

let x = MyBox::new(5);
println!("*x = {}", *x);  // Uses Deref

// Deref coercion: &MyBox<String> → &String → &str
let s = MyBox::new(String::from("hello"));
fn hello(name: &str) { println!("Hello, {}!", name); }
hello(&s);  // MyBox<String> → &String → &str automatically!
```

```rust
// Drop trait — runs code when value goes out of scope
struct Resource {
    name: String,
}

impl Drop for Resource {
    fn drop(&mut self) {
        println!("  💧 Dropping resource: {}", self.name);
    }
}

println!("Creating resources:");
{
    let r1 = Resource { name: "Database".into() };
    let r2 = Resource { name: "FileHandle".into() };
    println!("  Using {} and {}", r1.name, r2.name);
} // Dropped in REVERSE order!
println!("Resources dropped!");
```

```rust
// Early drop with std::mem::drop
let r = Resource { name: "Connection".into() };
println!("Before early drop");
std::mem::drop(r);  // Drop NOW, not at end of scope
// println!("{}", r.name);  // ❌ already dropped
println!("After early drop");
```

## 🎯 Key Takeaways

1. `Deref` enables `*x` and automatic deref coercion
2. Deref coercion: `&Box<String>` → `&String` → `&str`
3. `Drop` runs cleanup code when values go out of scope
4. Drop order is **reverse** of creation order
5. Use `std::mem::drop()` for early cleanup
