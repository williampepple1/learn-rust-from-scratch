# 🦀 Day 3: `Rc<T>` — Reference Counting
---

```rust
use std::rc::Rc;

// Rc allows MULTIPLE OWNERS of the same data
let a = Rc::new(String::from("shared data"));
let b = Rc::clone(&a);  // Increments ref count (cheap!)
let c = Rc::clone(&a);

println!("a: {}", a);
println!("b: {}", b);
println!("c: {}", c);
println!("Reference count: {}", Rc::strong_count(&a));
```

```rust
// Shared ownership — like a graph with shared nodes
#[derive(Debug)]
enum SharedList {
    Cons(i32, Rc<SharedList>),
    Nil,
}
use SharedList::{Cons as SCons, Nil as SNil};

let shared_tail = Rc::new(SCons(3, Rc::new(SCons(4, Rc::new(SNil)))));

// Two lists share the same tail!
let list_a = SCons(1, Rc::clone(&shared_tail));
let list_b = SCons(2, Rc::clone(&shared_tail));

println!("list_a: {:?}", list_a);
println!("list_b: {:?}", list_b);
println!("Shared tail refs: {}", Rc::strong_count(&shared_tail));
```

```rust
// Rc is READ-ONLY — can't mutate through Rc directly
let data = Rc::new(vec![1, 2, 3]);
// data.push(4);  // ❌ Rc gives &T, not &mut T
println!("Data: {:?}", data);
println!("\n⚠️ Rc is immutable! For mutation, combine with RefCell (tomorrow)");
```

## 🎯 Key Takeaways
1. `Rc<T>` enables **multiple ownership** (ref-counted)
2. `Rc::clone()` is cheap — just increments a counter
3. Data is freed when **last** Rc is dropped (counter → 0)
4. `Rc` is **read-only** — combine with `RefCell` for mutation
5. `Rc` is **single-threaded** only — use `Arc` for threads
