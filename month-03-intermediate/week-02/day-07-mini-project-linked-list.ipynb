# 🦀 Day 7: Mini-Project — Singly Linked List

```rust
use std::fmt;

type Link<T> = Option<Box<Node<T>>>;

struct Node<T> {
    data: T,
    next: Link<T>,
}

pub struct LinkedList<T> {
    head: Link<T>,
    size: usize,
}

impl<T> LinkedList<T> {
    pub fn new() -> Self { LinkedList { head: None, size: 0 } }

    pub fn push_front(&mut self, data: T) {
        let new_node = Box::new(Node { data, next: self.head.take() });
        self.head = Some(new_node);
        self.size += 1;
    }

    pub fn pop_front(&mut self) -> Option<T> {
        self.head.take().map(|node| {
            self.head = node.next;
            self.size -= 1;
            node.data
        })
    }

    pub fn peek(&self) -> Option<&T> {
        self.head.as_ref().map(|node| &node.data)
    }

    pub fn len(&self) -> usize { self.size }
    pub fn is_empty(&self) -> bool { self.size == 0 }

    pub fn iter(&self) -> Iter<T> {
        Iter { current: self.head.as_deref() }
    }
}

// Iterator
pub struct Iter<'a, T> {
    current: Option<&'a Node<T>>,
}

impl<'a, T> Iterator for Iter<'a, T> {
    type Item = &'a T;
    fn next(&mut self) -> Option<&'a T> {
        self.current.map(|node| {
            self.current = node.next.as_deref();
            &node.data
        })
    }
}

impl<T: fmt::Display> fmt::Display for LinkedList<T> {
    fn fmt(&self, f: &mut fmt::Formatter) -> fmt::Result {
        let items: Vec<String> = self.iter().map(|x| format!("{}", x)).collect();
        write!(f, "[{}]", items.join(" → "))
    }
}

impl<T> Drop for LinkedList<T> {
    fn drop(&mut self) {
        while self.pop_front().is_some() {}
    }
}

println!("LinkedList<T> implemented! ✅");
```

```rust
// Test it!
let mut list = LinkedList::new();
for i in [5, 4, 3, 2, 1] { list.push_front(i); }

println!("List: {}", list);
println!("Length: {}", list.len());
println!("Peek: {:?}", list.peek());

println!("Pop: {:?}", list.pop_front());
println!("After pop: {}", list);

// Iterator usage
let sum: i32 = list.iter().sum();
println!("Sum: {}", sum);

let doubled: Vec<i32> = list.iter().map(|&x| x * 2).collect();
println!("Doubled: {:?}", doubled);
```

```rust
// Test with strings
let mut names = LinkedList::new();
for name in ["Charlie", "Bob", "Alice"] {
    names.push_front(name.to_string());
}
println!("Names: {}", names);

for name in names.iter() {
    println!("  Hello, {}!", name);
}

println!("\n✅ All tests passed!");
```

## 🎯 Key Takeaways
1. `Option<Box<Node>>` is the standard linked list pattern in Rust
2. `take()` moves Option's contents, leaving None — essential for node manipulation
3. Custom `Drop` prevents stack overflow on large lists
4. Implement `Iterator` for ergonomic `for` loop and chain support
5. Smart pointers make recursive data structures safe and ergonomic
