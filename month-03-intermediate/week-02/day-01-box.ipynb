# 🦀 Day 1: `Box<T>` — Heap Allocation & Recursive Types

---

```rust
// Box puts data on the heap
let b = Box::new(5);
println!("Boxed value: {}", b);
println!("Dereferenced: {}", *b);

// Useful for large data — moves pointer, not data
let big_array = Box::new([0u8; 1_000_000]);
println!("Box size on stack: {} bytes", std::mem::size_of_val(&big_array));
println!("Data size on heap: {} bytes", big_array.len());
```

```rust
// Recursive types REQUIRE Box (unknown size at compile time)

// Cons list (linked list from Lisp)
#[derive(Debug)]
enum List {
    Cons(i32, Box<List>),
    Nil,
}

use List::{Cons, Nil};

let list = Cons(1, Box::new(Cons(2, Box::new(Cons(3, Box::new(Nil))))));
println!("List: {:?}", list);
```

```rust
// Binary tree with Box
#[derive(Debug)]
struct TreeNode {
    value: i32,
    left: Option<Box<TreeNode>>,
    right: Option<Box<TreeNode>>,
}

impl TreeNode {
    fn leaf(value: i32) -> Self {
        TreeNode { value, left: None, right: None }
    }
    fn with_children(value: i32, left: TreeNode, right: TreeNode) -> Self {
        TreeNode { value, left: Some(Box::new(left)), right: Some(Box::new(right)) }
    }
    fn depth(&self) -> usize {
        let left_d = self.left.as_ref().map_or(0, |n| n.depth());
        let right_d = self.right.as_ref().map_or(0, |n| n.depth());
        1 + left_d.max(right_d)
    }
    fn sum(&self) -> i32 {
        let left_s = self.left.as_ref().map_or(0, |n| n.sum());
        let right_s = self.right.as_ref().map_or(0, |n| n.sum());
        self.value + left_s + right_s
    }
}

let tree = TreeNode::with_children(1,
    TreeNode::with_children(2, TreeNode::leaf(4), TreeNode::leaf(5)),
    TreeNode::leaf(3),
);
println!("Tree: {:?}", tree);
println!("Depth: {}, Sum: {}", tree.depth(), tree.sum());
```

```rust
// Box for trait objects (dynamic dispatch)
trait Animal {
    fn speak(&self) -> &str;
}
struct Dog; struct Cat;
impl Animal for Dog { fn speak(&self) -> &str { "Woof!" } }
impl Animal for Cat { fn speak(&self) -> &str { "Meow!" } }

let animals: Vec<Box<dyn Animal>> = vec![Box::new(Dog), Box::new(Cat), Box::new(Dog)];
for a in &animals {
    println!("  {}", a.speak());
}
```

## 🎯 Key Takeaways

1. `Box<T>` allocates on the **heap**, pointer on the stack
2. Essential for **recursive types** (unknown compile-time size)
3. Used for **trait objects** (`Box<dyn Trait>`)
4. Zero overhead — just a pointer (8 bytes on 64-bit)
5. Automatically cleaned up when dropped
