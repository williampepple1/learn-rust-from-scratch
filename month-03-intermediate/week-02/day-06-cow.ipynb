# 🦀 Day 6: `Cow<T>` — Clone on Write
---

```rust
use std::borrow::Cow;

// Cow<str>: borrows when possible, clones only when mutation needed
fn normalize(input: &str) -> Cow<str> {
    if input.contains(' ') {
        Cow::Owned(input.replace(' ', "_"))  // Must allocate
    } else {
        Cow::Borrowed(input)  // Zero-cost borrow
    }
}

let a = normalize("hello");      // Borrowed — no allocation
let b = normalize("hello world"); // Owned — allocated new string
println!("a: {} (borrowed={})", a, matches!(a, Cow::Borrowed(_)));
println!("b: {} (borrowed={})", b, matches!(b, Cow::Borrowed(_)));
```

```rust
// Practical: efficient string processing
fn escape_html(input: &str) -> Cow<str> {
    if input.contains('<') || input.contains('>') || input.contains('&') {
        let escaped = input
            .replace('&', "&amp;")
            .replace('<', "&lt;")
            .replace('>', "&gt;");
        Cow::Owned(escaped)
    } else {
        Cow::Borrowed(input) // No special chars — zero-cost!
    }
}

let texts = ["Hello World", "<script>alert('xss')</script>", "Normal text", "A & B"];
for text in &texts {
    let escaped = escape_html(text);
    let kind = if matches!(escaped, Cow::Borrowed(_)) { "borrowed" } else { "owned" };
    println!("  [{}] {} → {}", kind, text, escaped);
}
```

```rust
// Cow in function parameters — accept both owned and borrowed
fn log_message(msg: Cow<str>) {
    println!("LOG: {}", msg);
}

log_message(Cow::Borrowed("static message"));         // No allocation
log_message(Cow::Owned(format!("dynamic {}", 42)));    // Already owned

println!("\n💡 Cow avoids cloning when data doesn't need modification!");
```

## 🎯 Key Takeaways
1. `Cow<T>` = borrow when possible, clone only when needed
2. `Cow::Borrowed(&T)` = zero-cost reference
3. `Cow::Owned(T)` = allocated only for modifications
4. Perfect for string processing where most inputs pass through unchanged
5. Combines the efficiency of borrowing with the flexibility of ownership
