```rust
// Custom Range with step
struct StepRange {
    current: i32,
    end: i32,
    step: i32,
}

impl StepRange {
    fn new(start: i32, end: i32, step: i32) -> Self {
        StepRange { current: start, end, step }
    }
}

impl Iterator for StepRange {
    type Item = i32;
    fn next(&mut self) -> Option<i32> {
        if (self.step > 0 && self.current < self.end) ||
           (self.step < 0 && self.current > self.end) {
            let val = self.current;
            self.current += self.step;
            Some(val)
        } else {
            None
        }
    }
}

let by2: Vec<i32> = StepRange::new(0, 20, 2).collect();
println!("0 to 20 by 2: {:?}", by2);

let down: Vec<i32> = StepRange::new(10, 0, -3).collect();
println!("10 to 0 by -3: {:?}", down);
```

```rust
// Iterator that generates from a formula
struct Geometric {
    current: f64,
    ratio: f64,
}

impl Geometric {
    fn new(start: f64, ratio: f64) -> Self {
        Geometric { current: start, ratio }
    }
}

impl Iterator for Geometric {
    type Item = f64;
    fn next(&mut self) -> Option<f64> {
        let val = self.current;
        self.current *= self.ratio;
        Some(val)  // Infinite iterator!
    }
}

let powers_of_2: Vec<f64> = Geometric::new(1.0, 2.0).take(10).collect();
println!("Powers of 2: {:?}", powers_of_2);

let decay: Vec<f64> = Geometric::new(100.0, 0.5).take(8).collect();
println!("Half-life: {:?}", decay);
```

```rust
// Making your struct iterable with IntoIterator
#[derive(Debug)]
struct Matrix {
    data: Vec<Vec<f64>>,
    rows: usize,
    cols: usize,
}

impl Matrix {
    fn new(data: Vec<Vec<f64>>) -> Self {
        let rows = data.len();
        let cols = if rows > 0 { data[0].len() } else { 0 };
        Matrix { data, rows, cols }
    }

    fn iter(&self) -> MatrixIter {
        MatrixIter { matrix: self, row: 0, col: 0 }
    }

    fn row_iter(&self) -> impl Iterator<Item = &Vec<f64>> {
        self.data.iter()
    }
}

struct MatrixIter<'a> {
    matrix: &'a Matrix,
    row: usize,
    col: usize,
}

impl<'a> Iterator for MatrixIter<'a> {
    type Item = (usize, usize, f64);
    fn next(&mut self) -> Option<Self::Item> {
        if self.row >= self.matrix.rows { return None; }
        let val = self.matrix.data[self.row][self.col];
        let result = (self.row, self.col, val);
        self.col += 1;
        if self.col >= self.matrix.cols {
            self.col = 0;
            self.row += 1;
        }
        Some(result)
    }
}

let m = Matrix::new(vec![
    vec![1.0, 2.0, 3.0],
    vec![4.0, 5.0, 6.0],
]);

println!("All elements:");
for (r, c, val) in m.iter() {
    println!("  [{},{}] = {}", r, c, val);
}

let sum: f64 = m.iter().map(|(_, _, v)| v).sum();
println!("Sum: {}", sum);
```

## 🎯 Key Takeaways

2. Return `None` to signal the end of iteration
3. Infinite iterators are fine — use `.take(n)` to limit
4. `IntoIterator` makes types usable in `for` loops
