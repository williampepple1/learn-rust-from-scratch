```rust
// zip — combine two iterators element-wise
let names = vec!["Alice", "Bob", "Charlie"];
let scores = vec![92, 85, 98];

let combined: Vec<(&&str, &i32)> = names.iter().zip(scores.iter()).collect();
println!("Zipped: {:?}", combined);

for (name, score) in names.iter().zip(scores.iter()) {
    println!("  {} scored {}", name, score);
}
```

```rust
// enumerate — adds index to each element
let fruits = vec!["apple", "banana", "cherry"];
for (i, fruit) in fruits.iter().enumerate() {
    println!("  {}. {}", i + 1, fruit);
}
```

```rust
// flat_map — map + flatten
let sentences = vec!["hello world", "foo bar baz", "rust is great"];
let words: Vec<&str> = sentences.iter().flat_map(|s| s.split_whitespace()).collect();
println!("All words: {:?}", words);

// flatten — flatten nested iterators
let nested = vec![vec![1, 2], vec![3, 4], vec![5]];
let flat: Vec<&i32> = nested.iter().flatten().collect();
println!("Flattened: {:?}", flat);
```

```rust
// fold — reduce to a single value with accumulator
let numbers = vec![1, 2, 3, 4, 5];

let sum = numbers.iter().fold(0, |acc, &x| acc + x);
let product = numbers.iter().fold(1, |acc, &x| acc * x);
let max = numbers.iter().fold(i32::MIN, |acc, &x| acc.max(x));

println!("Sum: {}, Product: {}, Max: {}", sum, product, max);

// Build a string with fold
let csv = numbers.iter().fold(String::new(), |acc, x| {
    if acc.is_empty() { format!("{}", x) }
    else { format!("{},{}", acc, x) }
});
println!("CSV: {}", csv);
```

```rust
// scan — like fold but yields intermediate results
let numbers = vec![1, 2, 3, 4, 5];

let running_sum: Vec<i32> = numbers.iter()
    .scan(0, |state, &x| {
        *state += x;
        Some(*state)
    })
    .collect();
println!("Running sum: {:?}", running_sum);  // [1, 3, 6, 10, 15]
```

```rust
// Practical: grade processing pipeline
let raw_scores = vec![
    ("Alice", vec![85, 92, 78, 95]),
    ("Bob", vec![72, 68, 81, 75]),
    ("Charlie", vec![91, 95, 88, 97]),
];

let results: Vec<(&&str, f64, &str)> = raw_scores.iter()
    .map(|(name, scores)| {
        let avg = scores.iter().sum::<i32>() as f64 / scores.len() as f64;
        let grade = match avg as u32 {
            90..=100 => "A",
            80..=89 => "B",
            70..=79 => "C",
            _ => "F",
        };
        (name, avg, grade)
    })
    .collect();

for (name, avg, grade) in &results {
    println!("  {} — avg: {:.1} — grade: {}", name, avg, grade);
}
```

## 🎯 Key Takeaways

1. `zip` pairs elements from two iterators
2. `enumerate` adds indices
3. `flat_map` = map + flatten for nested structures
4. `fold` reduces to a single value (like reduce in JS)
5. `scan` is like fold but yields intermediate values
