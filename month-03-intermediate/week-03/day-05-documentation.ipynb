## 📖 Doc Comment Types
```rust
/// This documents the NEXT item (function, struct, etc.)
/// Supports **markdown**!
///
/// # Examples
/// ```
/// let result = my_function(42);
/// assert_eq!(result, 84);
/// ```
fn my_function(x: i32) -> i32 { x * 2 }

//! This documents the ENCLOSING item (module, crate)
//! Usually at the top of lib.rs
```

````rust
/// Calculates the factorial of a non-negative integer.
///
/// # Arguments
/// * `n` - The number to calculate the factorial of
///
/// # Returns
/// The factorial of `n` (n!)
///
/// # Panics
/// This function will overflow for large values of `n`.
///
/// # Examples
/// ```
/// assert_eq!(factorial(0), 1);
/// assert_eq!(factorial(5), 120);
/// ```
fn factorial(n: u64) -> u64 {
    (1..=n).product()
}

// Doc tests are run by cargo test automatically!
assert_eq!(factorial(0), 1);
assert_eq!(factorial(5), 120);
assert_eq!(factorial(10), 3628800);
println!("factorial docs + tests verified! ✅");
````

````rust
// Common doc sections:
//
// # Examples      — usage examples (run as tests!)
// # Panics        — when the function panics
// # Errors        — what errors can be returned
// # Safety        — unsafe code requirements
//
// cargo doc             — generate HTML docs
// cargo doc --open      — generate and open in browser
// cargo test --doc      — run doc tests only

println!("Doc sections: Examples, Panics, Errors, Safety");
println!("Commands: cargo doc --open, cargo test --doc");
````

````rust
/// A 2D point with x and y coordinates.
///
/// # Examples
/// ```
/// let p = Point2D::new(3.0, 4.0);
/// assert_eq!(p.distance_to_origin(), 5.0);
/// ```
#[derive(Debug, Clone, Copy, PartialEq)]
struct Point2D {
    /// The x coordinate
    x: f64,
    /// The y coordinate
    y: f64,
}

impl Point2D {
    /// Creates a new point.
    fn new(x: f64, y: f64) -> Self { Point2D { x, y } }

    /// Calculates the distance from the origin (0, 0).
    fn distance_to_origin(&self) -> f64 {
        (self.x.powi(2) + self.y.powi(2)).sqrt()
    }

    /// Calculates the distance to another point.
    ///
    /// # Examples
    /// ```
    /// let a = Point2D::new(0.0, 0.0);
    /// let b = Point2D::new(3.0, 4.0);
    /// assert_eq!(a.distance_to(&b), 5.0);
    /// ```
    fn distance_to(&self, other: &Point2D) -> f64 {
        ((self.x - other.x).powi(2) + (self.y - other.y).powi(2)).sqrt()
    }
}

let p = Point2D::new(3.0, 4.0);
assert_eq!(p.distance_to_origin(), 5.0);
println!("Documented Point2D works! ✅");
````

## 🎯 Key Takeaways
1. `///` documents the next item, `//!` documents the enclosing item
2. Use markdown in doc comments (headings, code blocks, lists)
3. `# Examples` code blocks are **run as tests** by `cargo test`
4. Include `# Panics`, `# Errors`, `# Safety` sections as needed
5. `cargo doc --open` generates beautiful HTML documentation
