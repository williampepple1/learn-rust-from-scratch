# 🦀 Day 4: Test-Driven Development (TDD)
---

## 🔴🟢🔵 The TDD Cycle
1. **🔴 Red** — Write a failing test
2. **🟢 Green** — Write minimal code to pass
3. **🔵 Refactor** — Clean up while tests pass

```rust
// Let's build a RomanNumeral converter using TDD!

// Step 1 🔴: Write the test first
// assert_eq!(to_roman(1), "I");

// Step 2 🟢: Minimal implementation
fn to_roman(mut n: u32) -> String {
    let mappings = [
        (1000, "M"), (900, "CM"), (500, "D"), (400, "CD"),
        (100, "C"), (90, "XC"), (50, "L"), (40, "XL"),
        (10, "X"), (9, "IX"), (5, "V"), (4, "IV"), (1, "I"),
    ];
    let mut result = String::new();
    for &(value, symbol) in &mappings {
        while n >= value {
            result.push_str(symbol);
            n -= value;
        }
    }
    result
}

// Step 3: Comprehensive tests
let tests = [
    (1, "I"), (4, "IV"), (9, "IX"), (14, "XIV"),
    (42, "XLII"), (99, "XCIX"), (399, "CCCXCIX"),
    (1994, "MCMXCIV"), (3999, "MMMCMXCIX"),
];

for (input, expected) in &tests {
    let result = to_roman(*input);
    assert_eq!(result, *expected, "{} should be {}, got {}", input, expected, result);
    println!("  ✅ {} = {}", input, result);
}
println!("\nAll TDD tests passed! ✅");
```

```rust
// TDD Round 2: from_roman
fn from_roman(s: &str) -> Result<u32, String> {
    let mut result = 0u32;
    let mut prev = 0u32;
    for c in s.chars().rev() {
        let val = match c {
            'I' => 1, 'V' => 5, 'X' => 10, 'L' => 50,
            'C' => 100, 'D' => 500, 'M' => 1000,
            _ => return Err(format!("Invalid character: {}", c)),
        };
        if val < prev { result -= val; } else { result += val; }
        prev = val;
    }
    Ok(result)
}

// Round-trip tests
for n in [1, 4, 9, 42, 99, 1994, 3999] {
    let roman = to_roman(n);
    let back = from_roman(&roman).unwrap();
    assert_eq!(n, back, "Round-trip failed: {} → {} → {}", n, roman, back);
}
assert!(from_roman("INVALID").is_err());
println!("Round-trip tests passed! ✅");
```

## 🎯 Key Takeaways
1. **Red → Green → Refactor** guides development
2. Write the **test first**, then the code
3. Each test should test **one thing**
4. Round-trip tests verify encode/decode consistency
5. TDD produces code that's testable **by design**
