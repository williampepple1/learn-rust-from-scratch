## ⏱️ Benchmarking with `criterion`
```toml
[dev-dependencies]
criterion = { version = "0.5", features = ["html_reports"] }

[[bench]]
name = "my_benchmark"
harness = false
```

```rust
// benches/my_benchmark.rs
use criterion::{black_box, criterion_group, criterion_main, Criterion};

fn fibonacci(n: u64) -> u64 {
    match n { 0 => 0, 1 => 1, _ => fibonacci(n-1) + fibonacci(n-2) }
}

fn bench_fib(c: &mut Criterion) {
    c.bench_function("fib 20", |b| b.iter(|| fibonacci(black_box(20))));
}

criterion_group!(benches, bench_fib);
criterion_main!(benches);
```
```bash
cargo bench  # Runs benchmarks and generates HTML report
```

````rust
// Simple timing in notebook
use std::time::Instant;

fn time_it<F: FnOnce() -> R, R>(name: &str, f: F) -> R {
    let start = Instant::now();
    let result = f();
    let elapsed = start.elapsed();
    println!("  {} took {:?}", name, elapsed);
    result
}

// Compare approaches
let n = 1_000_000;

time_it("Iterator sum", || {
    let _: i64 = (1..=n as i64).sum();
});

time_it("Manual loop", || {
    let mut sum: i64 = 0;
    for i in 1..=n as i64 { sum += i; }
    let _ = sum;
});

time_it("Formula", || {
    let n = n as i64;
    let _ = n * (n + 1) / 2;
});

println!("\n💡 In real benchmarks, use criterion for statistical accuracy!");
````

````rust
// Compare data structure operations
use std::collections::{HashMap, BTreeMap};

let n = 100_000;

time_it("Vec push", || {
    let mut v = Vec::new();
    for i in 0..n { v.push(i); }
});

time_it("Vec with_capacity push", || {
    let mut v = Vec::with_capacity(n);
    for i in 0..n { v.push(i); }
});

time_it("HashMap insert", || {
    let mut m = HashMap::new();
    for i in 0..n { m.insert(i, i); }
});

time_it("BTreeMap insert", || {
    let mut m = BTreeMap::new();
    for i in 0..n { m.insert(i, i); }
});
````

## 🎯 Key Takeaways
1. Use `criterion` crate for statistical benchmarks
2. `black_box()` prevents compiler from optimizing away computation
3. `cargo bench` runs benchmarks and generates reports
4. Pre-allocation (`with_capacity`) can dramatically improve perf
5. Always benchmark before and after optimizations
