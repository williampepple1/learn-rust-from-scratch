# 🦀 Day 1: Unit Tests
---

```rust
// Basic test functions
fn add(a: i32, b: i32) -> i32 { a + b }
fn is_even(n: i32) -> bool { n % 2 == 0 }
fn divide(a: f64, b: f64) -> Result<f64, String> {
    if b == 0.0 { Err("Division by zero".into()) } else { Ok(a / b) }
}

// assert! — check a boolean condition
assert!(is_even(4));
assert!(!is_even(3));
assert!(add(2, 3) > 0);

// assert_eq! — check equality
assert_eq!(add(2, 3), 5);
assert_eq!(add(-1, 1), 0);
assert_eq!("hello".len(), 5);

// assert_ne! — check inequality
assert_ne!(add(2, 3), 6);
assert_ne!("hello", "world");

println!("All assertions passed! ✅");
```

```rust
// Custom failure messages
assert!(add(2, 3) == 5, "Expected 2+3 to equal 5, got {}", add(2, 3));
assert_eq!(add(2, 3), 5, "Addition failed: {} + {} should be {}", 2, 3, 5);

// Testing Result values
assert!(divide(10.0, 2.0).is_ok());
assert_eq!(divide(10.0, 2.0).unwrap(), 5.0);
assert!(divide(10.0, 0.0).is_err());

println!("Result tests passed! ✅");
```

```rust
// In a real Rust project, tests go in a #[cfg(test)] module:
//
// #[cfg(test)]
// mod tests {
//     use super::*;
//
//     #[test]
//     fn test_add() {
//         assert_eq!(add(2, 3), 5);
//     }
//
//     #[test]
//     fn test_add_negative() {
//         assert_eq!(add(-1, 1), 0);
//     }
// }
//
// Run with: cargo test

println!("In real projects: #[test] fn test_name() {{ assertions }}");
println!("Run with: cargo test");
```

```rust
// Practical: test a struct
#[derive(Debug, PartialEq)]
struct Calculator {
    history: Vec<f64>,
}

impl Calculator {
    fn new() -> Self { Calculator { history: Vec::new() } }
    fn add(&mut self, a: f64, b: f64) -> f64 { let r = a + b; self.history.push(r); r }
    fn last(&self) -> Option<f64> { self.history.last().copied() }
}

// Tests
let mut calc = Calculator::new();
assert_eq!(calc.add(2.0, 3.0), 5.0);
assert_eq!(calc.last(), Some(5.0));
assert_eq!(calc.history.len(), 1);

let mut calc2 = Calculator::new();
assert_eq!(calc2.last(), None);

println!("Calculator tests passed! ✅");
```

## 🎯 Key Takeaways
1. `assert!(condition)` — fails if false
2. `assert_eq!(left, right)` — fails if not equal (shows both values)
3. `assert_ne!(left, right)` — fails if equal
4. Custom messages: `assert!(cond, "message {}", var)`
5. `#[test]` attribute marks test functions in `cargo test`
