# 🦀 Day 2: Testing Panics & Results
---

```rust
// #[should_panic] — test that code panics
// In real code:
// #[test]
// #[should_panic(expected = "out of range")]
// fn test_out_of_range() {
//     let v = vec![1,2,3];
//     v[10]; // Should panic
// }

// Simulating should_panic test
let result = std::panic::catch_unwind(|| {
    let v = vec![1, 2, 3];
    let _ = v[10];  // Panics!
});
assert!(result.is_err(), "Expected a panic!");
println!("✅ Panic was caught as expected");
```

```rust
// Testing Result-returning functions
fn parse_positive(s: &str) -> Result<u32, String> {
    let n: i32 = s.parse().map_err(|e| format!("Parse error: {}", e))?;
    if n <= 0 { return Err(format!("{} is not positive", n)); }
    Ok(n as u32)
}

// Test happy path
assert_eq!(parse_positive("42"), Ok(42));
assert_eq!(parse_positive("1"), Ok(1));

// Test error paths
assert!(parse_positive("abc").is_err());
assert!(parse_positive("-5").is_err());
assert!(parse_positive("0").is_err());

// Check specific error messages
let err = parse_positive("-3").unwrap_err();
assert!(err.contains("not positive"), "Expected 'not positive', got: {}", err);

println!("All Result tests passed! ✅");
```

```rust
// Test-returning Result pattern
// In real code:
// #[test]
// fn test_parse() -> Result<(), String> {
//     let result = parse_positive("42")?;
//     assert_eq!(result, 42);
//     Ok(())
// }

// Test edge cases thoroughly
let test_cases: Vec<(&str, Result<u32, &str>)> = vec![
    ("42", Ok(42)),
    ("0", Err("not positive")),
    ("-1", Err("not positive")),
    ("abc", Err("Parse error")),
    ("", Err("Parse error")),
];

for (input, expected_kind) in &test_cases {
    let result = parse_positive(input);
    match expected_kind {
        Ok(val) => assert_eq!(result, Ok(*val), "Failed for input: {}", input),
        Err(contains) => {
            assert!(result.is_err(), "Expected error for '{}'", input);
            let err_msg = result.unwrap_err();
            assert!(err_msg.contains(contains), "Error '{}' should contain '{}'", err_msg, contains);
        }
    }
}
println!("All edge case tests passed! ✅");
```

## 🎯 Key Takeaways
1. `#[should_panic]` tests that code panics
2. `#[should_panic(expected = "msg")]` checks panic message
3. Tests can return `Result<(), E>` to use `?` operator
4. Test **both** happy path AND error paths
5. Table-driven tests (test_cases vec) are great for comprehensive coverage
