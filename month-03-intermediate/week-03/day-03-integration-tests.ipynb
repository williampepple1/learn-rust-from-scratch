## 📁 Project Structure for Tests
```
my_crate/
├── src/
│   └── lib.rs          # Library code
├── tests/              # Integration tests
│   ├── csv_tests.rs    # Each file = separate test binary
│   └── api_tests.rs
└── Cargo.toml
```

- Unit tests: **inside** `src/` files with `#[cfg(test)]`
- Integration tests: **outside** in `tests/` directory
- Integration tests only access **public** API

````rust
// Simulating integration test structure

// This is like our library's public API
mod my_library {
    pub fn add(a: i32, b: i32) -> i32 { a + b }
    pub fn factorial(n: u64) -> u64 {
        (1..=n).product()
    }
    pub fn is_palindrome(s: &str) -> bool {
        let cleaned: String = s.chars().filter(|c| c.is_alphanumeric()).map(|c| c.to_lowercase().next().unwrap()).collect();
        cleaned == cleaned.chars().rev().collect::<String>()
    }
    // Private — not accessible from integration tests
    fn internal_helper() -> bool { true }
}

// Integration test — uses the library as an external user would
use my_library::*;

assert_eq!(add(2, 3), 5);
assert_eq!(factorial(5), 120);
assert_eq!(factorial(0), 1);
assert!(is_palindrome("racecar"));
assert!(is_palindrome("A man a plan a canal Panama"));
assert!(!is_palindrome("hello"));

println!("All integration tests passed! ✅");
````

````rust
// Test organization — grouping related tests
mod string_tests {
    use super::my_library::*;

    pub fn run() {
        assert!(is_palindrome(""));
        assert!(is_palindrome("a"));
        assert!(is_palindrome("aba"));
        assert!(!is_palindrome("abc"));
        println!("  ✅ String tests passed");
    }
}

mod math_tests {
    use super::my_library::*;

    pub fn run() {
        assert_eq!(add(0, 0), 0);
        assert_eq!(add(-1, 1), 0);
        assert_eq!(factorial(1), 1);
        assert_eq!(factorial(10), 3628800);
        println!("  ✅ Math tests passed");
    }
}

println!("Running test suites:");
string_tests::run();
math_tests::run();
println!("All suites passed! ✅");
````

## 🏗️ Cargo Test Commands
```bash
cargo test                      # Run all tests
cargo test test_name             # Run specific test
cargo test -- --nocapture        # Show println output
cargo test --test csv_tests      # Run specific integration test file
cargo test -- --test-threads=1   # Run tests sequentially
```

## 🎯 Key Takeaways
1. Integration tests go in `tests/` directory
2. Each file in `tests/` is a **separate crate** — only sees public API
3. Shared test utilities: `tests/common/mod.rs`
4. `cargo test` runs both unit and integration tests
5. Use `#[ignore]` for slow tests, run with `cargo test -- --ignored`
