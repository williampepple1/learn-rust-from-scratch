# 🦀 Day 2: Terminal UI — crossterm & ratatui
---

Today we explore **terminal manipulation** and **TUI frameworks** in Rust.

- **crossterm**: Cross-platform terminal manipulation (raw mode, cursor, colors)
- **Styled output**: Colors, bold, underline with `style::Stylize`
- **ratatui**: Terminal UI library for building dashboards and interactive apps
- **Widgets**: Paragraph, List, Table, Gauge, Block
- **Layout**: Constraint-based layout system

```rust
:dep crossterm = "0.27"
```

```rust
use crossterm::style::{Color, Stylize, Attribute};

// Styled output — works in EvCxR!
println!("{}", "Bold text".bold());
println!("{}", "Red text".red());
println!("{}", "Green on black".green().on_dark_red());
println!("{}", "Underlined".attribute(Attribute::Underlined));
println!("{}", "Combined".bold().cyan());
```

## ratatui — Terminal UI Framework

ratatui builds on crossterm to create full TUI applications. It uses:
- **Widgets**: Paragraph, List, Table, Gauge, Block, Sparkline
- **Layout**: `Layout::default().direction(Direction::Vertical).constraints([...])`
- **Frame**: Render to a terminal buffer

**Note**: ratatui requires an event loop and terminal setup — it's meant for `cargo run` binaries, not EvCxR. Below is the structure of a typical ratatui app.

```rust
// ratatui dashboard example — Cargo binary structure
// Add to Cargo.toml: ratatui = "0.26", crossterm = "0.27"
//
// use ratatui::{
//     prelude::*,
//     widgets::{Block, Borders, Paragraph, List, ListItem, Gauge, Table, Row, Cell},
// };
//
// fn ui(frame: &mut Frame) {
//     let chunks = Layout::default()
//         .direction(Direction::Vertical)
//         .constraints([
//             Constraint::Length(3),
//             Constraint::Min(0),
//             Constraint::Length(3),
//         ])
//         .split(frame.area());
//
//     let title = Paragraph::new("Dashboard")
//         .block(Block::default().borders(Borders::ALL).title("Status"));
//     frame.render_widget(title, chunks[0]);
//
//     let items = vec![
//         ListItem::new("CPU: 45%"),
//         ListItem::new("Memory: 2.1GB / 8GB"),
//         ListItem::new("Disk: 120GB free"),
//     ];
//     let list = List::new(items).block(Block::default().borders(Borders::ALL));
//     frame.render_widget(list, chunks[1]);
//
//     let gauge = Gauge::default().gauge_style(Style::default().fg(Color::Green)).ratio(0.65);
//     frame.render_widget(gauge, chunks[2]);
// }
//
// In main(): enable_raw_mode(), enter_alternate_screen(), loop { terminal.draw(ui)?; }

println!("ratatui widgets: Paragraph, List, Table, Gauge, Block");
println!("Layout::constraints: Length(n), Min(n), Max(n), Percentage(n)");
```

## 🎯 Key Takeaways

1. **crossterm** provides cross-platform terminal manipulation (colors, cursor, raw mode)
2. `style::Stylize` trait adds `.red()`, `.bold()`, `.on_dark_blue()` etc. to strings
3. **ratatui** builds TUI apps with widgets and constraint-based layouts
4. Widgets: Paragraph, List, Table, Gauge, Block with borders and titles
5. Layout uses `Constraint::Length`, `Min`, `Max`, `Percentage`
6. Full TUI apps need `cargo run` — EvCxR can't run interactive event loops
