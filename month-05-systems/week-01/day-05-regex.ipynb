# 🦀 Day 5: Regular Expressions — regex crate
---

Today we explore **pattern matching** for strings with the regex crate.

- **Regex::new()**: Compile patterns (do once, use many times)
- **is_match(), find(), captures()**: Different levels of matching
- **Named capture groups**: `(?P<name>...)`
- **find_iter(), replace(), replace_all()**: Iterate and replace
- **RegexSet**: Match multiple patterns efficiently
- **Performance**: Compile once, reuse

```rust
:dep regex = "1"
```

```rust
use regex::Regex;

let re = Regex::new(r"\d{3}-\d{3}-\d{4}").unwrap();
let text = "Call 555-123-4567 or 555-987-6543";

println!("is_match: {}", re.is_match(text));
if let Some(m) = re.find(text) {
    println!("first match: '{}'", m.as_str());
}
for m in re.find_iter(text) {
    println!("match: '{}' at {}", m.as_str(), m.start());
}
```

```rust
// Capture groups
let re = Regex::new(r"(\d{3})-(\d{3})-(\d{4})").unwrap();
let text = "Call 555-123-4567";

if let Some(caps) = re.captures(text) {
    println!("Full: {}", caps.get(0).unwrap().as_str());
    println!("Area: {}", caps.get(1).unwrap().as_str());
    println!("Prefix: {}", caps.get(2).unwrap().as_str());
    println!("Line: {}", caps.get(3).unwrap().as_str());
}
```

```rust
// Named capture groups: (?P<name>...)
let re = Regex::new(r"(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})").unwrap();
let text = "Date: 2025-03-08";

if let Some(caps) = re.captures(text) {
    println!("Year: {}", &caps["year"]);
    println!("Month: {}", &caps["month"]);
    println!("Day: {}", &caps["day"]);
}
```

```rust
// replace and replace_all
let re = Regex::new(r"\b(foo|bar)\b").unwrap();
let text = "foo and bar went to the foo bar";

let replaced = re.replace_all(text, "***");
println!("{}", replaced);

// With replacement that uses capture groups
let re2 = Regex::new(r"(\w+) (\w+)").unwrap();
let swapped = re2.replace_all("hello world", "$2 $1");
println!("{}", swapped);
```

```rust
// RegexSet: match multiple patterns at once
use regex::RegexSet;

let set = RegexSet::new(&[
    r"\bERROR\b",
    r"\bWARN\b",
    r"\bINFO\b",
]).unwrap();

let lines = ["[ERROR] crash", "[WARN] slow", "[INFO] ok", "[DEBUG] trace"];
for line in lines {
    let matches: Vec<_> = set.matches(line).into_iter().collect();
    if !matches.is_empty() {
        println!("{} -> matched indices: {:?}", line, matches);
    }
}
```

## 🎯 Key Takeaways

1. **Regex::new()** compiles a pattern — do once, reuse for performance
2. **is_match()** for boolean; **find()** for first match; **captures()** for groups
3. Named captures `(?P<name>...)` enable `&caps["name"]`
4. **find_iter()** for all matches; **replace_all()** for substitutions
5. **RegexSet** matches multiple patterns in one pass
6. Use raw strings `r"..."` to avoid escaping backslashes
