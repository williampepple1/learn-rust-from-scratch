# 🦀 Day 1: Declarative Macros — macro_rules! Basics
---

Today we explore **declarative macros** — code that writes code (metaprogramming).

- **What macros are**: Code that generates code at compile time
- **macro_rules!** syntax: Matchers and transcribers
- **Fragment specifiers**: `expr`, `ident`, `ty`, `literal`, `tt`
- **Building macros**: From simple `say_hello!()` to `my_vec![]`
- **Built-in macros**: `stringify!`, `concat!`
- **Hygiene**: How macro-generated identifiers behave

```rust
// Simple macros: matcher => transcriber
macro_rules! say_hello {
    () => {
        println!("Hello from macro!");
    };
}

macro_rules! create_function {
    ($func_name:ident) => {
        fn $func_name() {
            println!("Called {:?}", stringify!($func_name));
        }
    };
}

say_hello!();
create_function!(my_special_fn);
my_special_fn();
```

```rust
// Fragment specifiers: expr, ident, ty, literal, tt
macro_rules! print_expr {
    ($e:expr) => { println!("expr = {:?}", $e) };
}

macro_rules! declare_var {
    ($name:ident : $t:ty = $v:expr) => {
        let $name: $t = $v;
    };
}

print_expr!(2 + 3);
declare_var!(x: i32 = 42);
println!("x = {}", x);
print_expr!(stringify!(hello));
```

```rust
// Building my_vec! like the standard vec! — multiple match arms
macro_rules! my_vec {
    () => { Vec::new() };
    ($($x:expr),* $(,)?) => {
        {
            let mut v = Vec::new();
            $( v.push($x); )*
            v
        }
    };
}

let empty: Vec<i32> = my_vec![];
let filled = my_vec![10, 20, 30];
println!("empty: {:?}, filled: {:?}", empty, filled);
```

```rust
// stringify! and concat! built-in macros
let s = stringify!(fn foo() { 1 + 2 });
println!("stringify: {}", s);

let c = concat!("Hello", ", ", "Rust", "!");
println!("concat: {}", c);

// Hygiene: macro-generated identifiers are "hygienic"
macro_rules! hygienic_example {
    () => {
        let x = 100;  // This x is in a different "hygienic" scope
        println!("Inside macro: {}", x);
    };
}

let x = 42;
hygienic_example!();
println!("Outside macro: {}", x);  // Still 42, not 100
```

## 🎯 Key Takeaways

1. Macros are **metaprogramming** — code that generates code at compile time
2. `macro_rules! name { (matcher) => { transcriber }; }` is the basic syntax
3. Fragment specifiers: `$x:expr`, `$x:ident`, `$x:ty`, `$x:literal`, `$x:tt`
4. Repetition: `$($x:expr),*` matches zero or more; `$($x:expr),+` matches one or more
5. `stringify!` turns tokens into a string; `concat!` concatenates literals
6. Macros are **hygienic** — they don't accidentally capture or shadow identifiers
