# 🦀 Day 6: Link-time optimization & binary size
---

Today we tune **release builds** — LTO, codegen-units, and binary size.

- **LTO**: Optimize across crate boundaries (fat vs thin)
- **codegen-units**: 1 = slower compile, better optimization
- **opt-level**: 0, 1, 2, 3, "s" (size), "z" (min size)
- **strip = true**: Remove debug symbols
- **panic = "abort"**: Smaller binary, no unwind tables
- **cargo-bloat**: Find what's taking space
- **#[cfg]**: Conditional compilation for dead code elimination

## Cargo.toml [profile.release] settings

```toml
[profile.release]
opt-level = 3          # max speed (default)
lto = "thin"          # or true for fat LTO (slower link, smaller)
codegen-units = 1      # better optimization, slower compile
strip = true           # remove debug symbols
panic = "abort"       # smaller binary, no unwind
```

For **minimum size**:
```toml
opt-level = "z"
lto = true
codegen-units = 1
strip = true
panic = "abort"
```

````rust
// #[cfg] for conditional compilation — dead code elimination
#[cfg(target_os = "windows")]
fn platform_name() -> &'static str {
    "Windows"
}

#[cfg(not(target_os = "windows"))]
fn platform_name() -> &'static str {
    "Not Windows"
}

#[cfg(feature = "extra")]
fn extra_feature() {
    println!("Extra feature enabled");
}

println!("Platform: {}", platform_name());
````

````rust
// Measuring binary size (run in terminal):
// cargo build --release
// ls -l target/release/your_binary   # Linux/Mac
// dir target\\release\\your_binary.exe  # Windows

// cargo-bloat: cargo install cargo-bloat
// cargo bloat --release -n 20  # top 20 crates by size

println!("Build with 'cargo build --release' and inspect binary size");
println!("Use 'cargo bloat --release' to see per-crate size breakdown");
````

## 🎯 Key Takeaways

1. **LTO** (link-time optimization) optimizes across crates; thin is faster, fat is smaller
2. **codegen-units = 1** improves optimization at the cost of compile time
3. **opt-level "z"** optimizes for minimum size; "s" is a balance
4. **strip = true** removes debug symbols, reducing binary size
5. **panic = "abort"** eliminates unwind tables for smaller binaries
6. **cargo-bloat** identifies which crates contribute most to size
