Today we dive into **criterion** — statistical benchmarking for Rust.

- **Why microbenchmarks are tricky**: Warmup, outliers, compiler optimization eliminating code
- **criterion basics**: `Criterion::benchmark_group`, `b.iter()`, `black_box()`
- **Parameterized benchmarks**: `BenchmarkGroup::bench_with_input`
- **Throughput**: `Throughput::Bytes`, `Throughput::Elements`
- **Statistical analysis**: Confidence intervals, t-tests for comparing implementations
- **HTML reports**: criterion generates beautiful reports in `target/criterion/`

````rust
// Manual benchmarking in EvCxR using Instant (like Day 1)
use std::time::Instant;

fn bubble_sort(arr: &mut [i32]) {
    let n = arr.len();
    for i in 0..n {
        for j in 0..n - i - 1 {
            if arr[j] > arr[j + 1] {
                arr.swap(j, j + 1);
            }
        }
    }
}

fn insertion_sort(arr: &mut [i32]) {
    for i in 1..arr.len() {
        let mut j = i;
        while j > 0 && arr[j - 1] > arr[j] {
            arr.swap(j - 1, j);
            j -= 1;
        }
    }
}

let mut data: Vec<i32> = (0..500).rev().collect();
let iterations = 100;

let mut bubble_data = data.clone();
let start = Instant::now();
for _ in 0..iterations {
    bubble_data.copy_from_slice(&data);
    bubble_sort(&mut bubble_data);
}
println!("Bubble sort avg: {:?}", start.elapsed() / iterations);

let mut insert_data = data.clone();
let start = Instant::now();
for _ in 0..iterations {
    insert_data.copy_from_slice(&data);
    insertion_sort(&mut insert_data);
}
println!("Insertion sort avg: {:?}", start.elapsed() / iterations);
````

````rust
// black_box prevents the optimizer from eliminating "unused" code
use std::hint::black_box;

fn compute(x: i32) -> i32 {
    x * x + 1
}

// Without black_box, compiler might optimize away the loop
let start = std::time::Instant::now();
for i in 0..1_000_000 {
    black_box(compute(black_box(i)));
}
println!("With black_box: {:?}", start.elapsed());
````

## 🎯 Key Takeaways

1. **criterion** provides statistical benchmarking with confidence intervals
2. **black_box()** prevents the optimizer from eliminating benchmarked code
3. **Parameterized benchmarks** with `bench_with_input` test multiple input sizes
4. **Throughput** (Bytes/Elements) reports ops per second for data-processing benchmarks
5. **HTML reports** in `target/criterion/` visualize results
6. For EvCxR, use `Instant` + multiple iterations as a lightweight alternative
